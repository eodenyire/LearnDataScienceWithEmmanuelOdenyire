# Week 5: Data Cleaning and Transformation

## Learning Objectives

```python
import pandas as pd
import numpy as np
```

## 1. Handling Missing Data

```python
data = {
    'name':   ['Alice', 'Bob', None, 'Diana', 'Eve', 'Frank'],
    'age':    [28, None, 23, 31, 27, None],
    'score':  [85.0, 92.0, 78.0, None, 88.0, 76.0],
    'city':   ['NYC', 'LA', 'Chicago', 'NYC', None, 'LA']
}
df = pd.DataFrame(data)
print(df)
print('\nMissing values:\n', df.isnull().sum())
```

```python
# Filling missing values
df['age'] = df['age'].fillna(df['age'].median())
df['score'] = df['score'].fillna(df['score'].mean())
df['name'] = df['name'].fillna('Unknown')
df['city'] = df['city'].fillna('Unknown')
print('After filling:\n', df)
```

## 2. Removing Duplicates

```python
df2 = pd.DataFrame({
    'id':     [1, 2, 2, 3, 4, 4, 5],
    'value':  [10, 20, 20, 30, 40, 40, 50]
})
print('With duplicates:', len(df2), 'rows')
df2 = df2.drop_duplicates()
print('Without duplicates:', len(df2), 'rows')
```

## 3. Renaming and Reindexing

```python
df = df.rename(columns={'name': 'employee_name', 'score': 'performance_score'})
df = df.reset_index(drop=True)
print(df.head())
```

## 4. Data Transformation

```python
# String transformations
df['employee_name'] = df['employee_name'].str.upper()
df['city'] = df['city'].str.strip()

# Map values
df['age_group'] = pd.cut(df['age'],
                          bins=[0, 25, 30, 100],
                          labels=['Young', 'Mid', 'Senior'])

# Apply custom function
df['grade'] = df['performance_score'].apply(
    lambda s: 'Excellent' if s >= 90 else ('Good' if s >= 80 else 'Average'))

print(df)
```

## 5. Merging DataFrames

```python
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name':   ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept_id':[10, 20, 10, 30]
})
departments = pd.DataFrame({
    'dept_id':  [10, 20, 30],
    'dept_name':['Engineering', 'Marketing', 'HR']
})
merged = pd.merge(employees, departments, on='dept_id', how='left')
print(merged)
```
