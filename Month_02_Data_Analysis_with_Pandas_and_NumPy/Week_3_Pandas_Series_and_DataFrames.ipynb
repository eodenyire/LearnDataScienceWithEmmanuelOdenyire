```python
import pandas as pd
import numpy as np

print(f'Pandas version: {pd.__version__}')
```

## 1. Pandas Series

```python
# Create a Series from a list
sales = pd.Series([250, 320, 180, 420, 310],
                  index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
                  name='Daily Sales ($)')
print(sales)
print(f'\nMean: {sales.mean():.1f}, Max: {sales.max()}')
```

## 2. Creating DataFrames

```python
data = {
    'name':       ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age':        [28, 34, 23, 31, 27],
    'department': ['Engineering', 'Marketing', 'Engineering', 'HR', 'Marketing'],
    'salary':     [75000, 60000, 70000, 65000, 62000],
    'years_exp':  [4, 8, 2, 6, 3]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Basic info
print(df.info())
print('\nShape:', df.shape)
print('\nDescribe:\n', df.describe())
```

## 3. Selecting Data

```python
# Select column(s)
print(df['name'])
print('\n', df[['name', 'salary']])

# loc: label-based
print('\nRow 2:', df.loc[2])

# iloc: integer-based
print('\nFirst 2 rows, cols 0-2:\n', df.iloc[:2, :3])
```

## 4. Filtering DataFrames

```python
# Boolean filter
eng = df[df['department'] == 'Engineering']
print('Engineers:\n', eng)

# Multiple conditions
senior = df[(df['years_exp'] >= 4) & (df['salary'] >= 65000)]
print('\nSenior & well-paid:\n', senior[['name', 'years_exp', 'salary']])
```

## 5. Modifying DataFrames

```python
# Add a new column
df['bonus'] = df['salary'] * 0.10
df['senior'] = df['years_exp'] >= 5
print(df[['name', 'salary', 'bonus', 'senior']])

# Apply a function
df['salary_band'] = df['salary'].apply(
    lambda x: 'High' if x >= 70000 else ('Mid' if x >= 60000 else 'Low'))
print('\nSalary bands:\n', df[['name', 'salary', 'salary_band']])
```
