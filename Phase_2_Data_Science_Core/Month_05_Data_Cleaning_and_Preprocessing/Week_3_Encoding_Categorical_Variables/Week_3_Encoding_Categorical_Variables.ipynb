# Week 3: Encoding Categorical Variables

## Learning Objectives
- Apply label encoding and ordinal encoding
- Use one-hot encoding (dummy variables)
- Implement target/mean encoding
- Handle high-cardinality categorical features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(42)
```

## 1. Label Encoding

```python
df = pd.DataFrame({
    'color':      ['Red', 'Blue', 'Green', 'Blue', 'Red', 'Green'],
    'size':       ['S', 'M', 'L', 'XL', 'M', 'S'],
    'quality':    ['Low', 'High', 'Medium', 'High', 'Low', 'Medium'],
    'price':      [10, 25, 20, 30, 12, 18]
})

le = LabelEncoder()
df['color_encoded'] = le.fit_transform(df['color'])
print('Label Encoding classes:', le.classes_)
print(df[['color', 'color_encoded']].drop_duplicates())
```

## 2. Ordinal Encoding

```python
ord_enc = OrdinalEncoder(categories=[['S', 'M', 'L', 'XL'], ['Low', 'Medium', 'High']])
df[['size_enc', 'quality_enc']] = ord_enc.fit_transform(df[['size', 'quality']])
print(df[['size', 'size_enc', 'quality', 'quality_enc']])
```

## 3. One-Hot Encoding

```python
df_ohe = pd.get_dummies(df[['color', 'size', 'price']], drop_first=False, prefix_sep='_')
print('After OHE:')
print(df_ohe)
print('Shape:', df_ohe.shape)
```

## 4. Target Encoding

```python
n = 200
categories = rng.choice(['A', 'B', 'C', 'D'], n)
target = np.where(categories == 'A', rng.normal(80, 5, n),
         np.where(categories == 'B', rng.normal(65, 5, n),
         np.where(categories == 'C', rng.normal(72, 5, n), rng.normal(55, 5, n))))
df_t = pd.DataFrame({'category': categories, 'target': target})

# Mean target per category (computed on train to avoid leakage)
mean_target = df_t.groupby('category')['target'].mean()
df_t['target_encoded'] = df_t['category'].map(mean_target)
print(mean_target.round(2))
```
