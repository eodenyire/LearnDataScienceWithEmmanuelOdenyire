# Day 8: Introduction to Feature Scaling

## Phase 2 | Month 5 | Week 2 | Day 8

**🎯 Goal:** Understand why feature scaling matters and which algorithms require it.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
N = 600
# Kenya loan approval — unscaled features with vastly different ranges
df = pd.DataFrame({
    'income_kes':   np.random.lognormal(10.8, 0.8, N),   # 10k–500k range
    'age':          np.random.normal(35, 10, N).clip(18, 70),  # 18–70 range
    'credit_score': np.random.normal(650, 80, N).clip(300, 850),  # 300–850 range
    'approved':     np.random.binomial(1, 0.55, N),
})

X = df[['income_kes','age','credit_score']].values
y = df['approved'].values

X_std = StandardScaler().fit_transform(X)

# Compare KNN accuracy with and without scaling
knn = KNeighborsClassifier(n_neighbors=5)
print('KNN (unscaled):', cross_val_score(knn, X, y, cv=5, scoring='accuracy').mean().round(4))
print('KNN (scaled):  ', cross_val_score(knn, X_std, y, cv=5, scoring='accuracy').mean().round(4))

print('\nLogistic Regression (unscaled):', cross_val_score(LogisticRegression(), X, y, cv=5).mean().round(4))
print('Logistic Regression (scaled):  ', cross_val_score(LogisticRegression(), X_std, y, cv=5).mean().round(4))
```

## Which Algorithms Need Scaling?

| Needs Scaling | Does NOT Need Scaling |
|--------------|---------------------|
| k-NN | Decision Trees |
| SVM | Random Forest |
| Logistic Regression | Gradient Boosting (XGBoost) |
| Neural Networks | Naïve Bayes |
| PCA | Rule-based systems |
| k-Means Clustering | |

**Rule:** Any algorithm that uses distances or gradient descent requires scaling.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 200
income = np.random.lognormal(10.8, 0.8, N)
age    = np.random.normal(35, 10, N).clip(18, 70)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(income, age, alpha=0.4, color='steelblue', s=30)
axes[0].set_title('Unscaled: income dominates distance', fontweight='bold')
axes[0].set_xlabel('Income (KES)')
axes[0].set_ylabel('Age')

from sklearn.preprocessing import StandardScaler
Xs = StandardScaler().fit_transform(np.column_stack([income, age]))
axes[1].scatter(Xs[:,0], Xs[:,1], alpha=0.4, color='tomato', s=30)
axes[1].set_title('Scaled: balanced feature space', fontweight='bold')
axes[1].set_xlabel('Income (z-score)')
axes[1].set_ylabel('Age (z-score)')

plt.tight_layout()
fig.savefig('/tmp/day08_scaling.png', dpi=120)
print('Saved /tmp/day08_scaling.png')
plt.close()
```
