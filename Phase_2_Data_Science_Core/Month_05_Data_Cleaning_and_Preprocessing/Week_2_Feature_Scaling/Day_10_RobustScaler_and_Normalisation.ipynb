# Day 10: RobustScaler and Normalisation

## Phase 2 | Month 5 | Week 2 | Day 10

**🎯 Goal:** Handle outlier-heavy data with RobustScaler and understand row-wise normalisation.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(7)
N = 400
# M-Pesa transaction dataset with heavy outliers
base_amount = np.random.lognormal(8, 1.5, N)
# Add 5% extreme outliers
base_amount[np.random.choice(N, int(N*0.05), replace=False)] *= 50

df = pd.DataFrame({'amount_kes': base_amount})

std_s    = StandardScaler().fit_transform(df[['amount_kes']]).ravel()
mm_s     = MinMaxScaler().fit_transform(df[['amount_kes']]).ravel()
robust_s = RobustScaler().fit_transform(df[['amount_kes']]).ravel()

fig, axes = plt.subplots(1, 4, figsize=(18, 4))
for ax, data, title in zip(axes,
        [base_amount, std_s, mm_s, robust_s],
        ['Original (outliers!)', 'StandardScaler', 'MinMaxScaler', 'RobustScaler']):
    ax.hist(np.clip(data, *np.percentile(data,[1,99])), bins=50,
            color='steelblue', alpha=0.7, edgecolor='white')
    ax.set_title(title, fontweight='bold', fontsize=9)
    ax.set_xlabel('Value')
plt.tight_layout()
fig.savefig('/tmp/day10_robust.png', dpi=120)
print('Saved /tmp/day10_robust.png')
plt.close()
print(f'RobustScaler uses: median={np.median(base_amount):,.0f}  IQR={np.percentile(base_amount,75)-np.percentile(base_amount,25):,.0f}')
```

## Row-wise Normalisation

`Normalizer` scales **each sample** (row) to have unit norm. Used for text/NLP features, cosine similarity, and neural networks.

```python
import numpy as np
from sklearn.preprocessing import Normalizer

# Text feature example: word frequency vectors
# Each row = a document, columns = word frequencies
X = np.array([[3, 0, 5, 2],
              [0, 8, 2, 0],
              [4, 1, 0, 6]])

norm_l2 = Normalizer(norm='l2').transform(X)
norm_l1 = Normalizer(norm='l1').transform(X)

print('Original row norms:', np.linalg.norm(X, axis=1).round(3))
print('L2 normalised row norms:', np.linalg.norm(norm_l2, axis=1).round(4))   # all 1.0
print('L1 normalised row norms:', np.abs(norm_l1).sum(axis=1).round(4))       # all 1.0
print('\nL2 normalised:\n', norm_l2.round(4))
```
