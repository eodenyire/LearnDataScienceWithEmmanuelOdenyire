```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 1000
# Safaricom customer data
df = pd.DataFrame({
    'customer_id':         range(1, N+1),
    'registration_date':   pd.date_range('2015-01-01', periods=N, freq='2D'),
    'last_transaction':    pd.date_range('2024-01-01', periods=N, freq='6H'),
    'total_transactions':  np.random.poisson(120, N),
    'total_amount_kes':    np.random.lognormal(11, 1.0, N),
    'num_top_up':          np.random.poisson(15, N),
    'num_withdrawal':      np.random.poisson(10, N),
    'num_send_money':      np.random.poisson(20, N),
    'data_gb_used':        np.random.lognormal(1.5, 0.8, N),
})

# Feature Engineering
reference_date = pd.Timestamp('2026-01-01')
df['tenure_days']      = (reference_date - df['registration_date']).dt.days
df['days_since_tx']    = (reference_date - df['last_transaction']).dt.days
df['avg_tx_value']     = df['total_amount_kes'] / df['total_transactions'].clip(1)
df['top_up_ratio']     = df['num_top_up'] / df['total_transactions'].clip(1)
df['send_ratio']       = df['num_send_money'] / df['total_transactions'].clip(1)
df['log_total_amount'] = np.log1p(df['total_amount_kes'])
df['is_heavy_data_user'] = (df['data_gb_used'] > df['data_gb_used'].quantile(0.80)).astype(int)

print('New features created:')
print(df[['tenure_days','days_since_tx','avg_tx_value','top_up_ratio','send_ratio',
          'log_total_amount','is_heavy_data_user']].describe().round(2))
```
