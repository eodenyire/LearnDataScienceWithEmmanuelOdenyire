# Week 2: Feature Scaling and Transformation

## Learning Objectives
- Apply min-max normalisation and standardisation
- Use robust scaling for outlier-resistant normalisation
- Apply log and power transformations
- Compare scaling effects on data distributions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, PowerTransformer

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Why Scaling Matters

```python
# Features with very different ranges
data = pd.DataFrame({
    'age':    rng.integers(18, 70, 200),
    'salary': rng.normal(60000, 15000, 200),
    'score':  rng.uniform(0, 100, 200)
}).astype(float)
print('Before scaling:')
print(data.describe().loc[['mean', 'std', 'min', 'max']].round(2))
```

## 2. Min-Max Normalisation

```python
mm = MinMaxScaler()
data_mm = pd.DataFrame(mm.fit_transform(data), columns=data.columns)
print('After Min-Max Scaling (range [0,1]):')
print(data_mm.describe().loc[['mean', 'std', 'min', 'max']].round(4))
```

## 3. Standard Scaling (Z-score)

```python
std = StandardScaler()
data_std = pd.DataFrame(std.fit_transform(data), columns=data.columns)
print('After Standard Scaling (mean=0, std=1):')
print(data_std.describe().loc[['mean', 'std']].round(4))
```

## 4. Robust Scaling

```python
rob = RobustScaler()
data_rob = pd.DataFrame(rob.fit_transform(data), columns=data.columns)
print('After Robust Scaling (IQR-based):')
print(data_rob.describe().loc[['mean', 'std', 'min', 'max']].round(4))
```

## 5. Log and Power Transformations

```python
skewed = rng.exponential(scale=3, size=300)
log_transformed = np.log1p(skewed)
pt = PowerTransformer(method='yeo-johnson')
power_transformed = pt.fit_transform(skewed.reshape(-1,1)).flatten()

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, d, t in zip(axes, [skewed, log_transformed, power_transformed],
                        ['Original (Skewed)', 'Log1p Transformed', 'Yeo-Johnson']):
    sns.histplot(d, kde=True, ax=ax)
    ax.set_title(t)
    sk = pd.Series(d).skew()
    ax.set_xlabel(f'Skewness: {sk:.2f}')
plt.tight_layout(); plt.show()
```
