# Day 12: Binning and Discretisation

## Phase 2 | Month 5 | Week 2 | Day 12

**🎯 Goal:** Convert continuous features into discrete bins for interpretability and non-linear modelling.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(3)
N = 800
# M-Shwari customer credit score data
credit_scores = np.random.normal(650, 90, N).clip(300, 850)

# Equal-width bins (pd.cut)
df = pd.DataFrame({'credit_score': credit_scores})
df['risk_band_equal_width'] = pd.cut(df['credit_score'],
    bins=[300, 450, 580, 680, 750, 850],
    labels=['Very High Risk','High Risk','Medium Risk','Low Risk','Very Low Risk'])

# Equal-frequency bins (pd.qcut)
df['risk_band_equal_freq'] = pd.qcut(df['credit_score'], q=5,
    labels=['VHR','HR','MR','LR','VLR'])

print(df['risk_band_equal_width'].value_counts().sort_index())
print('\n', df['risk_band_equal_freq'].value_counts().sort_index())
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(3)
N = 800
scores = np.random.normal(650, 90, N).clip(300, 850)
df = pd.DataFrame({'credit_score': scores})

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Original
axes[0].hist(scores, bins=40, color='steelblue', alpha=0.7, edgecolor='white')
axes[0].set_title('Original Continuous', fontweight='bold')
axes[0].set_xlabel('Credit Score')

# Equal-width
bands = pd.cut(scores, bins=5)
axes[1].bar(*zip(*bands.value_counts().sort_index().items()),
            color='tomato', alpha=0.8)
axes[1].set_title('Equal-Width Bins', fontweight='bold')
axes[1].tick_params(axis='x', rotation=30)

# KBinsDiscretizer — kmeans strategy
kbd = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='kmeans')
bins_km = kbd.fit_transform(scores.reshape(-1,1)).ravel()
axes[2].bar(*zip(*pd.Series(bins_km).value_counts().sort_index().items()),
            color='green', alpha=0.8)
axes[2].set_title('KBins (K-means) Strategy', fontweight='bold')
axes[2].set_xlabel('Bin')

plt.tight_layout()
fig.savefig('/tmp/day12_bins.png', dpi=120)
print('Saved /tmp/day12_bins.png')
plt.close()
```

## Why Discretise?

- **Interpretability**: 'High Risk' is clearer than 'credit_score=712'
- **Non-linear relationships**: tree-friendly representation
- **Robustness**: reduces impact of outliers
- **Business rules**: NHIF contribution tiers, tax brackets (PAYE Kenya)
