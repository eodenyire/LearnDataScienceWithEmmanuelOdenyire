```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 400
df = pd.DataFrame({
    'age':        np.where(np.random.rand(N)<0.12, np.nan, np.random.normal(38,12,N).clip(18,75)),
    'salary_kes': np.where(np.random.rand(N)<0.18, np.nan, np.random.lognormal(10.8,0.9,N)),
    'department': np.where(np.random.rand(N)<0.08, np.nan,
                           np.random.choice(['IT','Finance','HR','Ops','Sales'],N,p=[0.25,0.2,0.15,0.2,0.2])),
})

# Numeric imputers
age_mean_imp   = SimpleImputer(strategy='mean')
age_median_imp = SimpleImputer(strategy='median')

age_mean   = age_mean_imp.fit_transform(df[['age']]).ravel()
age_median = age_median_imp.fit_transform(df[['age']]).ravel()

# Categorical imputer
dept_imp = SimpleImputer(strategy='most_frequent')
dept_imputed = dept_imp.fit_transform(df[['department']]).ravel()

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for label, vals, color in [('Original (no NaN)', df['age'].dropna(), 'gray'),
                             ('Mean imputed', age_mean, 'steelblue'),
                             ('Median imputed', age_median, 'tomato')]:
    axes[0].hist(vals, bins=30, alpha=0.5, label=label, color=color if color!='gray' else None)
axes[0].set_title('Age Distribution — Imputation Comparison', fontweight='bold')
axes[0].legend()
axes[0].set_xlabel('Age')

axes[1].bar(*zip(*pd.Series(dept_imputed).value_counts().items()),
            color='green', alpha=0.7, edgecolor='white')
axes[1].set_title('Department (Most-Frequent Imputed)', fontweight='bold')
plt.tight_layout()
fig.savefig('/tmp/day04_imputation.png', dpi=120)
print('Saved')
plt.close()
print(f'Mean imputed age  : {age_mean.mean():.2f}')
print(f'Median imputed age: {age_median.mean():.2f}')
```

## Group-Wise (Conditional) Imputation

Imputing using the group mean is more accurate for MAR data — e.g., imputing salary using the department mean.

```python
import numpy as np
import pandas as pd

np.random.seed(42)
N = 400
df = pd.DataFrame({
    'salary_kes': np.where(np.random.rand(N)<0.18, np.nan, np.random.lognormal(10.8,0.9,N)),
    'department': np.random.choice(['IT','Finance','HR','Ops','Sales'],N,p=[0.25,0.2,0.15,0.2,0.2]),
})

# Global mean imputation
global_mean = df['salary_kes'].mean()
df['salary_global'] = df['salary_kes'].fillna(global_mean)

# Group mean imputation
df['salary_group'] = df.groupby('department')['salary_kes'].transform(lambda x: x.fillna(x.mean()))

print('Group means:')
print(df.groupby('department')['salary_kes'].mean().round(0))
print(f'\nGlobal mean: {global_mean:,.0f} KES')
print(f'Group-wise imputation varies by department — more realistic!')
```

## 📋 Summary

- ✓ **Mean**: good for symmetric numeric data without outliers
- ✓ **Median**: robust — preferred for skewed data (income, prices)
- ✓ **Mode / Most Frequent**: correct for categorical features
- ✓ **Group-wise imputation** reduces bias when missingness depends on category
- ✓ Always use `SimpleImputer` in a pipeline — never `fillna` directly on test data
