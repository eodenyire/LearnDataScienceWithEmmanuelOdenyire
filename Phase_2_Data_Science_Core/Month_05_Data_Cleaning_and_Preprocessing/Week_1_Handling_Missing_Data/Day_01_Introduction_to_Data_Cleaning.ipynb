### Dimensions of Data Quality

| Dimension | Question | Kenya Example |
|-----------|----------|---------------|
| **Completeness** | Are all values present? | Missing county in NHIF registration |
| **Consistency** | Are values logically consistent? | Phone: 0712345678 vs +254712345678 |
| **Accuracy** | Do values reflect reality? | Age = 200 in a hospital record |
| **Timeliness** | Are values up to date? | 2019 census population used in 2026 |
| **Uniqueness** | No duplicates? | Same patient registered twice in SHIF |

````python
import numpy as np
import pandas as pd
np.random.seed(42)
N = 500

# Simulate a messy KRA (Kenya Revenue Authority) tax-payer dataset
df = pd.DataFrame({
    'taxpayer_id': list(range(1, N+1)) + [50, 120, 300],         # duplicates
    'name':        ['Person_'+str(i) for i in range(1, N+1)] + ['Person_50','Person_120','Person_300'],
    'county':      np.random.choice(['Nairobi','Mombasa','Kisumu',None,'Nakuru'], N+3,
                                     p=[0.35,0.20,0.15,0.20,0.10]),
    'annual_income_kes': np.concatenate([
                        np.random.lognormal(12, 1.2, N-5),
                        [9_999_999_999, -5000, np.nan, 0, np.nan],  # outliers + missing
                        [np.nan, np.nan, np.nan]
                    ]),
    'phone': np.random.choice(['0712345678','+254712345678','07123456','NULL',None], N+3),
    'registration_date': pd.date_range('2018-01-01', periods=N+3, freq='D'),
})

print('Shape:', df.shape)
print('\n--- dtypes ---')
print(df.dtypes)
print('\n--- Missing values ---')
print(df.isnull().sum())
print('\n--- Duplicate taxpayer IDs ---')
print(df.duplicated(subset='taxpayer_id').sum())
````

````python
import numpy as np
import pandas as pd

np.random.seed(42)
N = 500
df = pd.DataFrame({
    'county': np.random.choice(['Nairobi','Mombasa','Kisumu',None,'Nakuru'], N, p=[0.35,0.20,0.15,0.20,0.10]),
    'annual_income_kes': np.concatenate([np.random.lognormal(12, 1.2, N-5),
                                          [9_999_999_999, -5000, np.nan, 0, np.nan]]),
    'phone': np.random.choice(['0712345678','+254712345678','07123456','NULL',None], N),
})

# Data quality report
print('=== Quick Data Quality Report ===')
for col in df.columns:
    n_miss = df[col].isnull().sum()
    pct    = n_miss / len(df) * 100
    dtype  = df[col].dtype
    n_uniq = df[col].nunique(dropna=True)
    print(f'{col:25s}  dtype={str(dtype):10s}  missing={n_miss:3d} ({pct:.1f}%)  unique={n_uniq}')
````
