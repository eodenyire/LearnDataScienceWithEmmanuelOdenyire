# Week 1: Handling Missing Data

## Learning Objectives
- Identify patterns in missing data (MCAR, MAR, MNAR)
- Use deletion and imputation strategies
- Apply advanced imputation with scikit-learn
- Evaluate the impact of missing data handling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer, KNNImputer

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Understanding Missing Data

```python
data = {
    'age':     [25, None, 35, 28, None, 42, 31, None, 29, 38],
    'salary':  [50000, 60000, None, 55000, 62000, None, 58000, 63000, None, 70000],
    'dept':    ['Eng', 'Mkt', 'Eng', None, 'HR', 'Eng', None, 'Mkt', 'HR', 'Eng'],
    'score':   [85, 92, None, 78, 88, 95, None, 83, 79, None]
}
df = pd.DataFrame(data)
print(df)
print('\nMissing count:\n', df.isnull().sum())
print('\nMissing %:')
print((df.isnull().sum() / len(df) * 100).round(1))
```

## 2. Deletion Strategies

```python
# Listwise deletion
df_dropped = df.dropna()
print(f'Original: {len(df)} rows')
print(f'After listwise deletion: {len(df_dropped)} rows')

# Threshold deletion: keep rows with at least 3 non-null values
df_thresh = df.dropna(thresh=3)
print(f'After threshold (thresh=3): {len(df_thresh)} rows')

# Column deletion: drop columns > 30% missing
missing_pct = df.isnull().mean()
cols_to_drop = missing_pct[missing_pct > 0.3].index
print(f'Columns with >30% missing: {list(cols_to_drop)}')
```

## 3. Imputation Strategies

```python
# Simple imputation
df_num = df[['age', 'salary', 'score']].copy()
imputers = {
    'Mean':   SimpleImputer(strategy='mean'),
    'Median': SimpleImputer(strategy='median'),
    'Const':  SimpleImputer(strategy='constant', fill_value=0)
}
for name, imp in imputers.items():
    result = imp.fit_transform(df_num)
    means = result.mean(axis=0).round(1)
    print(f'{name:7s} imputed means: age={means[0]}, salary={means[1]}, score={means[2]}')
```

```python
# KNN Imputation
knn_imp = KNNImputer(n_neighbors=3)
df_knn = pd.DataFrame(knn_imp.fit_transform(df_num), columns=df_num.columns)
print('KNN imputed:\n', df_knn.round(1))
```
