```python
import numpy as np
import pandas as pd

np.random.seed(1)
N = 600
df = pd.DataFrame({
    'loan_id':     range(1, N+1),
    'amount_kes':  np.random.lognormal(11, 1.0, N).round(0),
    'term_months': np.random.choice([6, 12, 24, 36], N),
    'interest_rt': np.where(np.random.rand(N)<0.10, np.nan, np.random.uniform(12,28,N).round(2)),
    'credit_score':np.where(np.random.rand(N)<0.25, np.nan, np.random.normal(650,80,N).round(0)),
    'defaulted':   np.random.choice([0, 1], N, p=[0.88, 0.12]),
})

# Pairwise (default) vs listwise deletion
df_listwise = df.dropna()                          # drop row if ANY column missing
df_pairwise_interest = df.dropna(subset=['interest_rt'])   # drop only when this column missing
df_thresh  = df.dropna(thresh=5)                   # keep rows with at least 5 non-NaN

print(f'Original        : {df.shape[0]:4d} rows')
print(f'Listwise (any)  : {df_listwise.shape[0]:4d} rows  (lost {df.shape[0]-df_listwise.shape[0]})')
print(f'Pairwise subset : {df_pairwise_interest.shape[0]:4d} rows  (lost {df.shape[0]-df_pairwise_interest.shape[0]})')
print(f'thresh=5        : {df_thresh.shape[0]:4d} rows  (lost {df.shape[0]-df_thresh.shape[0]})')

# Drop columns with >20% missing
col_miss_pct = df.isnull().mean()
cols_to_drop = col_miss_pct[col_miss_pct > 0.20].index.tolist()
print(f'\nColumns >20% missing: {cols_to_drop}')
```

## Rules of Thumb for Dropping

| Scenario | Recommendation |
|----------|----------------|
| Column > 40–50% missing | Consider dropping the column |
| Row missing only non-critical fields | Keep the row |
| Data is MCAR and n is large | Listwise deletion OK |
| Data is MAR or MNAR | Dropping causes bias — prefer imputation |
| n < 1000 and > 5% missing | Every row matters — impute instead |

```python
import numpy as np
import pandas as pd

np.random.seed(1)
N = 600
df = pd.DataFrame({
    'amount_kes':   np.random.lognormal(11, 1.0, N).round(0),
    'credit_score': np.where(np.random.rand(N)<0.25, np.nan, np.random.normal(650,80,N).round(0)),
    'defaulted':    np.random.choice([0, 1], N, p=[0.88, 0.12]),
})

# Check if default rate differs between complete and incomplete cases
complete_cases = df.dropna()
incomplete_cases = df[df.isnull().any(axis=1)]

print(f'Default rate (complete cases):   {complete_cases["defaulted"].mean():.3f}')
print(f'Default rate (incomplete cases): {incomplete_cases["defaulted"].mean():.3f}')
print(f'\nIf rates differ significantly, dropping causes bias!')

from scipy import stats
t, p = stats.ttest_ind(complete_cases['defaulted'], incomplete_cases['defaulted'])
print(f't={t:.4f}, p={p:.4f} → {"Biased!" if p<0.05 else "Safe to drop"} (α=0.05)')
```
