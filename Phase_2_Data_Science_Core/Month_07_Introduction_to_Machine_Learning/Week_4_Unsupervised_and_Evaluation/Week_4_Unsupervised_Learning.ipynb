# Week 4: Unsupervised Learning

## Learning Objectives
- Apply K-Means and DBSCAN clustering
- Evaluate clustering with silhouette score
- Use PCA for dimensionality reduction
- Visualise high-dimensional data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris, make_blobs, make_moons
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. K-Means Clustering

```python
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.5, random_state=42)

inertias, silhouettes = [], []
ks = range(2, 9)
for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X, labels))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(ks, inertias, marker='o'); axes[0].set_title('Elbow Method')
axes[1].plot(ks, silhouettes, marker='o', color='coral'); axes[1].set_title('Silhouette Score')
plt.tight_layout(); plt.show()
print('Best k:', ks[np.argmax(silhouettes)])
```

## 2. PCA — Dimensionality Reduction

```python
iris = load_iris()
X_iris = StandardScaler().fit_transform(iris.data)

pca = PCA()
pca.fit(X_iris)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(range(1,5), pca.explained_variance_ratio_)
axes[0].plot(range(1,5), pca.explained_variance_ratio_.cumsum(), 'r-o', label='Cumulative')
axes[0].set_title('Explained Variance')
axes[0].legend()

X_pca = PCA(n_components=2).fit_transform(X_iris)
axes[1].scatter(X_pca[:,0], X_pca[:,1], c=iris.target, cmap='Set1', alpha=0.7)
axes[1].set_title('PCA: 2 Components')
axes[1].set_xlabel('PC1'); axes[1].set_ylabel('PC2')
plt.tight_layout(); plt.show()
```

## 3. DBSCAN for Non-Convex Clusters

```python
X_moon, _ = make_moons(n_samples=200, noise=0.1, random_state=42)
labels_db = DBSCAN(eps=0.2, min_samples=5).fit_predict(X_moon)

plt.figure(figsize=(7, 4))
plt.scatter(X_moon[:,0], X_moon[:,1], c=labels_db, cmap='Set1', alpha=0.7)
num_clusters = len(set(labels_db)) - (1 if -1 in labels_db else 0)
plt.title(f'DBSCAN: {num_clusters} clusters')
plt.tight_layout(); plt.show()
```
