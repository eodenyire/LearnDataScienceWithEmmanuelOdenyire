# Week 1: Ensemble Methods

## Learning Objectives
- Understand bagging and boosting
- Use Random Forest and Extra Trees
- Train Gradient Boosting models
- Compare ensemble performance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import (RandomForestClassifier, ExtraTreesClassifier,
                               GradientBoostingClassifier, BaggingClassifier)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import roc_auc_score

sns.set_theme(style='whitegrid')
```

## 1. Bagging vs Single Tree

```python
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42)

models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Bagging':       BaggingClassifier(n_estimators=100, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Extra Trees':   ExtraTreesClassifier(n_estimators=100, random_state=42),
    'Gradient Boost':GradientBoostingClassifier(n_estimators=100, random_state=42)
}

for name, clf in models.items():
    cv  = cross_val_score(clf, X, y, cv=5, scoring='roc_auc').mean()
    clf.fit(X_train, y_train)
    auc = roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
    print(f'{name:20s}: CV={cv:.4f}, Test AUC={auc:.4f}')
```

## 2. Feature Importance

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
imp_df = pd.DataFrame({'feature': cancer.feature_names,
                        'importance': rf.feature_importances_}
                      ).sort_values('importance', ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(data=imp_df, x='importance', y='feature', palette='viridis')
plt.title('Top 10 Feature Importances')
plt.tight_layout(); plt.show()
```

## 3. XGBoost (if installed)

```python
try:
    import xgboost as xgb
    xg = xgb.XGBClassifier(n_estimators=200, learning_rate=0.1,
                             max_depth=3, eval_metric='auc',
                             random_state=42, use_label_encoder=False)
    xg.fit(X_train, y_train)
    print(f'XGBoost AUC: {roc_auc_score(y_test, xg.predict_proba(X_test)[:,1]):.4f}')
except ImportError:
    print('XGBoost not installed: pip install xgboost')
```
