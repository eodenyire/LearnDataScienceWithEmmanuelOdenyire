# Week 2: Univariate and Bivariate Analysis

## Learning Objectives
- Analyse individual variables (univariate)
- Explore relationships between pairs of variables (bivariate)
- Use appropriate plots for different variable types
- Compute and interpret descriptive statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

tips = sns.load_dataset('tips')
print(tips.head())
print('Shape:', tips.shape)
```

## 1. Univariate Analysis — Numeric Variables

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 9))
sns.histplot(tips['total_bill'], kde=True, bins=25, ax=axes[0,0])
axes[0,0].set_title('Total Bill Distribution')

stats.probplot(tips['total_bill'], plot=axes[0,1])
axes[0,1].set_title('Q-Q Plot: Total Bill')

sns.boxplot(y=tips['tip'], ax=axes[1,0])
axes[1,0].set_title('Tip Boxplot')

sns.violinplot(y=tips['total_bill'], ax=axes[1,1])
axes[1,1].set_title('Total Bill Violin Plot')
plt.tight_layout(); plt.show()
```

## 2. Univariate Analysis — Categorical Variables

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
tips['day'].value_counts().plot(kind='bar', ax=axes[0], rot=0, color='steelblue')
axes[0].set_title('Count by Day')

tips['time'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.0f%%')
axes[1].set_title('Meal Time Split')

tips.groupby('day')['total_bill'].mean().sort_values().plot(kind='barh', ax=axes[2], color='coral')
axes[2].set_title('Avg Bill by Day')
plt.tight_layout(); plt.show()
```

## 3. Bivariate Analysis — Numeric vs Numeric

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', ax=axes[0])
axes[0].set_title('Total Bill vs Tip')

corr = stats.pearsonr(tips['total_bill'], tips['tip'])
axes[0].set_xlabel(f'Total Bill (r={corr[0]:.2f}, p={corr[1]:.4f})')

sns.regplot(data=tips, x='total_bill', y='tip', ax=axes[1])
axes[1].set_title('Regression: Bill vs Tip')
plt.tight_layout(); plt.show()
```

## 4. Bivariate Analysis — Categorical vs Numeric

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0], palette='Set2')
axes[0].set_title('Total Bill by Day')

sns.pointplot(data=tips, x='day', y='tip', hue='sex', ax=axes[1])
axes[1].set_title('Average Tip by Day and Gender')
plt.tight_layout(); plt.show()
```
