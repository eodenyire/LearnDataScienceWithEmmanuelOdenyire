# Week 4: Correlation Analysis

## Learning Objectives
- Compute Pearson, Spearman, and Kendall correlations
- Visualise correlation matrices
- Understand correlation vs causation
- Detect multicollinearity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Pearson Correlation

```python
n = 100
x = rng.normal(0, 1, n)
y_strong = 2*x + rng.normal(0, 0.5, n)
y_weak   = 0.5*x + rng.normal(0, 2, n)
y_none   = rng.normal(0, 1, n)

for label, y in [('Strong', y_strong), ('Weak', y_weak), ('None', y_none)]:
    r, p = stats.pearsonr(x, y)
    print(f'{label:6s}: r={r:+.3f}, p={p:.4f}')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, (label, y) in zip(axes, [('Strong +', y_strong), ('Weak +', y_weak), ('None', y_none)]):
    r, _ = stats.pearsonr(x, y)
    ax.scatter(x, y, alpha=0.5)
    ax.set_title(f'{label} (r={r:.2f})')
    ax.set_xlabel('x'); ax.set_ylabel('y')
plt.tight_layout(); plt.show()
```

## 2. Correlation Matrix

```python
iris = sns.load_dataset('iris')
corr_matrix = iris.drop('species', axis=1).corr()
print(corr_matrix.round(3))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=axes[0])
axes[0].set_title('Pearson Correlation Matrix (Iris)')

# Spearman
spearman = iris.drop('species', axis=1).corr(method='spearman')
sns.heatmap(spearman, annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=axes[1])
axes[1].set_title('Spearman Correlation Matrix')
plt.tight_layout(); plt.show()
```

## 3. Pair Plot

```python
sns.pairplot(iris, hue='species', diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pair Plot — Iris Dataset', y=1.02)
plt.show()
```

## 4. Multicollinearity Detection

```python
from sklearn.linear_model import LinearRegression

# Variance Inflation Factor (VIF)
def vif(X):
    vif_data = {}
    cols = X.columns if hasattr(X, 'columns') else [f'f{i}' for i in range(X.shape[1])]
    X_arr = X.values if hasattr(X, 'values') else X
    for i, col in enumerate(cols):
        x_i = X_arr[:, i]
        X_rest = np.delete(X_arr, i, axis=1)
        r2 = LinearRegression().fit(X_rest, x_i).score(X_rest, x_i)
        vif_data[col] = 1 / (1 - r2) if r2 < 1 else float('inf')
    return vif_data

X_iris = iris.drop('species', axis=1)
vif_scores = vif(X_iris)
for col, v in vif_scores.items():
    print(f'{col}: VIF = {v:.2f}', '(multicollinear)' if v > 5 else '')
```
