## 1. What is Data Science?

Data Science is an interdisciplinary field that uses:
- **Statistics** for understanding data patterns
- **Programming** for automation and analysis
- **Domain Knowledge** for context and insights

### Key Components:
- Data Collection
- Data Cleaning
- Data Analysis
- Data Visualization
- Machine Learning
- Communication of Results

## 2. Why Python for Data Science?

Python is the most popular language for data science because:
- **Easy to Learn**: Clean and readable syntax
- **Rich Ecosystem**: NumPy, Pandas, Scikit-learn, TensorFlow, etc.
- **Community Support**: Large community and extensive documentation
- **Versatile**: Web development, automation, data science, AI

## 4. Basic Python Syntax

### Variables
Variables store data for later use:

```python
# Creating variables
name = "Emmanuel"
age = 25
is_data_scientist = True

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Is Data Scientist: {is_data_scientist}")
```

### Simple Calculations

```python
# Basic arithmetic
a = 10
b = 3

print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Integer Division: {a // b}")
print(f"Modulus: {a % b}")
print(f"Exponentiation: {a ** b}")
```
