## 1. What are Variables?

Variables are containers for storing data values. In Python, you don't need to declare variable types explicitly.

```python
# Creating variables
x = 5
name = "Alice"
pi = 3.14159
is_student = True

print(x, name, pi, is_student)
```

## 2. Python Data Types

### Numeric Types

```python
# Integer
age = 25
print(f"Integer: {age}, Type: {type(age)}")

# Float
temperature = 36.6
print(f"Float: {temperature}, Type: {type(temperature)}")

# Complex
complex_num = 3 + 4j
print(f"Complex: {complex_num}, Type: {type(complex_num)}")
```

### String Type

```python
# Single quotes
message1 = 'Hello, World!'

# Double quotes
message2 = "Hello, Python!"

# Triple quotes for multi-line
message3 = """This is a
multi-line
string"""

print(message1)
print(message2)
print(message3)
```

### Boolean Type

```python
is_sunny = True
is_raining = False

print(f"Is sunny: {is_sunny}")
print(f"Is raining: {is_raining}")
print(f"Type: {type(is_sunny)}")
```

## 3. Type Checking and Conversion

```python
# Type checking
x = 10
print(f"Type of x: {type(x)}")

# Type conversion
x_float = float(x)
print(f"x as float: {x_float}, Type: {type(x_float)}")

x_str = str(x)
print(f"x as string: {x_str}, Type: {type(x_str)}")

# Converting string to integer
num_str = "100"
num_int = int(num_str)
print(f"String to int: {num_int}, Type: {type(num_int)}")
```

## 4. Variable Naming Rules

**Rules:**
- Must start with a letter or underscore
- Can contain letters, numbers, and underscores
- Case-sensitive (age, Age, and AGE are different)
- Cannot use Python keywords

**Best Practices:**
- Use descriptive names: `student_age` instead of `sa`
- Use snake_case for variables: `my_variable`
- Use UPPERCASE for constants: `MAX_SIZE = 100`

```python
# Good variable names
student_name = "John"
total_score = 95
is_passed = True

# Constants
MAX_ATTEMPTS = 3
PI = 3.14159

print(student_name, total_score, is_passed)
print(MAX_ATTEMPTS, PI)
```

## 5. Multiple Assignment

```python
# Assign multiple variables at once
x, y, z = 1, 2, 3
print(f"x={x}, y={y}, z={z}")

# Assign same value to multiple variables
a = b = c = 100
print(f"a={a}, b={b}, c={c}")
```
