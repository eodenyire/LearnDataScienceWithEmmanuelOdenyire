# Week 4: Mini Data Analyzer Project
## Project Objectives
- Build a command-line data analysis tool in pure Python
- Apply functions, OOP, file handling, and data structures
- Generate a text-based report with statistics and insights


```python
import csv, json, statistics, os
from collections import Counter

class DataAnalyzer:
    """A mini data analyzer for CSV files."""

    def __init__(self, filepath):
        self.filepath = filepath
        self.data = []
        self.headers = []
        self._load()

    def _load(self):
        with open(self.filepath, 'r') as f:
            reader = csv.DictReader(f)
            self.headers = reader.fieldnames or []
            self.data = list(reader)
        print(f'Loaded {len(self.data)} rows, {len(self.headers)} columns')

    def numeric_stats(self, column):
        values = []
        for row in self.data:
            try:
                values.append(float(row[column]))
            except (ValueError, KeyError):
                pass
        if not values:
            return {}
        return {
            'count': len(values),
            'mean': statistics.mean(values),
            'median': statistics.median(values),
            'stdev': statistics.stdev(values) if len(values) > 1 else 0,
            'min': min(values),
            'max': max(values)
        }

    def categorical_summary(self, column):
        values = [row.get(column, '') for row in self.data]
        return Counter(values)

print('DataAnalyzer class defined')
```

```python
# Create sample dataset
sample_csv = '/tmp/sales_data.csv'
with open(sample_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['date', 'product', 'region', 'sales', 'units'])
    import random
    random.seed(42)
    products = ['Widget A', 'Widget B', 'Gadget X']
    regions = ['North', 'South', 'East', 'West']
    for i in range(50):
        writer.writerow([
            f'2024-{(i%12)+1:02d}-01',
            random.choice(products),
            random.choice(regions),
            round(random.uniform(100, 5000), 2),
            random.randint(1, 50)
        ])

analyzer = DataAnalyzer(sample_csv)
print(f'Headers: {analyzer.headers}')
```

```python
# Run analysis
sales_stats = analyzer.numeric_stats('sales')
print('\nSales Statistics:')
for k, v in sales_stats.items():
    print(f'  {k:8s}: {v:.2f}')

print('\nProduct Distribution:')
for product, count in analyzer.categorical_summary('product').most_common():
    bar = '█' * count
    print(f'  {product:12s}: {bar} ({count})')

print('\nRegion Distribution:')
for region, count in analyzer.categorical_summary('region').most_common():
    bar = '█' * count
    print(f'  {region:8s}: {bar} ({count})')
```

```python
# Generate report
def generate_report(analyzer, output_path):
    report = {
        'file': analyzer.filepath,
        'rows': len(analyzer.data),
        'columns': analyzer.headers,
        'analysis': {}
    }
    for col in ['sales', 'units']:
        report['analysis'][col] = analyzer.numeric_stats(col)
    for col in ['product', 'region']:
        report['analysis'][col] = dict(analyzer.categorical_summary(col))

    with open(output_path, 'w') as f:
        json.dump(report, f, indent=2)
    print(f'Report saved to {output_path}')
    return report

report = generate_report(analyzer, '/tmp/analysis_report.json')
print(f'\nReport keys: {list(report.keys())}')
```
