## 1. Conditional Statements (if / elif / else)

```python
temperature = 32

if temperature > 35:
    print('It is hot!')
elif temperature > 25:
    print('It is warm.')
elif temperature > 15:
    print('It is mild.')
else:
    print('It is cold.')
```

```python
# Nested conditions
score = 78

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f'Score: {score}, Grade: {grade}')
```

## 2. For Loops

Iterate over sequences:

```python
# Loop over a list
fruits = ['apple', 'banana', 'cherry', 'date']
for fruit in fruits:
    print(f'I like {fruit}')

# Loop with range
print('\nSquares 1-5:')
for i in range(1, 6):
    print(f'{i}^2 = {i**2}')
```

```python
# Enumerate for index + value
languages = ['Python', 'R', 'SQL', 'Julia']
for idx, lang in enumerate(languages, start=1):
    print(f'{idx}. {lang}')

# Nested loop
print('\nMultiplication table (1-3):')
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i}x{j}={i*j}', end='  ')
    print()
```

## 3. While Loops

```python
# Basic while loop
count = 0
while count < 5:
    print(f'Count: {count}')
    count += 1

# break and continue
print('\nNumbers 1-10, skipping 5:')
for n in range(1, 11):
    if n == 5:
        continue
    if n == 8:
        break
    print(n, end=' ')
print()
```

## 4. Functions

```python
# Defining a function
def greet(name):
    """Return a greeting message."""
    return f'Hello, {name}!'

print(greet('Emmanuel'))
print(greet('Data Science Learner'))
```

```python
# Function with default parameters
def calculate_bmi(weight_kg, height_m, name='Unknown'):
    """Calculate Body Mass Index."""
    bmi = weight_kg / (height_m ** 2)
    if bmi < 18.5:
        category = 'Underweight'
    elif bmi < 25:
        category = 'Normal'
    elif bmi < 30:
        category = 'Overweight'
    else:
        category = 'Obese'
    return bmi, category

bmi, cat = calculate_bmi(70, 1.75, 'Alice')
print(f'BMI: {bmi:.1f} ({cat})')
```
