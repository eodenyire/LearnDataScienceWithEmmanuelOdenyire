# Week 1: Descriptive Statistics

## Learning Objectives
- Calculate measures of central tendency (mean, median, mode)
- Compute measures of spread (variance, std, IQR)
- Detect and handle outliers
- Understand data distributions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_theme(style='whitegrid')
```

## 1. Measures of Central Tendency

```python
data = [72, 85, 90, 68, 74, 92, 88, 76, 83, 79, 95, 71, 84, 77, 86]
arr = np.array(data)

print(f'Mean:   {arr.mean():.2f}')
print(f'Median: {np.median(arr):.2f}')
print(f'Mode:   {stats.mode(arr, keepdims=True).mode[0]}')

# Trimmed mean (robust)
print(f'Trimmed Mean (10%): {stats.trim_mean(arr, 0.10):.2f}')
```

## 2. Measures of Spread

```python
print(f'Range:    {arr.max() - arr.min()}')
print(f'Variance: {arr.var():.2f}')
print(f'Std Dev:  {arr.std():.2f}')
q1, q3 = np.percentile(arr, [25, 75])
iqr = q3 - q1
print(f'IQR:      {iqr:.2f}')
print(f'Q1={q1}, Q3={q3}')
```

## 3. Outlier Detection

```python
# Z-score method
z_scores = np.abs(stats.zscore(arr))
outliers_z = arr[z_scores > 2]
print('Outliers (Z > 2):', outliers_z)

# IQR method
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers_iqr = arr[(arr < lower) | (arr > upper)]
print('Outliers (IQR):', outliers_iqr)
print(f'Fence: [{lower:.1f}, {upper:.1f}]')
```

## 4. Visualising Distributions

```python
rng = np.random.default_rng(42)
normal_data = rng.normal(70, 12, 300)
skewed_data = rng.exponential(scale=10, size=300)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
sns.histplot(normal_data, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Normal Distribution')

sns.histplot(skewed_data, kde=True, ax=axes[1], color='coral')
axes[1].set_title('Skewed Distribution')

axes[2].boxplot([normal_data, skewed_data], labels=['Normal', 'Skewed'])
axes[2].set_title('Box Plots')
plt.tight_layout()
plt.show()
```
