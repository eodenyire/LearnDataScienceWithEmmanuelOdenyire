# Week 2: Probability Theory and Distributions

## Learning Objectives

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Basic Probability

```python
# Simulating a fair coin
n_flips = 10000
flips = rng.integers(0, 2, n_flips)  # 0=heads, 1=tails
p_heads = flips.mean()
print(f'P(Heads) after {n_flips} flips: {1 - p_heads:.4f}')

# Dice rolls
dice = rng.integers(1, 7, 10000)
print(f'P(6) after 10000 rolls: {(dice == 6).mean():.4f}  (expected: {1/6:.4f})')
```

## 2. Discrete Distributions

```python
# Binomial: P(X=k) for n=10 trials, p=0.4
n, p = 10, 0.4
x = np.arange(0, n+1)
pmf = stats.binom.pmf(x, n, p)

# Poisson: avg 3 events per interval
lam = 3
x_p = np.arange(0, 12)
poisson_pmf = stats.poisson.pmf(x_p, lam)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(x, pmf, color='steelblue', alpha=0.8)
axes[0].set_title(f'Binomial(n={n}, p={p})')
axes[0].set_xlabel('k'); axes[0].set_ylabel('P(X=k)')

axes[1].bar(x_p, poisson_pmf, color='coral', alpha=0.8)
axes[1].set_title(f'Poisson(λ={lam})')
axes[1].set_xlabel('k'); axes[1].set_ylabel('P(X=k)')
plt.tight_layout(); plt.show()
```

## 3. Continuous Distributions

```python
x_cont = np.linspace(-4, 4, 200)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for mu, std, label in [(0, 1, 'Standard'), (0, 2, 'σ=2'), (1, 1, 'μ=1')]:
    axes[0].plot(x_cont, stats.norm.pdf(x_cont, mu, std), label=label, lw=2)
axes[0].set_title('Normal Distributions')
axes[0].legend(); axes[0].set_xlabel('x'); axes[0].set_ylabel('PDF')

# Exponential
x_exp = np.linspace(0, 5, 200)
for scale, label in [(0.5, 'λ=2'), (1, 'λ=1'), (2, 'λ=0.5')]:
    axes[1].plot(x_exp, stats.expon.pdf(x_exp, scale=scale), label=label, lw=2)
axes[1].set_title('Exponential Distributions')
axes[1].legend(); axes[1].set_xlabel('x'); axes[1].set_ylabel('PDF')
plt.tight_layout(); plt.show()
```

## 4. Central Limit Theorem

```python
# Population: heavily skewed (exponential)
population = rng.exponential(scale=2, size=100000)

sample_means = [rng.choice(population, 50).mean() for _ in range(1000)]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(population, bins=60, color='coral', density=True, alpha=0.7)
axes[0].set_title('Population (Exponential)')

axes[1].hist(sample_means, bins=40, color='steelblue', density=True, alpha=0.7)
axes[1].axvline(np.mean(sample_means), color='red', lw=2, label=f'Mean={np.mean(sample_means):.2f}')
axes[1].set_title('Sampling Distribution of Mean (n=50)')
axes[1].legend()
plt.tight_layout(); plt.show()
print(f'Population mean: {population.mean():.2f}, Sampling mean: {np.mean(sample_means):.2f}')
```
