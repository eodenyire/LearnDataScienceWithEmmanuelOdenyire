# Large-Scale Data Processing with Apache Spark
## Learning Objectives
- Understand Spark's distributed computing model
- Use PySpark DataFrames for data manipulation
- Perform aggregations, joins, and window functions


```python
# Spark architecture concepts
spark_concepts = {
    'Driver': 'Orchestrates job execution, holds SparkContext',
    'Executor': 'JVM process on worker node that runs tasks',
    'RDD': 'Resilient Distributed Dataset - immutable distributed collection',
    'DataFrame': 'Distributed table with named columns (like pandas)',
    'Partition': 'Chunk of data processed in parallel',
    'DAG': 'Directed Acyclic Graph of transformations',
    'Lazy Evaluation': 'Transformations are not executed until an action is called'
}
for k, v in spark_concepts.items():
    print(f'  {k:20s}: {v}')
```

```python
# PySpark setup (local mode)
# pip install pyspark

# from pyspark.sql import SparkSession
# spark = SparkSession.builder.master('local[*]').appName('BootcampSpark').getOrCreate()

# Simulate PySpark operations with pandas for demonstration
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'customer_id': range(1000),
    'region': np.random.choice(['East', 'West', 'North', 'South'], 1000),
    'revenue': np.random.exponential(scale=500, size=1000),
    'transactions': np.random.poisson(lam=5, size=1000)
})
print(df.head())
print(f'Shape: {df.shape}')
```

```python
# PySpark-equivalent operations (shown with pandas)
# In PySpark: df.groupBy('region').agg(F.sum('revenue'), F.avg('transactions'))

region_summary = df.groupby('region').agg(
    total_revenue=('revenue', 'sum'),
    avg_transactions=('transactions', 'mean'),
    customer_count=('customer_id', 'count')
).round(2)
print('Region Summary:')
print(region_summary)
```

```python
# Window functions (rank within region)
from pandas import Series

df['revenue_rank'] = df.groupby('region')['revenue'].rank(method='dense', ascending=False)
top_customers = df[df['revenue_rank'] <= 3].sort_values(['region', 'revenue_rank'])
print('Top 3 customers per region by revenue:')
print(top_customers[['region', 'customer_id', 'revenue', 'revenue_rank']].head(12))
```
