# Transformers and Fine-Tuning with Hugging Face
## Learning Objectives
- Understand the transformer architecture (attention mechanism)
- Use pretrained models from Hugging Face Hub
- Fine-tune BERT for text classification


```python
# pip install transformers datasets torch
# Transformer concepts
import numpy as np

attention_concepts = {
    'Self-Attention': 'Each token attends to all other tokens in the sequence',
    'Multi-Head Attention': 'Multiple attention heads capture different relationships',
    'Positional Encoding': 'Adds position information since Transformers have no recurrence',
    'Feed-Forward Network': 'Applied to each position after attention',
    'Layer Normalization': 'Stabilizes training in deep transformer networks',
    'BERT': 'Bidirectional Encoder - pre-trained on masked LM + NSP',
    'GPT': 'Autoregressive decoder - pre-trained on causal LM'
}
for k, v in attention_concepts.items():
    print(f'  {k:25s}: {v}')
```

```python
# Using Hugging Face pipeline (zero-shot)
# from transformers import pipeline

# Text classification
# classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
# results = classifier(['I love data science!', 'This dataset is terrible.'])
# for r in results:
#     print(f"Label: {r['label']}, Score: {r['score']:.4f}")

# Demo output
demo_results = [
    {'text': 'I love data science!', 'label': 'POSITIVE', 'score': 0.9998},
    {'text': 'This dataset is terrible.', 'label': 'NEGATIVE', 'score': 0.9995}
]
for r in demo_results:
    print(f"Text: {r['text'][:40]:40s} -> {r['label']} ({r['score']:.4f})")
```

```python
# Fine-tuning workflow
fine_tuning_steps = '''
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset

# 1. Load dataset and tokenizer
dataset = load_dataset('imdb')
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

# 2. Tokenize
def tokenize(batch):
    return tokenizer(batch['text'], truncation=True, padding=True, max_length=512)
tokenized = dataset.map(tokenize, batched=True)

# 3. Load model
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)

# 4. Training arguments
args = TrainingArguments(output_dir='./results', num_train_epochs=3, per_device_train_batch_size=16)

# 5. Train
trainer = Trainer(model=model, args=args, train_dataset=tokenized['train'], eval_dataset=tokenized['test'])
trainer.train()
'''
print('Fine-tuning workflow:')
print(fine_tuning_steps)
```

```python
# Attention visualization (conceptual)
def scaled_dot_product_attention(Q, K, V):
    """Simplified attention mechanism."""
    d_k = Q.shape[-1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)
    weights = np.exp(scores) / np.exp(scores).sum(axis=-1, keepdims=True)  # softmax
    return np.dot(weights, V), weights

np.random.seed(42)
Q = np.random.randn(5, 64)  # 5 query tokens, 64-dim
K = np.random.randn(5, 64)  # 5 key tokens
V = np.random.randn(5, 64)  # 5 value tokens
output, attn_weights = scaled_dot_product_attention(Q, K, V)
print(f'Attention output shape: {output.shape}')
print(f'Attention weight matrix:\n{attn_weights.round(3)}')
```
