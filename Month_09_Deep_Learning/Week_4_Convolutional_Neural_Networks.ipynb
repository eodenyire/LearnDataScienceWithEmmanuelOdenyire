```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import convolve2d

try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    TF = True
    print(f'TensorFlow {tf.__version__}')
except ImportError:
    TF = False
    print('TensorFlow not installed.')
```

## 1. Convolution — Intuition

```python
image = np.zeros((5, 5))
image[1:4, 1:4] = 1.0
kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]], dtype=float)
conv_out = convolve2d(image, kernel, mode='same')

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
axes[0].imshow(image, cmap='gray');    axes[0].set_title('Input')
axes[1].imshow(kernel, cmap='RdBu');   axes[1].set_title('Kernel')
axes[2].imshow(conv_out, cmap='gray'); axes[2].set_title('Output')
for ax in axes: ax.axis('off')
plt.tight_layout(); plt.show()
```

## 2. CNN for MNIST with Keras

```python
if TF:
    (X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
    X_train = X_train.astype('float32')[..., np.newaxis] / 255.0
    X_test  = X_test.astype('float32')[..., np.newaxis] / 255.0

    cnn = keras.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(10, activation='softmax')
    ])
    cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    cnn.summary()
else:
    print('TensorFlow not available.')
```

```python
if TF:
    history = cnn.fit(X_train[:10000], y_train[:10000], epochs=5,
                       batch_size=128, validation_split=0.1, verbose=1)
    _, test_acc = cnn.evaluate(X_test, y_test, verbose=0)
    print(f'Test Accuracy: {test_acc:.4f}')
else:
    print('TensorFlow not available.')
```

## 3. Transfer Learning (conceptual)

```python
if TF:
    base = keras.applications.MobileNetV2(input_shape=(96,96,3), include_top=False, weights=None)
    base.trainable = False
    inputs = keras.Input(shape=(96,96,3))
    x = base(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    outputs = layers.Dense(10, activation='softmax')(x)
    transfer = keras.Model(inputs, outputs)
    print('Trainable params:', sum(tf.size(p).numpy() for p in transfer.trainable_variables))
else:
    print('TensorFlow not available.')
```
