```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import (SelectKBest, f_classif,
                                          RFE, SelectFromModel)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

sns.set_theme(style='whitegrid')
```

## 1. Filter Methods — SelectKBest

```python
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42)

selector = SelectKBest(f_classif, k=10)
X_train_k = selector.fit_transform(X_train, y_train)
selected = np.array(cancer.feature_names)[selector.get_support()]
print('Selected features:')
for f in selected: print(' -', f)
```

## 2. RFE — Recursive Feature Elimination

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rfe = RFE(estimator=rf, n_features_to_select=10)
rfe.fit(X_train, y_train)
rfe_features = np.array(cancer.feature_names)[rfe.support_]
print('RFE selected:', rfe_features)
```

## 3. Comparing Approaches

```python
rf_full = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
X_test_k = selector.transform(X_test)
rf_k = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train_k, y_train)

print(f'Full features AUC:     {roc_auc_score(y_test, rf_full.predict_proba(X_test)[:,1]):.4f}')
print(f'SelectKBest (10) AUC:  {roc_auc_score(y_test, rf_k.predict_proba(X_test_k)[:,1]):.4f}')
```

## 4. Feature Engineering

```python
rng = np.random.default_rng(42)
n = 500
df = pd.DataFrame({
    'length': rng.uniform(10, 100, n),
    'width':  rng.uniform(5, 50, n),
    'height': rng.uniform(2, 20, n),
    'weight': rng.uniform(0.5, 10, n),
})
df['volume']       = df['length'] * df['width'] * df['height']
df['density']      = df['weight'] / df['volume']
df['aspect_ratio'] = df['length'] / df['width']
df['log_weight']   = np.log1p(df['weight'])
print(df.head(3).round(3))
```
