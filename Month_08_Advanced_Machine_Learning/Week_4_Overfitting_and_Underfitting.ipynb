# Week 4: Handling Overfitting and Underfitting

## Learning Objectives
- Diagnose with learning curves
- Apply regularisation techniques
- Compare regularised vs non-regularised models
- Use cross-validation for reliable evaluation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer, make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, Ridge, Lasso
from sklearn.model_selection import (train_test_split, learning_curve,
                                      validation_curve)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Demonstrating Overfitting

```python
X, y = make_classification(n_samples=200, n_features=20, n_informative=5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('Depth | Train Acc | Test Acc | Gap')
for depth in [1, 3, 5, 10, None]:
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)
    dt.fit(X_train, y_train)
    tr = dt.score(X_train, y_train)
    te = dt.score(X_test, y_test)
    print(f'{str(depth):5s} | {tr:.4f}    | {te:.4f}   | {tr-te:.4f}')
```

## 2. Regularisation for Linear Models

```python
from sklearn.datasets import make_regression
X_reg, y_reg = make_regression(n_samples=200, n_features=30, n_informative=5,
                                noise=20, random_state=42)
X_tr, X_te, y_tr, y_te = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_tr_s = scaler.fit_transform(X_tr)
X_te_s = scaler.transform(X_te)

for name, model in [('No regularisation', Ridge(alpha=0)),
                     ('Ridge alpha=1',     Ridge(alpha=1)),
                     ('Ridge alpha=100',   Ridge(alpha=100)),
                     ('Lasso alpha=0.1',   Lasso(alpha=0.1))]:
    model.fit(X_tr_s, y_tr)
    tr_r2 = r2_score(y_tr, model.predict(X_tr_s))
    te_r2 = r2_score(y_te, model.predict(X_te_s))
    print(f'{name:25s}: Train={tr_r2:.4f}, Test={te_r2:.4f}')
```

## 3. Learning Curve Comparison

```python
cancer = load_breast_cancer()
X_c, y_c = cancer.data, cancer.target

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, (model, title) in zip(axes, [
    (DecisionTreeClassifier(max_depth=None, random_state=42), 'Overfitting Tree'),
    (RandomForestClassifier(max_depth=5, n_estimators=50, random_state=42), 'Regularised RF')
]):
    ts, train_s, val_s = learning_curve(model, X_c, y_c, cv=5, scoring='accuracy',
                                         train_sizes=np.linspace(0.1, 1.0, 10))
    ax.plot(ts, train_s.mean(1), label='Train')
    ax.plot(ts, val_s.mean(1), label='Validation')
    ax.set_title(title); ax.legend()
plt.tight_layout(); plt.show()
```
