# Week 1: Setting Up Your Data Science Environment
## Learning Objectives

```python
# Verify Python environment
import sys
print(f'Python version: {sys.version}')
print('Environment ready!')
```

```python
# Install and verify core libraries
import importlib
libraries = ['numpy', 'pandas', 'matplotlib', 'sklearn']
for lib in libraries:
    try:
        importlib.import_module(lib)
        print(f'  ✓ {lib} installed')
    except ImportError:
        print(f'  ✗ {lib} NOT installed')
```

```python
# Data science mindset: curiosity-driven exploration
import numpy as np

# Ask a question -> collect data -> explore -> model -> communicate
data = np.random.normal(loc=50, scale=10, size=100)
print(f'Sample mean: {data.mean():.2f}')
print(f'Sample std:  {data.std():.2f}')
print(f'Min: {data.min():.2f}, Max: {data.max():.2f}')
```

```python
# Jupyter tips: keyboard shortcuts
shortcuts = {
    'Shift+Enter': 'Run cell and move to next',
    'Ctrl+Enter': 'Run cell in place',
    'A': 'Insert cell above (command mode)',
    'B': 'Insert cell below (command mode)',
    'D+D': 'Delete cell (command mode)',
    'M': 'Convert to Markdown (command mode)',
    'Y': 'Convert to Code (command mode)'
}
for k, v in shortcuts.items():
    print(f'  {k:20s} -> {v}')
```
