```python
# Git workflow simulation using Python
# (In practice, run these in your terminal)
git_commands = [
    'git init my-ds-project',
    'cd my-ds-project',
    'git add .',
    'git commit -m "Initial commit"',
    'git remote add origin https://github.com/username/my-ds-project.git',
    'git push -u origin main'
]
print('Essential Git commands:')
for cmd in git_commands:
    print(f'  $ {cmd}')
```

```python
# Understanding .gitignore for data science projects
gitignore_content = '''
# Python
__pycache__/
*.py[cod]
.env

# Jupyter
.ipynb_checkpoints/

# Data (large files)
data/raw/*.csv
*.parquet

# Models
models/*.pkl
'''
print('Sample .gitignore for data science:')
print(gitignore_content)
```

```python
# Branching strategy for data science
branch_strategy = {
    'main': 'Production-ready code',
    'develop': 'Integration branch',
    'feature/eda-analysis': 'Feature branch for EDA',
    'experiment/random-forest': 'Experiment branch',
    'hotfix/data-bug': 'Quick bug fix'
}
for branch, desc in branch_strategy.items():
    print(f'  {branch:35s} -> {desc}')
```
