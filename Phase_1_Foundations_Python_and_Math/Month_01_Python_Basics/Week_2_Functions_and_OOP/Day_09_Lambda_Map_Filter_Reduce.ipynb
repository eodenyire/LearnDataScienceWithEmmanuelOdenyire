## 🌅 Morning Session — Core Concepts

```python
# Lambda functions
square = lambda x: x**2
add = lambda x, y: x + y
print(square(5))  # 25
print(add(3, 4))  # 7
```

```python
# map() - apply function to all items
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print('Squared:', squared)
```

```python
# filter() - keep items matching condition
evens = list(filter(lambda x: x % 2 == 0, range(1, 21)))
print('Evens 1-20:', evens)
```

```python
from functools import reduce
# reduce() - combine items into single value
total = reduce(lambda acc, x: acc + x, [1, 2, 3, 4, 5])
product = reduce(lambda acc, x: acc * x, [1, 2, 3, 4, 5])
print(f'Sum: {total}, Product: {product}')
```

## ☀️ Afternoon Session — Practice Exercises

```python
# Practice: Sort list of dicts by a key
students = [
    {'name': 'Alice', 'gpa': 3.8},
    {'name': 'Bob', 'gpa': 3.2},
    {'name': 'Carol', 'gpa': 3.9},
]
sorted_students = sorted(students, key=lambda s: s['gpa'], reverse=True)
for s in sorted_students:
    print(f"{s['name']}: {s['gpa']}")
```

## 🌙 Evening Session — Projects & Review

```python
# Evening: Data pipeline using map/filter/reduce
data = [15, -3, 42, 7, -8, 23, 0, 55, -1, 30]
# Step 1: filter positives
pos = list(filter(lambda x: x > 0, data))
# Step 2: square them
squared = list(map(lambda x: x**2, pos))
# Step 3: compute sum
total = reduce(lambda a, b: a + b, squared)
print(f'Positive values: {pos}')
print(f'Squared: {squared}')
print(f'Sum of squares: {total}')
```
