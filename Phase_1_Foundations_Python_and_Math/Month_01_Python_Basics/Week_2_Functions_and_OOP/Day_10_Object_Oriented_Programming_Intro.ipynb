## 🌅 Morning Session — Core Concepts

```python
# Defining a class
class DataPoint:
    """Represents a single data observation."""
    
    def __init__(self, label, value):
        self.label = label
        self.value = value
    
    def __str__(self):
        return f'DataPoint({self.label}={self.value})'
    
    def is_above(self, threshold):
        return self.value > threshold

dp = DataPoint('Temperature', 37.5)
print(dp)
print(dp.is_above(36.5))
```

```python
# Class with multiple methods
class Dataset:
    def __init__(self, name):
        self.name = name
        self.data = []
    
    def add(self, value):
        self.data.append(value)
    
    def mean(self):
        return sum(self.data) / len(self.data) if self.data else 0
    
    def summary(self):
        return {
            'name': self.name,
            'count': len(self.data),
            'mean': self.mean(),
            'min': min(self.data),
            'max': max(self.data)
        }

ds = Dataset('Sales')
for v in [10, 25, 30, 15, 42, 8, 19]:
    ds.add(v)
print(ds.summary())
```

## ☀️ Afternoon Session — Practice Exercises

### Practice: Create a `Student` class with:
- `name`, `student_id`, `grades` (list)
- Methods: `add_grade()`, `average()`, `letter_grade()`, `__str__()`

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = []
    
    def add_grade(self, grade):
        self.grades.append(grade)
    
    def average(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0
    
    def letter_grade(self):
        avg = self.average()
        return 'A' if avg >= 90 else 'B' if avg >= 80 else 'C' if avg >= 70 else 'D' if avg >= 60 else 'F'
    
    def __str__(self):
        return f'Student({self.name}, ID={self.student_id}, GPA={self.average():.2f} [{self.letter_grade()}])'

s = Student('Emmanuel', 'DS2026001')
for g in [88, 92, 85, 95, 90]:
    s.add_grade(g)
print(s)
```
