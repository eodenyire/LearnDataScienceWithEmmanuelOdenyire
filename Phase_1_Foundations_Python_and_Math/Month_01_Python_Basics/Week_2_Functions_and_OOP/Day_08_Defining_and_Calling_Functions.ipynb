## 🌅 Morning Session — Core Concepts

```python
# Defining functions
def greet(name, greeting='Hello'):
    """Greet a person by name."""
    return f'{greeting}, {name}!'

print(greet('Emmanuel'))
print(greet('Alice', 'Hi'))
```

```python
# Multiple return values
def min_max(numbers):
    """Return the min and max of a list."""
    return min(numbers), max(numbers)

data = [3, 1, 4, 1, 5, 9, 2, 6]
lo, hi = min_max(data)
print(f'Min: {lo}, Max: {hi}')
```

```python
# *args and **kwargs
def summarize(*args, **kwargs):
    print('Positional args:', args)
    print('Keyword args:', kwargs)

summarize(1, 2, 3, name='Data Science', year=2026)
```

## ☀️ Afternoon Session — Practice Exercises

### Practice Problems
1. Write a function that checks if a number is prime
2. Write a recursive function to compute factorial
3. Write a function with `*args` that computes the average of any number of values

```python
# Exercise 1: Prime checker
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

primes = [n for n in range(2, 50) if is_prime(n)]
print('Primes up to 50:', primes)
```

```python
# Exercise 2: Factorial (recursive)
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)

for i in range(10):
    print(f'{i}! = {factorial(i)}')
```

```python
# Exercise 3: Variable args average
def average(*values):
    return sum(values) / len(values)

print(average(10, 20, 30, 40, 50))
```

### Evening Project: Function Library
Create a module of useful math/utility functions:
- `is_palindrome(s)` – check if string is a palindrome
- `flatten(lst)` – flatten a nested list
- `chunk(lst, size)` – split a list into chunks

```python
def is_palindrome(s):
    s = s.lower().replace(' ', '')
    return s == s[::-1]

test_words = ['racecar', 'hello', 'madam', 'python', 'level']
for word in test_words:
    print(f"'{word}' is palindrome: {is_palindrome(word)}")
```

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

nested = [1, [2, 3], [4, [5, 6]], 7]
print('Flattened:', flatten(nested))
```
