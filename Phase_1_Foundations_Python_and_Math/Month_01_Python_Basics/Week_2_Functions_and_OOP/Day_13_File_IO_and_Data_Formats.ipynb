## 🌅 Morning Session — Core Concepts

```python
import json
import csv
from pathlib import Path

# Writing a JSON file
data = {'bootcamp': 'Data Science', 'year': 2026, 'phases': 3}
with open('/tmp/bootcamp.json', 'w') as f:
    json.dump(data, f, indent=2)

# Reading it back
with open('/tmp/bootcamp.json', 'r') as f:
    loaded = json.load(f)
print('JSON loaded:', loaded)
```

```python
# CSV operations
students = [['name', 'score', 'grade'], ['Alice', 92, 'A'], ['Bob', 78, 'C']]
with open('/tmp/students.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(students)

with open('/tmp/students.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(dict(row))
```
