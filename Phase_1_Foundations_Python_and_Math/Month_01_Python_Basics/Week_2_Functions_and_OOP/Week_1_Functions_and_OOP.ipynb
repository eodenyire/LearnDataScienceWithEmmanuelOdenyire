# Week 1: Python Functions and Object-Oriented Programming
## Learning Objectives
- Write reusable functions with parameters and return values
- Use decorators and lambda functions
- Implement classes with attributes and methods
- Apply inheritance and polymorphism


```python
# Functions: basics
def calculate_statistics(data):
    """Calculate basic statistics for a dataset."""
    import statistics
    return {
        'mean': statistics.mean(data),
        'median': statistics.median(data),
        'stdev': statistics.stdev(data),
        'min': min(data),
        'max': max(data)
    }

scores = [85, 92, 78, 95, 88, 76, 91, 83, 79, 94]
stats = calculate_statistics(scores)
for k, v in stats.items():
    print(f'  {k:8s}: {v:.2f}')
```

```python
# Lambda functions and higher-order functions
data = [{'name': 'Alice', 'score': 92}, {'name': 'Bob', 'score': 85},
        {'name': 'Carol', 'score': 78}, {'name': 'Dave', 'score': 95}]

# Sort by score descending
sorted_data = sorted(data, key=lambda x: x['score'], reverse=True)
print('Ranked by score:')
for rank, student in enumerate(sorted_data, 1):
    print(f'  {rank}. {student["name"]:10s}: {student["score"]}')

# Filter and map
high_scorers = list(filter(lambda x: x['score'] >= 90, data))
print(f'\nHigh scorers (≥90): {[s["name"] for s in high_scorers]}')
```

```python
# Object-Oriented Programming
class DataSet:
    def __init__(self, name, data):
        self.name = name
        self.data = data
        self._stats = None  # lazy computation

    @property
    def stats(self):
        if self._stats is None:
            import statistics
            self._stats = {'mean': statistics.mean(self.data), 'std': statistics.stdev(self.data)}
        return self._stats

    def __len__(self):
        return len(self.data)

    def __repr__(self):
        return f'DataSet(name={self.name!r}, n={len(self)})'

ds = DataSet('exam_scores', [85, 92, 78, 95, 88])
print(ds)
print(f'Stats: {ds.stats}')
```

```python
# Inheritance
class NumericalDataSet(DataSet):
    def normalize(self):
        mn, mx = min(self.data), max(self.data)
        return [(x - mn) / (mx - mn) for x in self.data]

    def zscore(self):
        mean = self.stats['mean']
        std = self.stats['std']
        return [(x - mean) / std for x in self.data]

nds = NumericalDataSet('temps', [20, 25, 30, 22, 28, 35])
print(f'Normalized: {[round(v,3) for v in nds.normalize()]}')
print(f'Z-scores:   {[round(v,3) for v in nds.zscore()]}')
```
