```python
# Week 2 Integration Project: Data Science Toolkit Class
class DataScienceToolkit:
    """
    A simple data analysis toolkit demonstrating OOP concepts.
    Supports loading, cleaning, and summarizing data.
    """
    def __init__(self, name):
        self.name = name
        self.data = []
    
    def load_from_list(self, values):
        self.data = [v for v in values if v is not None]
        print(f'Loaded {len(self.data)} values')
    
    def clean(self):
        before = len(self.data)
        self.data = list(set(self.data))  # remove duplicates
        print(f'Removed {before - len(self.data)} duplicates')
    
    def stats(self):
        if not self.data:
            return {}
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        mean = sum(sorted_data) / n
        median = sorted_data[n//2] if n % 2 else (sorted_data[n//2-1] + sorted_data[n//2]) / 2
        return {'count': n, 'mean': round(mean, 2), 'median': median,
                'min': sorted_data[0], 'max': sorted_data[-1]}
    
    def __repr__(self):
        return f'DataScienceToolkit(name={self.name!r}, records={len(self.data)})'

toolkit = DataScienceToolkit('SalesAnalyzer')
toolkit.load_from_list([23, 45, 12, 45, 67, None, 34, 12, 89, 23, None])
toolkit.clean()
print(toolkit)
print('Stats:', toolkit.stats())
```
