## Learning Objectives
- Master `for` and `while` loops
- Use `range()`, `enumerate()`, and `zip()`
- Understand `break`, `continue`, and `pass`
- Write list comprehensions

```python
# For loop basics
for i in range(10):
    print(f'Iteration {i}')
```

```python
# While loop
count = 0
while count < 5:
    print(f'Count: {count}')
    count += 1
```

```python
# List comprehension
squares = [x**2 for x in range(1, 11)]
print('Squares:', squares)
```

```python
# enumerate() and zip()
fruits = ['apple', 'banana', 'cherry']
for idx, fruit in enumerate(fruits, start=1):
    print(f'{idx}. {fruit}')

names = ['Alice', 'Bob']
scores = [95, 87]
for name, score in zip(names, scores):
    print(f'{name}: {score}')
```

## Afternoon Practice Exercises

1. Write a loop that prints all even numbers from 1-50
2. Create a multiplication table using nested loops
3. Use list comprehension to filter words longer than 4 characters
4. Find all prime numbers up to 100 using a loop

```python
# Exercise 1: Even numbers
evens = [n for n in range(1, 51) if n % 2 == 0]
print('Even numbers 1-50:', evens)
```

```python
# Exercise 2: Multiplication table
print('Multiplication Table (1-5):')
for i in range(1, 6):
    row = '  '.join(f'{i*j:3}' for j in range(1, 6))
    print(row)
```

## Evening Mini-Project: Word Frequency Counter

Build a word frequency counter that:
- Takes a paragraph as input
- Counts how often each word appears
- Displays the top 5 most common words

```python
# Mini-Project: Word Frequency Counter
from collections import Counter

text = """data science is the study of data to extract meaningful insights
data scientists use python for data analysis and machine learning"""

words = text.lower().split()
word_count = Counter(words)

print('Top 5 most common words:')
for word, count in word_count.most_common(5):
    print(f'  {word}: {count} times')
```
