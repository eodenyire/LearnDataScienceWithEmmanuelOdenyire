## Week 1 Capstone Challenge

**Build a Simple Student Grade Calculator:**
- Store 5 students with names and scores
- Calculate average, min, max
- Assign letter grades
- Print a formatted report

```python
# Week 1 Capstone: Student Grade Calculator
students = [
    {'name': 'Alice', 'scores': [85, 92, 78, 90, 88]},
    {'name': 'Bob', 'scores': [70, 75, 80, 65, 72]},
    {'name': 'Carol', 'scores': [95, 98, 92, 97, 94]},
    {'name': 'David', 'scores': [55, 60, 58, 62, 57]},
    {'name': 'Emma', 'scores': [80, 82, 85, 79, 84]},
]

def get_letter_grade(avg):
    if avg >= 90: return 'A'
    elif avg >= 80: return 'B'
    elif avg >= 70: return 'C'
    elif avg >= 60: return 'D'
    else: return 'F'

print('=' * 50)
print(f'{'Student':<12} {'Average':>8} {'Min':>6} {'Max':>6} {'Grade':>6}')
print('=' * 50)

all_averages = []
for student in students:
    avg = sum(student['scores']) / len(student['scores'])
    all_averages.append(avg)
    letter = get_letter_grade(avg)
    print(f"{student['name']:<12} {avg:>8.2f} {min(student['scores']):>6} {max(student['scores']):>6} {letter:>6}")

print('=' * 50)
print(f"{'Class Average':<12} {sum(all_averages)/len(all_averages):>8.2f}")
print('=' * 50)
```
