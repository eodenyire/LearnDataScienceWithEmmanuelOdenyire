## 🌅 Morning Session — Core Concepts

```python
# List operations
fruits = ['banana', 'apple', 'cherry', 'mango']
fruits.sort()
print('Sorted:', fruits)

# Slicing
print('First two:', fruits[:2])
print('Last two:', fruits[-2:])
print('Reversed:', fruits[::-1])
```

```python
# Named tuples
from collections import namedtuple
Student = namedtuple('Student', ['name', 'gpa', 'major'])
s1 = Student('Alice', 3.8, 'Data Science')
print(s1)
print(f'Name: {s1.name}, GPA: {s1.gpa}')
```

## ☀️ Afternoon Session — Practice Exercises

```python
# Practice: Implement a stack and queue using lists
class Stack:
    def __init__(self):
        self._items = []
    def push(self, item): self._items.append(item)
    def pop(self): return self._items.pop()
    def peek(self): return self._items[-1] if self._items else None
    def is_empty(self): return len(self._items) == 0

s = Stack()
for v in [1, 2, 3, 4, 5]:
    s.push(v)
print('Peek:', s.peek())
while not s.is_empty():
    print('Pop:', s.pop())
```
