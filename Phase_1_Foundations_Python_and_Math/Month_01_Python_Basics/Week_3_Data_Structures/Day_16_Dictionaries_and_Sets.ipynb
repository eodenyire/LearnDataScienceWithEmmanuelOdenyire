## 🌅 Morning Session — Core Concepts

```python
from collections import defaultdict, Counter

# Dictionary operations
word_freq = {}
text = 'data science machine learning data python machine data'
for word in text.split():
    word_freq[word] = word_freq.get(word, 0) + 1
print('Word frequencies:', word_freq)

# Counter - same thing, cleaner
counter = Counter(text.split())
print('Top 3:', counter.most_common(3))
```

```python
# Set operations
python_skills = {'python', 'pandas', 'numpy', 'matplotlib'}
ml_skills = {'python', 'scikit-learn', 'tensorflow', 'pandas'}

print('Common:', python_skills & ml_skills)
print('All skills:', python_skills | ml_skills)
print('Only Python skills:', python_skills - ml_skills)
```
