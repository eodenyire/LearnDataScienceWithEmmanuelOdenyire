## 🌅 Morning Session — Core Concepts

```python
# Generator function
def count_up(start, end):
    n = start
    while n <= end:
        yield n
        n += 1

for num in count_up(1, 5):
    print(num, end=' ')
print()
```

```python
import itertools
# itertools examples
print('Chain:', list(itertools.chain([1, 2], [3, 4], [5])))
print('Cycle (first 8):', list(itertools.islice(itertools.cycle([1, 2, 3]), 8)))
print('Combinations:', list(itertools.combinations([1,2,3,4], 2)))
```
