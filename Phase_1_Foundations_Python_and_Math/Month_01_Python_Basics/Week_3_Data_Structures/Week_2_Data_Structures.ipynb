# Week 2: Python Data Structures
## Learning Objectives
- Master lists, tuples, sets, and dictionaries
- Use comprehensions for concise transformations
- Understand when to use each structure
- Apply built-in algorithms (sorting, searching)


```python
# Lists: ordered, mutable
import time

students = ['Alice', 'Bob', 'Carol', 'Dave', 'Eve']
students.append('Frank')           # O(1)
students.insert(2, 'Grace')        # O(n)
students.remove('Bob')             # O(n)
print(f'Students: {students}')

# List comprehension
squares = [x**2 for x in range(10) if x % 2 == 0]
print(f'Even squares: {squares}')

# Slicing
print(f'First 3: {students[:3]}, Last 2: {students[-2:]}, Reversed: {students[::-1]}')
```

```python
# Tuples: ordered, immutable (great for records)
coordinate = (40.7128, -74.0060, 'New York')  # lat, lon, city
lat, lon, city = coordinate  # unpacking
print(f'City: {city}, Location: ({lat}, {lon})')

# Named tuples for structured data
from collections import namedtuple
Student = namedtuple('Student', ['name', 'score', 'grade'])
students_records = [Student('Alice', 92, 'A'), Student('Bob', 75, 'B'), Student('Carol', 88, 'B+')]
for s in students_records:
    print(f'  {s.name:10s}: {s.score} ({s.grade})')
```

```python
# Dictionaries: key-value, O(1) lookup
word_freq = {}
text = 'data science is the science of extracting insights from data'
for word in text.split():
    word_freq[word] = word_freq.get(word, 0) + 1

# Sort by frequency
sorted_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
print('Word frequencies:')
for word, count in sorted_freq[:5]:
    print(f'  {word:12s}: {count}')

# Dict comprehension
lengths = {word: len(word) for word in text.split()}
print(f'\nAvg word length: {sum(lengths.values())/len(lengths):.1f}')
```

```python
# Sets: unordered, unique elements
ml_skills = {'Python', 'Statistics', 'ML', 'SQL', 'Visualization'}
de_skills = {'Python', 'SQL', 'Spark', 'Kafka', 'Airflow'}

print(f'Common skills:  {ml_skills & de_skills}')
print(f'ML only:        {ml_skills - de_skills}')
print(f'All skills:     {ml_skills | de_skills}')
print(f'Not in both:    {ml_skills ^ de_skills}')

# Deduplication
duplicates = [1, 2, 2, 3, 3, 3, 4, 5, 5]
unique = list(set(duplicates))
print(f'\nUnique values: {sorted(unique)}')
```

```python
# Performance comparison: list vs set lookup
import time

big_list = list(range(1_000_000))
big_set = set(range(1_000_000))
target = 999_999

t0 = time.perf_counter()
_ = target in big_list
list_time = time.perf_counter() - t0

t0 = time.perf_counter()
_ = target in big_set
set_time = time.perf_counter() - t0

print(f'List lookup: {list_time*1000:.3f} ms (O(n))')
print(f'Set lookup:  {set_time*1000:.3f} ms (O(1))')
print(f'Speedup: {list_time/set_time:.1f}x')
```
