# Week 3: File Handling and Python Modules
## Learning Objectives
- Read and write text, CSV, and JSON files
- Handle exceptions gracefully
- Create and import custom modules
- Use popular standard library modules


```python
import os, json, csv

# Writing files
with open('/tmp/sample_data.txt', 'w') as f:
    f.write('Name,Score,Grade\n')
    f.write('Alice,92,A\n')
    f.write('Bob,75,B\n')
    f.write('Carol,88,B+\n')

# Reading files
with open('/tmp/sample_data.txt', 'r') as f:
    content = f.read()
print('File content:')
print(content)
```

```python
# CSV handling
import csv

# Write CSV
rows = [['Name','Score','Grade'],['Alice',92,'A'],['Bob',75,'B'],['Carol',88,'B+']]
with open('/tmp/students.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(rows)

# Read CSV with DictReader
with open('/tmp/students.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"  {row['Name']:10s}: Score={row['Score']}, Grade={row['Grade']}")
```

```python
# JSON handling
config = {
    'project': 'Data Science Bootcamp',
    'version': '1.0',
    'settings': {'epochs': 100, 'lr': 0.001, 'batch_size': 32},
    'features': ['age', 'income', 'education']
}

# Write JSON
with open('/tmp/config.json', 'w') as f:
    json.dump(config, f, indent=2)

# Read JSON
with open('/tmp/config.json', 'r') as f:
    loaded = json.load(f)
print(f"Project: {loaded['project']}")
print(f"Settings: {loaded['settings']}")
```

```python
# Exception handling
def safe_read_csv(filepath):
    try:
        import pandas as pd
        df = pd.read_csv(filepath)
        return df
    except FileNotFoundError:
        print(f'Error: File {filepath!r} not found')
    except pd.errors.EmptyDataError:
        print(f'Error: File {filepath!r} is empty')
    except Exception as e:
        print(f'Unexpected error: {type(e).__name__}: {e}')
    return None

# Test it
result = safe_read_csv('/tmp/students.csv')
if result is not None:
    print(result)

bad = safe_read_csv('/tmp/nonexistent.csv')
```

```python
# Standard library modules
import pathlib, datetime, collections

# pathlib for path operations
p = pathlib.Path('/tmp')
csvs = list(p.glob('*.csv'))
print(f'CSV files in /tmp: {[f.name for f in csvs]}')

# datetime
now = datetime.datetime.now()
print(f'Current time: {now.strftime("%Y-%m-%d %H:%M:%S")}')

# collections
from collections import Counter, defaultdict
words = 'the quick brown fox jumps over the lazy dog the'.split()
freq = Counter(words)
print(f'Most common: {freq.most_common(3)}')
```
