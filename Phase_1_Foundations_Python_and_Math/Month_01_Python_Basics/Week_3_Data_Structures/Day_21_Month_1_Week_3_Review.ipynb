```python
# Integration Challenge: Data Pipeline
from dataclasses import dataclass
from typing import List
from collections import Counter

@dataclass
class Customer:
    id: int
    name: str
    purchases: List[float]
    
    def total(self) -> float:
        return sum(self.purchases)
    
    def tier(self) -> str:
        t = self.total()
        return 'Gold' if t > 500 else 'Silver' if t > 200 else 'Bronze'

customers = [
    Customer(1, 'Alice', [120, 250, 90, 200]),
    Customer(2, 'Bob', [50, 75, 30]),
    Customer(3, 'Carol', [300, 450, 120]),
]

for c in sorted(customers, key=lambda x: x.total(), reverse=True):
    print(f'{c.name}: ${c.total():.0f} [{c.tier()}]')

tier_counts = Counter(c.tier() for c in customers)
print('\nTier Distribution:', dict(tier_counts))
```
