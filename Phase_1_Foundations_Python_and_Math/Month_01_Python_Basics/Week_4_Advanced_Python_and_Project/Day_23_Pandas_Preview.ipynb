## Practice: Build Your Month 1 Project

Create a **Personal Finance Tracker** that:
1. Stores income/expense records as dataclasses
2. Calculates monthly summaries
3. Identifies spending categories using Counter
4. Saves/loads data from JSON
5. Generates a text-based report

```python
from dataclasses import dataclass, field
from collections import defaultdict, Counter
from datetime import date
import json

@dataclass
class Transaction:
    date: str
    category: str
    amount: float
    description: str
    type: str = 'expense'  # 'income' or 'expense'

class FinanceTracker:
    def __init__(self):
        self.transactions = []
    
    def add(self, date, category, amount, description, ttype='expense'):
        self.transactions.append(Transaction(date, category, amount, description, ttype))
    
    def summary(self):
        income = sum(t.amount for t in self.transactions if t.type == 'income')
        expenses = sum(t.amount for t in self.transactions if t.type == 'expense')
        by_category = defaultdict(float)
        for t in self.transactions:
            if t.type == 'expense':
                by_category[t.category] += t.amount
        return {'income': income, 'expenses': expenses, 'net': income - expenses,
                'by_category': dict(by_category)}

tracker = FinanceTracker()
tracker.add('2026-01-05', 'Salary', 5000, 'Monthly salary', 'income')
tracker.add('2026-01-10', 'Food', 150, 'Groceries')
tracker.add('2026-01-15', 'Transport', 80, 'Bus pass')
tracker.add('2026-01-20', 'Education', 200, 'Bootcamp materials')

summary = tracker.summary()
print(f"Income: ${summary['income']:,.0f}")
print(f"Expenses: ${summary['expenses']:,.0f}")
print(f"Net: ${summary['net']:,.0f}")
print(f"By category: {summary['by_category']}")
```
