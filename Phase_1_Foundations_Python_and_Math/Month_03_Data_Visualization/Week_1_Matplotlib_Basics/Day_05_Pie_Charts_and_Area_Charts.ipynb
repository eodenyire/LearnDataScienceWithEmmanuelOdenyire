# Day 5: Pie Charts and Area Charts

## Phase 1 | Month 3 | Week 1 | Day 5

**🎯 Goal:** Create pie/donut charts for proportional data and stacked area charts for cumulative trends.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Pie chart — Kenya energy mix ──────────────────────────────────────────────
sources  = ['Geothermal', 'Hydro', 'Wind', 'Thermal', 'Solar', 'Other']
shares   = [43.9, 28.4, 11.4, 9.8, 4.2, 2.3]
explode  = [0.05, 0, 0, 0, 0, 0]   # explode geothermal slice
colors   = ['#E74C3C', '#3498DB', '#27AE60', '#F39C12', '#F1C40F', '#95A5A6']

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    shares, labels=sources, autopct='%1.1f%%', explode=explode,
    colors=colors, startangle=140, pctdistance=0.82,
    wedgeprops=dict(edgecolor='white', linewidth=1.5))

for at in autotexts:
    at.set_fontsize(9)
    at.set_fontweight('bold')

ax.set_title('Kenya Electricity Generation Mix 2023\n(Total: ~11,000 MW)', fontsize=13, fontweight='bold')
fig.tight_layout()
fig.savefig('/tmp/day05_pie.png', dpi=120)
print('Pie chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Donut chart ────────────────────────────────────────────────────────────────
mobile_ops  = ['Safaricom', 'Airtel Kenya', 'Telkom Kenya', 'Others']
market_share = [64.7, 24.3, 7.2, 3.8]
colors = ['#00A651', '#FF0000', '#0000FF', '#808080']

fig, ax = plt.subplots(figsize=(7, 6))
wedges, texts, autotexts = ax.pie(
    market_share, labels=mobile_ops, autopct='%1.1f%%',
    colors=colors, startangle=90, pctdistance=0.75,
    wedgeprops=dict(width=0.5, edgecolor='white', linewidth=2))  # width<1 = donut

ax.text(0, 0, 'Kenya\nMobile\nMarket', ha='center', va='center', fontsize=11, fontweight='bold')
ax.set_title('Kenya Mobile Market Share Q3 2023', fontsize=13, fontweight='bold')
fig.tight_layout()
fig.savefig('/tmp/day05_donut.png', dpi=120)
print('Donut chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Stacked area chart — Kenya banking sector assets ─────────────────────────
years = np.arange(2016, 2024)
commercial = np.array([3200, 3480, 3750, 3920, 4100, 4380, 4620, 4900])
microfinance = np.array([280, 310, 340, 365, 390, 420, 450, 480])
saccos       = np.array([600, 660, 730, 810, 890, 980, 1070, 1160])
mobile_banking = np.array([120, 200, 310, 430, 570, 730, 900, 1100])

fig, ax = plt.subplots(figsize=(11, 6))
ax.stackplot(years,
             commercial, microfinance, saccos, mobile_banking,
             labels=['Commercial Banks', 'Microfinance', 'SACCOs', 'Mobile Banking'],
             colors=['#2196F3', '#4CAF50', '#FF9800', '#9C27B0'],
             alpha=0.85)

ax.set_title('Kenya Financial Sector Assets (KES Billion) 2016–2023', fontsize=13, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Assets (KES Billion)')
ax.legend(loc='upper left')
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day05_area.png', dpi=120)
print('Stacked area chart saved')
plt.close()
```
