5. Adding value labels on bars

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Vertical bar chart ────────────────────────────────────────────────────────
countries  = ['Nigeria','Ethiopia','Egypt','DRC','Tanzania','Kenya','South Africa']
population = [223.8, 126.5, 104.2, 99.0, 65.5, 54.0, 59.6]
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(countries)))

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(countries, population, color=colors, edgecolor='white', linewidth=0.8)

# Add value labels on top
for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, h + 1.5, f'{h:.0f}M',
            ha='center', va='bottom', fontsize=9, fontweight='bold')

ax.set_title('Top African Countries by Population (2023)', fontsize=13, fontweight='bold')
ax.set_ylabel('Population (Millions)')
ax.set_ylim(0, 250)
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day03_bar.png', dpi=120)
print('Bar chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Grouped bar chart — Kenya county budgets ──────────────────────────────────
counties      = ['Nairobi', 'Mombasa', 'Kisumu', 'Nakuru', 'Eldoret']
dev_budget    = [15.2, 6.4, 4.8, 5.1, 4.2]   # KES billion
recurrent     = [28.5, 9.2, 6.9, 8.3, 6.5]

x = np.arange(len(counties))
w = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x - w/2, dev_budget, w, label='Development Budget', color='#2196F3', edgecolor='white')
ax.bar(x + w/2, recurrent,  w, label='Recurrent Budget',   color='#FF5722', edgecolor='white')

ax.set_xticks(x)
ax.set_xticklabels(counties)
ax.set_title('Kenya County Budgets (KES Billion) 2023/24', fontsize=13, fontweight='bold')
ax.set_ylabel('Budget (KES Billion)')
ax.legend()
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day03_grouped.png', dpi=120)
print('Grouped bar chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Histogram — M-Pesa transaction amounts ───────────────────────────────────
np.random.seed(42)
# Simulate transaction amounts: most are small (airtime top-ups), some large (rent)
transactions = np.concatenate([
    np.random.exponential(scale=500, size=8000),   # small transactions
    np.random.normal(loc=5000, scale=1000, size=1500),  # medium
    np.random.normal(loc=20000, scale=3000, size=500)   # large
])
transactions = transactions[transactions > 0]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Linear scale
axes[0].hist(transactions, bins=50, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].set_title('M-Pesa Transaction Amounts (Linear)', fontweight='bold')
axes[0].set_xlabel('Amount (KES)')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.3)

# Log scale
axes[1].hist(transactions, bins=50, color='tomato', edgecolor='white', alpha=0.8)
axes[1].set_xscale('log')
axes[1].set_title('M-Pesa Transaction Amounts (Log Scale)', fontweight='bold')
axes[1].set_xlabel('Amount (KES, log scale)')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.3)

fig.suptitle('Distribution of M-Pesa Transactions', fontsize=13, fontweight='bold')
fig.tight_layout()
fig.savefig('/tmp/day03_hist.png', dpi=120)
print(f'Histogram saved | Mean: KES {transactions.mean():,.0f} | Median: KES {np.median(transactions):,.0f}')
plt.close()
```
