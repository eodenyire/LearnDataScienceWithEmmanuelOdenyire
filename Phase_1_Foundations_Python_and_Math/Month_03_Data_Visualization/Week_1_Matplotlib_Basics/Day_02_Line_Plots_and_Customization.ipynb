```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Line style options ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle('Line Style Options', fontsize=14, fontweight='bold')

x = np.linspace(0, 10, 50)
styles = [('-', 'o', 'steelblue'), ('--', 's', 'tomato'), ('-.', '^', 'green'), (':', 'D', 'purple')]
labels = ['Solid (-)', 'Dashed (--)', 'Dash-dot (-.)', 'Dotted (:)']

for ax, (ls, mk, col), lbl in zip(axes.flat, styles, labels):
    ax.plot(x, np.sin(x + np.random.rand()), linestyle=ls, marker=mk,
            color=col, markevery=5, linewidth=2, label=lbl)
    ax.set_title(lbl)
    ax.grid(True, alpha=0.3)

fig.tight_layout()
fig.savefig('/tmp/day02_styles.png', dpi=120)
print('Line styles figure saved')
plt.close()
```

## Dual-Axis (Twin Axes) Plots

Used when two metrics have very different scales — e.g., M-Pesa transaction **volume** (millions) and **value** (KES billions) on the same time axis.

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

quarters = ['Q1-21','Q2-21','Q3-21','Q4-21','Q1-22','Q2-22','Q3-22','Q4-22','Q1-23']
mpesa_volume = [500, 540, 575, 610, 645, 680, 710, 750, 795]   # millions of transactions
mpesa_value  = [1800, 1950, 2100, 2250, 2380, 2500, 2620, 2780, 2950]  # KES billions

fig, ax1 = plt.subplots(figsize=(11, 5))

color1 = 'steelblue'
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Transaction Volume (Millions)', color=color1)
ax1.plot(quarters, mpesa_volume, 'o-', color=color1, linewidth=2, markersize=6)
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tomato'
ax2.set_ylabel('Transaction Value (KES Billions)', color=color2)
ax2.plot(quarters, mpesa_value, 's--', color=color2, linewidth=2, markersize=6)
ax2.tick_params(axis='y', labelcolor=color2)

ax1.set_title('M-Pesa Growth: Volume vs Value (2021-2023)', fontsize=13, fontweight='bold')
plt.xticks(rotation=30)
fig.tight_layout()
fig.savefig('/tmp/day02_twin.png', dpi=120)
print('Twin-axis M-Pesa chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# Custom tick labels and limits
nse_dates = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
safaricom_price = [34.2, 35.1, 33.8, 32.5, 33.0, 34.5, 36.2, 37.0, 35.8, 34.2, 35.5, 36.8]
equity_price    = [48.5, 50.2, 47.8, 46.0, 47.5, 49.0, 51.2, 52.5, 50.0, 48.8, 50.0, 52.0]

fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(nse_dates, safaricom_price, 'o-', color='#00A651', linewidth=2.5, label='Safaricom (SCOM)', markersize=5)
ax.plot(nse_dates, equity_price,    's-', color='#003087', linewidth=2.5, label='Equity Bank (EQTY)', markersize=5)

ax.fill_between(nse_dates, safaricom_price, alpha=0.07, color='#00A651')
ax.fill_between(nse_dates, equity_price,    alpha=0.07, color='#003087')

ax.set_title('NSE Stock Prices 2023 — Safaricom vs Equity Bank', fontsize=13, fontweight='bold')
ax.set_xlabel('Month 2023')
ax.set_ylabel('Share Price (KES)')
ax.set_ylim(28, 58)
ax.legend(loc='upper right')
ax.grid(True, linestyle='--', alpha=0.35)
fig.tight_layout()
fig.savefig('/tmp/day02_nse.png', dpi=120)
print('NSE stock chart saved')
plt.close()
```
