# Day 4: Scatter Plots and Bubble Charts

## Phase 1 | Month 3 | Week 1 | Day 4

**🎯 Goal:** Visualise relationships between two continuous variables, encode a third variable using size, and a fourth using color.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Basic scatter: GDP vs Life Expectancy ─────────────────────────────────────
np.random.seed(7)
countries = ['Nigeria','Kenya','South Africa','Egypt','Ghana','Ethiopia',
             'Tanzania','Uganda','Rwanda','Senegal','Côte d\'Ivoire','Zambia']
gdp_pc  = [2086, 2082, 7055, 3500, 2202, 925,  1108, 817,  930,  1529, 2286, 1305]
life_ex = [53,   67,   64,   72,   64,   68,   67,   63,   69,   68,   58,   63]
pop     = [223.8, 54.0, 59.6, 104.2, 33.5, 126.5, 65.5, 47.2, 14.1, 17.2, 27.5, 19.5]

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(gdp_pc, life_ex, s=[p*3 for p in pop],
                     c=life_ex, cmap='RdYlGn', alpha=0.75, edgecolors='grey', linewidth=0.5)

for i, country in enumerate(countries):
    ax.annotate(country, (gdp_pc[i], life_ex[i]),
                textcoords='offset points', xytext=(5, 3), fontsize=8)

cb = fig.colorbar(scatter, ax=ax, pad=0.01)
cb.set_label('Life Expectancy (years)')
ax.set_xlabel('GDP per Capita (USD)')
ax.set_ylabel('Life Expectancy (years)')
ax.set_title('African Countries: GDP per Capita vs Life Expectancy (bubble = population)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.25)
fig.tight_layout()
fig.savefig('/tmp/day04_bubble.png', dpi=120)
print('Bubble chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Correlation scatter with regression line ──────────────────────────────────
np.random.seed(42)
n = 80
studyhours = np.random.uniform(1, 10, n)
scores = 40 + 5.5 * studyhours + np.random.normal(0, 8, n)
scores = np.clip(scores, 0, 100)

# Regression line
m, b = np.polyfit(studyhours, scores, 1)
x_line = np.linspace(1, 10, 100)

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(studyhours, scores, alpha=0.6, color='steelblue', edgecolors='white', s=60, label='Students')
ax.plot(x_line, m * x_line + b, 'r--', linewidth=2, label=f'Regression: y={m:.1f}x+{b:.1f}')

ax.set_xlabel('Daily Study Hours')
ax.set_ylabel('Exam Score (%)')
ax.set_title('Study Hours vs Exam Performance — Kenya Universities', fontsize=12, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day04_scatter_reg.png', dpi=120)

corr = np.corrcoef(studyhours, scores)[0, 1]
print(f'Correlation coefficient r = {corr:.3f}')
plt.close()
```
