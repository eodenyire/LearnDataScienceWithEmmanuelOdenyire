| Day | Topic | Key Functions |

| 6 | Subplots | `plt.subplots()`, `GridSpec` |

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# ════════════════════════════════════════════════════════════════════
# MINI PROJECT: Kenya Economy at a Glance — 5-Year Review Dashboard
# ════════════════════════════════════════════════════════════════════

years = [2019, 2020, 2021, 2022, 2023]
gdp_growth     = [5.4, -0.3, 7.5, 4.8, 5.6]       # % growth
inflation      = [5.2, 5.4, 6.1, 7.9, 7.7]         # average annual %
unemployment   = [5.3, 7.2, 5.7, 5.4, 5.1]         # %
tourists       = [2.0, 0.48, 0.87, 1.48, 1.97]     # millions
exports        = [62.0, 56.0, 68.5, 82.0, 88.5]    # KES billion (monthly avg)
mpesa_users    = [23.6, 28.4, 31.5, 34.2, 38.1]    # millions

fig = plt.figure(figsize=(15, 10), constrained_layout=True)
fig.suptitle('Kenya Economy at a Glance: 2019–2023', fontsize=16, fontweight='bold', y=1.01)

gs = gridspec.GridSpec(2, 3, figure=fig)

# ── 1. GDP Growth ─────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
colors1 = ['#27AE60' if g > 0 else '#E74C3C' for g in gdp_growth]
ax1.bar(years, gdp_growth, color=colors1, edgecolor='white')
ax1.axhline(0, color='black', linewidth=0.8)
ax1.set_title('GDP Growth Rate (%)', fontweight='bold')
ax1.set_ylabel('%')
ax1.grid(axis='y', alpha=0.3)
for y, g in zip(years, gdp_growth):
    ax1.text(y, g + (0.15 if g >= 0 else -0.4), f'{g:.1f}%', ha='center', fontsize=8, fontweight='bold')

# ── 2. Inflation vs Unemployment ──────────────────────────────────────────────
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(years, inflation,    'o-', color='tomato',   linewidth=2, label='Inflation')
ax2.plot(years, unemployment, 's-', color='steelblue', linewidth=2, label='Unemployment')
ax2.set_title('Inflation & Unemployment (%)', fontweight='bold')
ax2.set_ylabel('%')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

# ── 3. Tourist Arrivals ───────────────────────────────────────────────────────
ax3 = fig.add_subplot(gs[0, 2])
ax3.fill_between(years, tourists, alpha=0.4, color='#9B59B6')
ax3.plot(years, tourists, 'o-', color='#9B59B6', linewidth=2, markersize=7)
ax3.set_title('International Tourist Arrivals (M)', fontweight='bold')
ax3.set_ylabel('Millions')
ax3.grid(True, alpha=0.3)
ax3.annotate('COVID\nCrash', xy=(2020, 0.48), xytext=(2020.3, 0.8),
             arrowprops=dict(arrowstyle='->', color='red'), color='red', fontsize=9)

# ── 4. Exports ────────────────────────────────────────────────────────────────
ax4 = fig.add_subplot(gs[1, 0])
ax4.bar(years, exports, color='#E67E22', alpha=0.85, edgecolor='white')
ax4.set_title('Monthly Export Earnings (KES B)', fontweight='bold')
ax4.set_ylabel('KES Billion')
ax4.grid(axis='y', alpha=0.3)

# ── 5. M-Pesa Users ───────────────────────────────────────────────────────────
ax5 = fig.add_subplot(gs[1, 1])
ax5.plot(years, mpesa_users, 'o-', color='#00A651', linewidth=2.5, markersize=8)
ax5.fill_between(years, mpesa_users, alpha=0.15, color='#00A651')
ax5.set_title('M-Pesa Registered Users (M)', fontweight='bold')
ax5.set_ylabel('Millions')
ax5.grid(True, alpha=0.3)

# ── 6. Budget Allocation Pie ──────────────────────────────────────────────────
ax6 = fig.add_subplot(gs[1, 2])
budget_items = ['Education', 'Debt', 'Infrastructure', 'Health', 'Agriculture', 'Security', 'Others']
budget_vals  = [27, 25, 18, 12, 7, 11, 0]
budget_vals[-1] = 100 - sum(budget_vals[:-1])
colors6 = ['#3498DB','#E74C3C','#27AE60','#F39C12','#1ABC9C','#8E44AD','#95A5A6']
ax6.pie(budget_vals, labels=budget_items, autopct='%1.0f%%', colors=colors6,
        textprops={'fontsize': 8}, wedgeprops={'edgecolor':'white', 'linewidth':1})
ax6.set_title('Budget Allocation 2023/24', fontweight='bold')

fig.savefig('/tmp/day07_kenya_dashboard.png', dpi=130, bbox_inches='tight')
print('Kenya Economy Dashboard saved to /tmp/day07_kenya_dashboard.png')
plt.close()
```
