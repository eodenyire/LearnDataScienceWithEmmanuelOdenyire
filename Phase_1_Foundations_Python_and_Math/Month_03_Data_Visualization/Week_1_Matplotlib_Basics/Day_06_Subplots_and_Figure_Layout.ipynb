# Day 6: Subplots and Figure Layout

## Phase 1 | Month 3 | Week 1 | Day 6

**🎯 Goal:** Arrange multiple plots in one figure using subplots, GridSpec, and constrained_layout.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── 2×2 subplot grid ──────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(11, 8), constrained_layout=True)
fig.suptitle('Kenya Economic Dashboard 2023', fontsize=15, fontweight='bold')

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# 1. Inflation
axes[0,0].plot(months, [9.0,9.2,9.9,9.9,8.0,7.9,7.9,6.7,6.8,6.9,6.8,6.6],
               'o-', color='tomato', linewidth=2)
axes[0,0].set_title('Inflation Rate (%)')
axes[0,0].set_ylabel('%')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(True, alpha=0.3)

# 2. Forex KES/USD
axes[0,1].plot(months, [122,123,128,132,135,138,141,143,148,151,153,154],
               's-', color='steelblue', linewidth=2)
axes[0,1].set_title('KES/USD Exchange Rate')
axes[0,1].set_ylabel('KES per USD')
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].grid(True, alpha=0.3)

# 3. NSE 20 Index
axes[1,0].bar(months, [1750,1780,1690,1640,1600,1580,1560,1570,1550,1540,1560,1590],
              color='#27AE60', alpha=0.8)
axes[1,0].set_title('NSE 20 Share Index')
axes[1,0].set_ylabel('Index Points')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Tourist arrivals
axes[1,1].fill_between(months, [120,115,130,95,110,115,140,155,135,145,160,175],
                        alpha=0.5, color='purple')
axes[1,1].plot(months, [120,115,130,95,110,115,140,155,135,145,160,175],
               color='purple', linewidth=2)
axes[1,1].set_title('Tourist Arrivals (000s)')
axes[1,1].set_ylabel('Thousands')
axes[1,1].tick_params(axis='x', rotation=45)

fig.savefig('/tmp/day06_dashboard.png', dpi=120)
print('Dashboard saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# ── GridSpec: unequal subplot sizes ──────────────────────────────────────────
fig = plt.figure(figsize=(12, 8), constrained_layout=True)
gs = gridspec.GridSpec(2, 3, figure=fig)

ax_main  = fig.add_subplot(gs[0, :2])   # top-left: spans 2 columns
ax_right = fig.add_subplot(gs[0, 2])    # top-right: 1 column
ax_bl    = fig.add_subplot(gs[1, 0])
ax_bm    = fig.add_subplot(gs[1, 1])
ax_br    = fig.add_subplot(gs[1, 2])

x = np.linspace(0, 4*np.pi, 200)
ax_main.plot(x, np.sin(x), 'b-', linewidth=2)
ax_main.set_title('Main Chart (wide)', fontweight='bold')
ax_main.grid(True, alpha=0.3)

ax_right.bar(['A','B','C'], [3,7,5], color=['#E74C3C','#3498DB','#27AE60'])
ax_right.set_title('Side Bar', fontweight='bold')

for ax, color in zip([ax_bl, ax_bm, ax_br], ['tomato','steelblue','green']):
    ax.hist(np.random.randn(300), bins=20, color=color, alpha=0.75)
    ax.set_title('Distribution', fontweight='bold')

fig.suptitle('GridSpec Layout Demo', fontsize=14, fontweight='bold')
fig.savefig('/tmp/day06_gridspec.png', dpi=120)
print('GridSpec figure saved')
plt.close()
```
