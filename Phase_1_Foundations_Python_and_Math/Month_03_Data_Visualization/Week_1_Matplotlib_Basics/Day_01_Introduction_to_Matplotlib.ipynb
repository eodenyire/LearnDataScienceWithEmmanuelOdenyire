### 🌍 Why This Matters
Every chart you see in the Daily Nation, Standard Media, or CBK reports can be recreated in Matplotlib. Data Scientists at Safaricom, NCBA, and Equity Bank use it daily to communicate insights to executives.

```python
import matplotlib
matplotlib.use('Agg')   # non-interactive backend for scripts
import matplotlib.pyplot as plt
import numpy as np

print(f'Matplotlib version: {matplotlib.__version__}')

# ── Anatomy of a Figure ──────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot([1, 2, 3, 4, 5], [2, 4, 1, 6, 3], color='steelblue', linewidth=2, marker='o', label='Series A')

# Labels & title
ax.set_title('Figure Anatomy Demo', fontsize=16, fontweight='bold')
ax.set_xlabel('X Axis Label', fontsize=12)
ax.set_ylabel('Y Axis Label', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)

# Annotate a point
ax.annotate('Peak', xy=(4, 6), xytext=(3.2, 5.5),
            arrowprops=dict(arrowstyle='->', color='red'), fontsize=11, color='red')

fig.tight_layout()
fig.savefig('/tmp/day01_anatomy.png', dpi=120)
print('Figure saved to /tmp/day01_anatomy.png')
print('Key components: Figure -> Axes -> Axis -> Artists (lines, text, patches)')
plt.close()
```

## The Two Matplotlib APIs

| API | When to use | How |
|-----|------------|-----|
| `pyplot` (functional) | Quick exploratory plots | `plt.plot(x, y)` |
| Object-Oriented (OO) | Production code, subplots | `fig, ax = plt.subplots(); ax.plot(x, y)` |

**Best practice:** Always use the OO API in production code — it gives you explicit control over every element.

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# pyplot API (quick)
x = np.linspace(0, 2 * np.pi, 100)
plt.figure(figsize=(6, 3))
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('pyplot API')
plt.legend()
plt.savefig('/tmp/day01_pyplot.png', dpi=100)
plt.close()
print('pyplot API plot saved')

# Object-Oriented API (recommended)
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, np.sin(x), label='sin(x)', color='navy')
ax.plot(x, np.cos(x), label='cos(x)', color='crimson')
ax.set_title('Object-Oriented API')
ax.legend()
fig.savefig('/tmp/day01_oo.png', dpi=100)
plt.close()
print('OO API plot saved')
```

## 🌍 Real-World Example: Kenya Monthly Temperature

We'll plot average monthly temperatures for Nairobi — the same type of chart used by the Kenya Meteorological Department.

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
nairobi_temp = [25.5, 26.0, 25.8, 24.2, 23.1, 21.5, 20.9, 21.2, 23.5, 24.8, 24.2, 24.7]
mombasa_temp = [30.1, 30.8, 31.2, 30.5, 29.1, 27.3, 26.5, 26.8, 28.0, 29.2, 29.8, 30.0]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(months, nairobi_temp, 'o-', color='steelblue', linewidth=2, markersize=6, label='Nairobi (1795m)')
ax.plot(months, mombasa_temp, 's-', color='tomato',    linewidth=2, markersize=6, label='Mombasa (Coast)')

ax.set_title('Average Monthly Temperatures — Kenya 2023', fontsize=14, fontweight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (°C)')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.4)
ax.set_ylim(18, 34)

fig.tight_layout()
fig.savefig('/tmp/day01_kenya_temp.png', dpi=120)
print('Kenya temperature chart saved')
print(f'Nairobi avg: {np.mean(nairobi_temp):.1f}°C')
print(f'Mombasa avg: {np.mean(mombasa_temp):.1f}°C')
plt.close()
```
