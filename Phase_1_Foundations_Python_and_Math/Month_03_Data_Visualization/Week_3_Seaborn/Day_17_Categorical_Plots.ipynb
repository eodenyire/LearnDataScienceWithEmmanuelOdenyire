# Day 17: Categorical Plots

## Phase 1 | Month 3 | Week 3 | Day 17

**🎯 Goal:** Visualise distributions and relationships when one variable is categorical.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set2')
np.random.seed(42)

# Simulated crop yield dataset
n = 240
df = pd.DataFrame({
    'crop'   : np.random.choice(['Maize','Tea','Coffee','Wheat','Rice'], n),
    'county' : np.random.choice(['Nakuru','Nyeri','Meru','Kisii','Trans Nzoia'], n),
    'yield_kg': np.random.normal(2500, 600, n).clip(500, 5000),
    'season' : np.random.choice(['Long Rains','Short Rains'], n),
})

fig, axes = plt.subplots(1, 3, figsize=(16, 5), constrained_layout=True)
fig.suptitle('Kenya Crop Yield Analysis', fontsize=14, fontweight='bold')

sns.boxplot(data=df, x='crop', y='yield_kg', palette='Set2', ax=axes[0])
axes[0].set_title('Box Plot')
axes[0].set_xlabel('Crop')
axes[0].set_ylabel('Yield (kg/ha)')
axes[0].tick_params(axis='x', rotation=20)

sns.violinplot(data=df, x='crop', y='yield_kg', palette='Set2', inner='quartile', ax=axes[1])
axes[1].set_title('Violin Plot')
axes[1].set_xlabel('Crop')
axes[1].tick_params(axis='x', rotation=20)

sns.stripplot(data=df, x='crop', y='yield_kg', hue='season', jitter=True,
              alpha=0.5, ax=axes[2], palette='Set1', size=4)
axes[2].set_title('Strip Plot by Season')
axes[2].set_xlabel('Crop')
axes[2].tick_params(axis='x', rotation=20)

fig.savefig('/tmp/day17_categorical.png', dpi=120)
print('Categorical plots saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='muted')
np.random.seed(42)
n = 240
df = pd.DataFrame({
    'crop'   : np.random.choice(['Maize','Tea','Coffee','Wheat','Rice'], n),
    'county' : np.random.choice(['Nakuru','Nyeri','Meru','Kisii','Trans Nzoia'], n),
    'yield_kg': np.random.normal(2500, 600, n).clip(500, 5000),
    'season' : np.random.choice(['Long Rains','Short Rains'], n),
})

# Bar plot (shows mean + CI automatically)
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=df, x='crop', y='yield_kg', hue='season',
            palette='Set1', capsize=0.1, errwidth=1.5, ax=ax)
ax.set_title('Mean Crop Yield by Season (with 95% CI)', fontsize=12, fontweight='bold')
ax.set_xlabel('Crop')
ax.set_ylabel('Average Yield (kg/ha)')
ax.legend(title='Season')
fig.tight_layout()
fig.savefig('/tmp/day17_barplot.png', dpi=120)
print('Bar plot with CI saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='ticks')
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'operator': np.random.choice(['Safaricom','Airtel','Telkom'], n, p=[0.65, 0.25, 0.10]),
    'data_gb' : np.random.exponential(2.5, n).clip(0.1, 20),
    'region'  : np.random.choice(['Nairobi','Coast','Western','Rift Valley'], n),
    'age_group': np.random.choice(['18-25','26-35','36-50','50+'], n),
})

# Count plot
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
sns.countplot(data=df, x='operator', hue='region', palette='Set2', ax=axes[0])
axes[0].set_title('Subscribers by Operator and Region', fontweight='bold')
axes[0].set_xlabel('Mobile Operator')
axes[0].set_ylabel('Count')
axes[0].legend(title='Region', fontsize=8)

sns.boxplot(data=df, x='age_group', y='data_gb', hue='operator',
            palette=['#00A651','#FF0000','#0000FF'], ax=axes[1])
axes[1].set_title('Data Usage by Age Group and Operator', fontweight='bold')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Monthly Data (GB)')
axes[1].legend(title='Operator')

fig.tight_layout()
fig.savefig('/tmp/day17_mobile.png', dpi=120)
print('Mobile operator chart saved')
plt.close()
```
