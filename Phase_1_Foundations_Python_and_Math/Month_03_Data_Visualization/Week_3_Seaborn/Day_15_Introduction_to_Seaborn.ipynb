### 🌍 Why Seaborn?
Seaborn's statistical plotting turns 20 lines of Matplotlib code into one line. At Safaricom's data team, they use Seaborn for exploratory analysis; Matplotlib for final polished reports.

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='muted')

# ── Simulate African mobile banking dataset ───────────────────────────────────
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'country'    : np.random.choice(['Kenya','Nigeria','Ghana','Rwanda','Tanzania'], n),
    'age'        : np.random.randint(18, 65, n),
    'income_usd' : np.abs(np.random.normal(450, 200, n)),
    'tx_per_month': np.random.poisson(12, n),
    'gender'     : np.random.choice(['Male','Female'], n),
    'has_account': np.random.choice([0, 1], n, p=[0.3, 0.7]),
})
df['tx_value'] = df['income_usd'] * df['tx_per_month'] * np.random.uniform(0.1, 0.5, n)

print(df.head())
print(f'\nShape: {df.shape}')
print(df.dtypes)
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='muted')
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'country'    : np.random.choice(['Kenya','Nigeria','Ghana','Rwanda','Tanzania'], n),
    'age'        : np.random.randint(18, 65, n),
    'income_usd' : np.abs(np.random.normal(450, 200, n)),
    'tx_per_month': np.random.poisson(12, n),
    'gender'     : np.random.choice(['Male','Female'], n),
})

# Figure-level: relplot (scatter)
fig = sns.relplot(data=df, x='income_usd', y='tx_per_month',
                  hue='country', style='gender', size='age',
                  sizes=(30, 120), alpha=0.75, height=5, aspect=1.4)
fig.set(title='Mobile Transactions vs Income by Country', xlabel='Monthly Income (USD)', ylabel='Transactions/Month')
fig.savefig('/tmp/day15_relplot.png', dpi=120, bbox_inches='tight')
print('relplot saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set2')
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'country'    : np.random.choice(['Kenya','Nigeria','Ghana','Rwanda','Tanzania'], n),
    'age'        : np.random.randint(18, 65, n),
    'income_usd' : np.abs(np.random.normal(450, 200, n)),
    'tx_per_month': np.random.poisson(12, n),
    'gender'     : np.random.choice(['Male','Female'], n),
})

# catplot: box per country
fig = sns.catplot(data=df, x='country', y='income_usd', hue='gender',
                  kind='box', height=5, aspect=1.6, palette='Set2')
fig.set(title='Income Distribution by Country and Gender', ylabel='Monthly Income (USD)')
fig.savefig('/tmp/day15_catplot.png', dpi=120, bbox_inches='tight')
print('catplot saved')
plt.close()
```
