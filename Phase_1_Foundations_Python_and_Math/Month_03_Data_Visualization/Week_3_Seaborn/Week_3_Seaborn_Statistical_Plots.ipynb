# Week 3: Seaborn Statistical Plots

## Learning Objectives
- Create distribution plots with Seaborn
- Build categorical and relational plots
- Use FacetGrid for multi-panel figures
- Apply Seaborn themes and palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='husl')
print(f'Seaborn version: {sns.__version__}')
```

## 1. Distribution Plots

```python
rng = np.random.default_rng(42)
data = pd.DataFrame({
    'score': np.concatenate([rng.normal(70, 10, 100), rng.normal(85, 8, 100)]),
    'group': ['A'] * 100 + ['B'] * 100
})

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(data=data, x='score', hue='group', kde=True, ax=axes[0], bins=20)
axes[0].set_title('Histogram with KDE')

sns.boxplot(data=data, x='group', y='score', ax=axes[1], palette='Set2')
axes[1].set_title('Box Plot by Group')
plt.tight_layout()
plt.show()
```

## 2. Categorical Plots

```python
tips = sns.load_dataset('tips')
print(tips.head())

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', ax=axes[0], errorbar='sd')
axes[0].set_title('Total Bill by Day and Gender')

sns.violinplot(data=tips, x='day', y='tip', hue='sex', ax=axes[1], split=True)
axes[1].set_title('Tip Distribution by Day and Gender')
plt.tight_layout()
plt.show()
```

## 3. Relational and Regression Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', ax=axes[0])
axes[0].set_title('Bill vs Tip by Meal Time')

sns.regplot(data=tips, x='total_bill', y='tip', ax=axes[1], scatter_kws={'alpha': 0.5})
axes[1].set_title('Bill vs Tip (Regression Line)')
plt.tight_layout()
plt.show()
```

## 4. Heatmap and Pair Plot

```python
iris = sns.load_dataset('iris')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[0])
axes[0].set_title('Correlation Heatmap (Iris)')

sns.boxplot(data=iris, x='species', y='petal_length', ax=axes[1], palette='Set3')
axes[1].set_title('Petal Length by Species')
plt.tight_layout()
plt.show()
```
