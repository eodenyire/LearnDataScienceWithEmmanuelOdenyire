# Day 18: Relational Plots

## Phase 1 | Month 3 | Week 3 | Day 18

**🎯 Goal:** Visualise relationships between two continuous variables using scatter and line plots, with multi-variable encoding.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='deep')
np.random.seed(42)
n = 400

df = pd.DataFrame({
    'country'  : np.random.choice(['Kenya','Nigeria','Ghana','Ethiopia','Tanzania','South Africa'], n),
    'gdp_pc'   : np.random.lognormal(7.5, 0.8, n),
    'internet' : np.random.beta(2, 5, n) * 100,
    'mobile'   : np.random.beta(4, 3, n) * 100,
    'year'     : np.random.choice(range(2015, 2024), n),
    'pop_m'    : np.random.uniform(10, 230, n),
})

# Basic scatter with regression line
fig, ax = plt.subplots(figsize=(9, 6))
sns.regplot(data=df, x='gdp_pc', y='internet', scatter_kws={'alpha':0.4,'s':30},
            line_kws={'color':'red','linewidth':2}, ax=ax)
ax.set_title('Internet Penetration vs GDP per Capita — Africa', fontsize=12, fontweight='bold')
ax.set_xlabel('GDP per Capita (USD, log scale)')
ax.set_ylabel('Internet Penetration (%)')
ax.set_xscale('log')
fig.tight_layout()
fig.savefig('/tmp/day18_regplot.png', dpi=120)
print('Regression scatter saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set1')
np.random.seed(42)
n = 400
df = pd.DataFrame({
    'country'  : np.random.choice(['Kenya','Nigeria','Ghana','Ethiopia','Tanzania'], n),
    'gdp_pc'   : np.random.lognormal(7.5, 0.8, n),
    'internet' : np.random.beta(2, 5, n) * 100,
    'mobile'   : np.random.beta(4, 3, n) * 100,
    'pop_m'    : np.random.uniform(10, 230, n),
})

# scatterplot with hue, size, style
fig, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(data=df, x='gdp_pc', y='internet',
                hue='country', size='pop_m', style='country',
                sizes=(30, 200), alpha=0.7, ax=ax)
ax.set_title('Internet Penetration vs GDP per Capita (bubble = population)', fontsize=12, fontweight='bold')
ax.set_xscale('log')
ax.set_xlabel('GDP per Capita (USD)')
ax.set_ylabel('Internet Penetration (%)')
ax.legend(title='Country', bbox_to_anchor=(1.01, 1), borderaxespad=0)
fig.tight_layout()
fig.savefig('/tmp/day18_scatter.png', dpi=120, bbox_inches='tight')
print('Multi-variable scatter saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid')
# Time series line plot
years = list(range(2015, 2024))
data_rows = []
for country, base, growth in [('Kenya', 26, 4.5), ('Nigeria', 45, 3.8), ('Ghana', 35, 5.1),
                                ('Ethiopia', 12, 8.2), ('Tanzania', 20, 5.5)]:
    for i, yr in enumerate(years):
        val = base + growth * i + np.random.normal(0, 2)
        data_rows.append({'country': country, 'year': yr, 'mobile_pct': np.clip(val, 0, 100)})
df_ts = pd.DataFrame(data_rows)

fig, ax = plt.subplots(figsize=(11, 6))
sns.lineplot(data=df_ts, x='year', y='mobile_pct', hue='country',
             marker='o', linewidth=2, ax=ax)
ax.set_title('Mobile Internet Penetration Growth 2015–2023', fontsize=12, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Mobile Internet Penetration (%)')
ax.legend(title='Country', bbox_to_anchor=(1.01, 1))
ax.set_ylim(0, 100)
fig.tight_layout()
fig.savefig('/tmp/day18_lineplot.png', dpi=120, bbox_inches='tight')
print('Line plot saved')
plt.close()
```
