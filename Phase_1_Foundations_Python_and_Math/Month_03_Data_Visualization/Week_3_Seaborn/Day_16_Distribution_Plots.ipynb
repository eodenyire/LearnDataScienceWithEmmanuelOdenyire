# Day 16: Distribution Plots

## Phase 1 | Month 3 | Week 3 | Day 16

**🎯 Goal:** Visualise data distributions with histograms, KDE, ECDF, rug plots, and violin plots.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='muted')
np.random.seed(42)

# Dataset: Kenyan household survey (simulated)
n = 500
df = pd.DataFrame({
    'income'     : np.concatenate([np.random.lognormal(5.8, 0.6, 350),
                                   np.random.lognormal(7.0, 0.4, 150)]),
    'age'        : np.random.normal(32, 10, n).clip(18, 70),
    'region'     : np.random.choice(['Urban','Rural','Peri-urban'], n, p=[0.45, 0.35, 0.20]),
    'education'  : np.random.choice(['Primary','Secondary','Tertiary'], n, p=[0.25, 0.45, 0.30]),
})

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Income Distribution — Kenya Household Survey', fontsize=13, fontweight='bold')

sns.histplot(data=df, x='income', bins=40, ax=axes[0], color='steelblue')
axes[0].set_title('Histogram')
axes[0].set_xlabel('Monthly Income (KES)')

sns.kdeplot(data=df, x='income', hue='region', fill=True, ax=axes[1], alpha=0.4)
axes[1].set_title('KDE by Region')
axes[1].set_xlabel('Monthly Income (KES)')

sns.ecdfplot(data=df, x='income', hue='region', ax=axes[2])
axes[2].set_title('ECDF by Region')
axes[2].set_xlabel('Monthly Income (KES)')

fig.tight_layout()
fig.savefig('/tmp/day16_distributions.png', dpi=120)
print('Distribution plots saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid')
np.random.seed(42)
n = 500
df = pd.DataFrame({
    'income' : np.concatenate([np.random.lognormal(5.8, 0.6, 350), np.random.lognormal(7.0, 0.4, 150)]),
    'age'    : np.random.normal(32, 10, n).clip(18, 70),
    'education': np.random.choice(['Primary','Secondary','Tertiary'], n, p=[0.25, 0.45, 0.30]),
})

# Histogram + KDE + rug on same axes
fig, ax = plt.subplots(figsize=(9, 5))
sns.histplot(data=df, x='income', kde=True, bins=35, ax=ax,
             color='steelblue', alpha=0.6, label='Histogram + KDE')
sns.rugplot(data=df.sample(100), x='income', ax=ax, color='tomato', alpha=0.3, height=0.04)
ax.set_title('Household Income Distribution with KDE and Rug Plot', fontsize=12, fontweight='bold')
ax.set_xlabel('Monthly Income (KES)')
ax.set_ylabel('Count')
ax.legend()
fig.tight_layout()
fig.savefig('/tmp/day16_hist_kde_rug.png', dpi=120)
print('Histogram + KDE + rug plot saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set2')
np.random.seed(42)
n = 500
df = pd.DataFrame({
    'income' : np.concatenate([np.random.lognormal(5.8, 0.6, 350), np.random.lognormal(7.0, 0.4, 150)]),
    'education': np.random.choice(['Primary','Secondary','Tertiary'], n, p=[0.25, 0.45, 0.30]),
})

# Violin plot: shows full distribution shape
fig, ax = plt.subplots(figsize=(8, 5))
sns.violinplot(data=df, x='education', y='income', palette='Set2', inner='box', ax=ax)
ax.set_title('Income Distribution by Education Level', fontsize=12, fontweight='bold')
ax.set_xlabel('Education Level')
ax.set_ylabel('Monthly Income (KES)')
fig.tight_layout()
fig.savefig('/tmp/day16_violin.png', dpi=120)
print('Violin plot saved')
plt.close()
```
