# Day 20: Pair Plots and FacetGrid

## Phase 1 | Month 3 | Week 3 | Day 20

**🎯 Goal:** Explore all pairwise relationships in a dataset using `pairplot` and build custom multi-panel figures with `FacetGrid`.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set1')
np.random.seed(42)

# African Development Indicators dataset
n = 250
df = pd.DataFrame({
    'gdp_pc'     : np.random.lognormal(7.5, 0.7, n),
    'life_exp'   : np.random.normal(65, 8, n).clip(45, 82),
    'literacy'   : np.random.beta(5, 2, n) * 100,
    'mobile_pct' : np.random.beta(3, 2, n) * 100,
    'region'     : np.random.choice(['East Africa','West Africa','Southern Africa','North Africa'], n),
})

g = sns.pairplot(df, hue='region', vars=['gdp_pc','life_exp','literacy','mobile_pct'],
                 diag_kind='kde', plot_kws={'alpha':0.5, 's':30},
                 diag_kws={'fill':True})
g.fig.suptitle('Pairwise Relationships — African Development Indicators', y=1.02, fontsize=13, fontweight='bold')
g.savefig('/tmp/day20_pairplot.png', dpi=120, bbox_inches='tight')
print('Pair plot saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='muted')
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'county'   : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n),
    'sector'   : np.random.choice(['Agri','Tech','Finance','Tourism'], n),
    'revenue'  : np.random.lognormal(4.5, 0.8, n),
    'employees': np.random.randint(5, 500, n),
    'year'     : np.random.choice([2021, 2022, 2023], n),
})

# FacetGrid with histogram per county
g = sns.FacetGrid(df, col='county', col_wrap=3, height=3.5, aspect=1.1, sharey=False)
g.map_dataframe(sns.histplot, x='revenue', bins=25, color='steelblue', kde=True)
g.set_axis_labels('Revenue (KES M)', 'Count')
g.set_titles(col_template='{col_name}')
g.fig.suptitle('Revenue Distribution by County — Kenya SME Survey', y=1.03, fontsize=13, fontweight='bold')
g.tight_layout()
g.savefig('/tmp/day20_facetgrid.png', dpi=120, bbox_inches='tight')
print('FacetGrid saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid')
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'county'   : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n),
    'sector'   : np.random.choice(['Agri','Tech','Finance','Tourism'], n),
    'revenue'  : np.random.lognormal(4.5, 0.8, n),
    'employees': np.random.randint(5, 500, n),
})

# FacetGrid: scatter employees vs revenue, faceted by sector
g = sns.FacetGrid(df, col='sector', hue='county', height=3.5, aspect=1.0)
g.map_dataframe(sns.scatterplot, x='employees', y='revenue', alpha=0.5, s=30)
g.add_legend(title='County')
g.set_axis_labels('Employees', 'Revenue (KES M)')
g.set_titles(col_template='{col_name} Sector')
g.fig.suptitle('Employees vs Revenue by Sector', y=1.03, fontsize=13, fontweight='bold')
g.savefig('/tmp/day20_facet_scatter.png', dpi=120, bbox_inches='tight')
print('FacetGrid scatter saved')
plt.close()
```
