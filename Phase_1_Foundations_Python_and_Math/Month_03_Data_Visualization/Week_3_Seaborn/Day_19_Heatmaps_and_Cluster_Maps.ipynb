```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='white')
np.random.seed(42)

# Economic indicators correlation matrix
countries = ['Kenya','Nigeria','Ghana','Ethiopia','Tanzania','Rwanda','Uganda','Senegal']
indicators = ['GDP Growth','Inflation','Unemployment','FDI Inflow','Export Growth','Literacy']

corr = np.array([
    [ 1.00,-0.62,-0.71, 0.83, 0.77, 0.68],
    [-0.62, 1.00, 0.53,-0.59,-0.50,-0.44],
    [-0.71, 0.53, 1.00,-0.65,-0.61,-0.55],
    [ 0.83,-0.59,-0.65, 1.00, 0.72, 0.64],
    [ 0.77,-0.50,-0.61, 0.72, 1.00, 0.59],
    [ 0.68,-0.44,-0.55, 0.64, 0.59, 1.00],
])

fig, ax = plt.subplots(figsize=(8, 6))
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)  # upper triangle mask (show lower)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r',
            xticklabels=indicators, yticklabels=indicators,
            vmin=-1, vmax=1, center=0,
            linewidths=0.5, mask=mask, ax=ax,
            cbar_kws={'label': 'Pearson r'})
ax.set_title('Economic Indicator Correlations — Sub-Saharan Africa', fontsize=12, fontweight='bold')
fig.tight_layout()
fig.savefig('/tmp/day19_heatmap.png', dpi=120)
print('Correlation heatmap saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='white')
np.random.seed(42)

# Monthly crop yield heatmap (county × month)
counties = ['Nakuru','Trans Nzoia','Uasin Gishu','Nyandarua','Kirinyaga','Meru','Embu','Kisii']
months   = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
data = np.random.normal(2500, 400, (len(counties), len(months)))
# Add seasonal pattern
for i, m in enumerate(months):
    if m in ['Apr','May','Nov','Dec']:  # rainy seasons — higher yield
        data[:, i] += 500
    elif m in ['Jan','Feb','Jul','Aug']:  # dry — lower
        data[:, i] -= 400
data = data.clip(800, 4500)

df_heat = pd.DataFrame(data, index=counties, columns=months)

fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(df_heat, annot=True, fmt='.0f', cmap='YlGn',
            linewidths=0.5, linecolor='white', ax=ax,
            cbar_kws={'label': 'Yield (kg/ha)'})
ax.set_title('Monthly Maize Yield by County (kg/ha)', fontsize=12, fontweight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('County')
fig.tight_layout()
fig.savefig('/tmp/day19_yield_heatmap.png', dpi=120)
print('Crop yield heatmap saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='white')
np.random.seed(42)

# Clustermap: group similar countries/months automatically
counties = ['Nakuru','Trans Nzoia','Uasin Gishu','Nyandarua','Kirinyaga','Meru','Embu','Kisii']
months   = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
data = np.random.normal(2500, 400, (8, 12))
for i, m in enumerate(months):
    if m in ['Apr','May','Nov','Dec']: data[:, i] += 500
    elif m in ['Jan','Feb','Jul','Aug']: data[:, i] -= 400
data = data.clip(800, 4500)
df_heat = pd.DataFrame(data, index=counties, columns=months)

g = sns.clustermap(df_heat, cmap='YlGn', figsize=(12, 8),
                   linewidths=0.4, annot=True, fmt='.0f',
                   cbar_kws={'label': 'Yield (kg/ha)'})
g.fig.suptitle('Clustered Crop Yield Heatmap (auto-grouped counties & months)',
               fontsize=12, fontweight='bold', y=1.02)
g.savefig('/tmp/day19_clustermap.png', dpi=120, bbox_inches='tight')
print('Cluster map saved')
plt.close()
```
