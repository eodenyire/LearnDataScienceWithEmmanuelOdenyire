```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set2', font_scale=1.05)
np.random.seed(2024)

# ════════════════════════════════════════════════════════════════════
# WEEK 3 PROJECT: Kenya Health & Demographics Dashboard
# ════════════════════════════════════════════════════════════════════

n = 500
df = pd.DataFrame({
    'county'     : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Nyeri',
                                      'Meru','Kakamega','Kisii','Eldoret','Garissa'], n),
    'age'        : np.random.gamma(shape=3, scale=10, size=n).clip(1, 80).astype(int),
    'gender'     : np.random.choice(['Male','Female'], n),
    'bmi'        : np.random.normal(23.5, 4.2, n).clip(14, 45),
    'income_kes' : np.abs(np.random.lognormal(9.5, 0.9, n)),
    'hospital_visits': np.random.poisson(2.5, n),
    'has_insurance'  : np.random.choice(['Yes','No'], n, p=[0.35, 0.65]),
    'region'     : np.random.choice(['Urban','Rural','Peri-urban'], n, p=[0.4, 0.4, 0.2]),
})
df['age_group'] = pd.cut(df['age'], bins=[0,17,35,60,100],
                          labels=['Child (0-17)','Young Adult (18-35)','Adult (36-60)','Senior (60+)'])

fig, axes = plt.subplots(2, 3, figsize=(16, 10), constrained_layout=True)
fig.suptitle('Kenya Health & Demographics Survey Dashboard', fontsize=15, fontweight='bold')

# 1. Age distribution by gender
sns.histplot(data=df, x='age', hue='gender', bins=30, kde=True, ax=axes[0,0],
             palette=['#3498DB','#E74C3C'], alpha=0.7)
axes[0,0].set_title('Age Distribution by Gender')
axes[0,0].set_xlabel('Age (years)')

# 2. BMI by region (violin)
sns.violinplot(data=df, x='region', y='bmi', palette='Set2', inner='box', ax=axes[0,1])
axes[0,1].set_title('BMI Distribution by Region')
axes[0,1].axhline(25, color='red', linestyle='--', alpha=0.7, label='Overweight threshold')
axes[0,1].legend(fontsize=8)

# 3. Insurance coverage (count)
sns.countplot(data=df, x='has_insurance', hue='region', palette='Set1', ax=axes[0,2])
axes[0,2].set_title('Health Insurance Coverage by Region')
axes[0,2].set_xlabel('Has Health Insurance')

# 4. Hospital visits by age group (bar + CI)
sns.barplot(data=df, x='age_group', y='hospital_visits', hue='has_insurance',
            palette=['#27AE60','#E74C3C'], capsize=0.1, ax=axes[1,0])
axes[1,0].set_title('Hospital Visits by Age Group')
axes[1,0].set_xlabel('Age Group')
axes[1,0].tick_params(axis='x', rotation=15)
axes[1,0].legend(title='Insurance', fontsize=8)

# 5. Income vs BMI scatter
sns.scatterplot(data=df.sample(150), x='income_kes', y='bmi', hue='region',
                style='gender', size='hospital_visits', sizes=(20, 120),
                alpha=0.6, ax=axes[1,1])
axes[1,1].set_xscale('log')
axes[1,1].set_title('Income vs BMI')
axes[1,1].set_xlabel('Monthly Income (KES, log scale)')
axes[1,1].legend(fontsize=7, ncol=2)

# 6. Correlation heatmap
numeric_cols = ['age','bmi','income_kes','hospital_visits']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
            linewidths=0.5, ax=axes[1,2], cbar_kws={'shrink': 0.8})
axes[1,2].set_title('Variable Correlations')

fig.savefig('/tmp/day21_health_dashboard.png', dpi=130, bbox_inches='tight')
print('Kenya Health Dashboard saved to /tmp/day21_health_dashboard.png')
plt.close()
```
