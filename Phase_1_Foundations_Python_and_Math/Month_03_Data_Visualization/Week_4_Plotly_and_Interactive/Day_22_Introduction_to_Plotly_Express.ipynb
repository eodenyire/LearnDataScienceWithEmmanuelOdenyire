```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
n = 200
df = pd.DataFrame({
    'country'   : np.random.choice(['Kenya','Nigeria','Ghana','Ethiopia','Tanzania','Rwanda','Uganda'], n),
    'gdp_pc'    : np.random.lognormal(7.5, 0.8, n),
    'life_exp'  : np.random.normal(65, 8, n).clip(45, 82),
    'population': np.random.uniform(10, 230, n),
    'continent' : 'Africa',
    'year'      : np.random.choice(range(2015, 2024), n),
    'literacy'  : np.random.beta(4, 2, n) * 100,
})

# Interactive scatter plot — hover shows all columns
fig = px.scatter(
    df, x='gdp_pc', y='life_exp',
    size='population', color='country',
    hover_name='country',
    hover_data={'gdp_pc':':.0f', 'life_exp':':.1f', 'literacy':':.1f'},
    log_x=True,
    size_max=40,
    title='African Countries: GDP per Capita vs Life Expectancy',
    labels={'gdp_pc': 'GDP per Capita (USD)', 'life_exp': 'Life Expectancy (years)'}
)
fig.update_layout(template='plotly_white', height=500)
fig.write_html('/tmp/day22_scatter.html')
fig.write_image('/tmp/day22_scatter.png', scale=1.5)
print('Interactive scatter saved to /tmp/day22_scatter.html')
print('Static PNG saved to /tmp/day22_scatter.png')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
# Time series line chart
years = list(range(2015, 2024))
data_rows = []
for country, base, growth in [('Kenya',26,4.5),('Nigeria',45,3.8),('Ghana',35,5.1),
                                ('Ethiopia',12,8.2),('Tanzania',20,5.5),('Rwanda',28,7.1)]:
    for i, yr in enumerate(years):
        val = base + growth * i + np.random.normal(0, 2)
        data_rows.append({'Country': country, 'Year': yr,
                           'Mobile Internet (%)': np.clip(val, 0, 100)})
df_ts = pd.DataFrame(data_rows)

fig = px.line(
    df_ts, x='Year', y='Mobile Internet (%)', color='Country',
    markers=True,
    title='Mobile Internet Penetration Growth — East Africa 2015–2023',
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.update_layout(height=480, legend_title_text='Country')
fig.write_html('/tmp/day22_line.html')
fig.write_image('/tmp/day22_line.png', scale=1.5)
print('Line chart saved')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
countries  = ['Nigeria','South Africa','Egypt','Algeria','Ethiopia','Morocco','Kenya','Angola']
gdp_bn     = [477, 405, 396, 240, 156, 142, 113, 124]
pop_m      = [224, 60, 104, 45, 127, 37, 54, 35]
region     = ['West','Southern','North','North','East','North','East','Southern']

df_bar = pd.DataFrame({'Country':countries, 'GDP_Billion':gdp_bn, 'Pop_M':pop_m, 'Region':region})

fig = px.bar(
    df_bar.sort_values('GDP_Billion', ascending=True),
    x='GDP_Billion', y='Country',
    color='Region', orientation='h',
    text='GDP_Billion',
    hover_data={'Pop_M': True},
    title='Top African Economies by GDP (USD Billion, 2023)',
    labels={'GDP_Billion': 'GDP (USD Billion)'},
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig.update_traces(texttemplate='$%{text:.0f}B', textposition='outside')
fig.update_layout(height=480, showlegend=True)
fig.write_html('/tmp/day22_bar.html')
fig.write_image('/tmp/day22_bar.png', scale=1.5)
print('Bar chart saved')
```
