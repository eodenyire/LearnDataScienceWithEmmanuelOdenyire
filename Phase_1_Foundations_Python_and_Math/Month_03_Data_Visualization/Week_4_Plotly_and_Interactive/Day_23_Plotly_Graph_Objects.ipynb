### Plotly Express vs Graph Objects

| Feature | Plotly Express | Graph Objects |
|---------|---------------|---------------|
| Syntax | One-liner, high-level | Verbose, full control |
| Customisation | Limited | Complete |
| Use case | Quick EDA | Production dashboards |
| Learning curve | Low | Medium |

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(42)

# ── NSE Stock Price Chart ──────────────────────────────────────────────────────
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
safaricom  = [34.2, 35.1, 33.8, 32.5, 33.0, 34.5, 36.2, 37.0, 35.8, 34.2, 35.5, 36.8]
equity     = [48.5, 50.2, 47.8, 46.0, 47.5, 49.0, 51.2, 52.5, 50.0, 48.8, 50.0, 52.0]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=months, y=safaricom,
    mode='lines+markers',
    name='Safaricom (SCOM)',
    line=dict(color='#00A651', width=2.5),
    marker=dict(size=7),
    hovertemplate='<b>%{x}</b><br>Price: KES %{y:.1f}<extra>SCOM</extra>'
))

fig.add_trace(go.Scatter(
    x=months, y=equity,
    mode='lines+markers',
    name='Equity Bank (EQTY)',
    line=dict(color='#003087', width=2.5),
    marker=dict(size=7),
    hovertemplate='<b>%{x}</b><br>Price: KES %{y:.1f}<extra>EQTY</extra>'
))

fig.update_layout(
    title='NSE Stock Prices 2023 — Safaricom vs Equity Bank',
    xaxis_title='Month',
    yaxis_title='Share Price (KES)',
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    height=450
)
fig.write_html('/tmp/day23_stocks.html')
fig.write_image('/tmp/day23_stocks.png', scale=1.5)
print('Stock chart saved')
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

np.random.seed(42)
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
turnover  = [12.5, 14.2, 11.8, 10.5, 9.8, 11.2, 10.9, 12.4, 11.7, 10.8, 13.5, 15.2]
nse20     = [1750, 1780, 1730, 1690, 1660, 1640, 1625, 1610, 1590, 1570, 1585, 1600]

# ── Subplots with shared x-axis ────────────────────────────────────────────────
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.06,
                    subplot_titles=['NSE 20 Index', 'Monthly Turnover (KES Billion)'])

fig.add_trace(go.Scatter(x=months, y=nse20, mode='lines+markers',
                          line=dict(color='navy', width=2),
                          fill='tozeroy', fillcolor='rgba(0,0,128,0.08)',
                          name='NSE 20'), row=1, col=1)

fig.add_trace(go.Bar(x=months, y=turnover,
                      marker_color=['#27AE60' if t > np.mean(turnover) else '#E74C3C' for t in turnover],
                      name='Turnover'), row=2, col=1)

fig.update_layout(title='NSE Market Dashboard 2023',
                  template='plotly_white', height=600, showlegend=False)
fig.update_yaxes(title_text='Index Points', row=1, col=1)
fig.update_yaxes(title_text='KES Billion', row=2, col=1)
fig.write_html('/tmp/day23_subplots.html')
fig.write_image('/tmp/day23_subplots.png', scale=1.5)
print('Subplot dashboard saved')
```

```python
import plotly.graph_objects as go
import numpy as np

# ── Waterfall chart: company P&L ──────────────────────────────────────────────
fig = go.Figure(go.Waterfall(
    name='P&L', orientation='v',
    measure=['relative','relative','relative','relative','relative','total',
             'relative','relative','relative','total'],
    x=['Revenue','Cost of Goods','Gross Profit','Operating Costs','EBITDA',
       'Net Operating Income','Depreciation','Finance Costs','Tax','Net Profit'],
    textposition='outside',
    text=['+145B','-62B','','−38B','','','−8B','−5B','−12B',''],
    y=[145, -62, 0, -38, 0, 45, -8, -5, -12, 0],
    connector={'line':{'color':'rgb(63, 63, 63)'}},
    increasing={'marker':{'color':'#27AE60'}},
    decreasing={'marker':{'color':'#E74C3C'}},
    totals={'marker':{'color':'#2196F3'}}
))
fig.update_layout(
    title='Safaricom PLC — Simplified P&L Waterfall (KES Billion, FY2023)',
    template='plotly_white',
    height=500
)
fig.write_html('/tmp/day23_waterfall.html')
fig.write_image('/tmp/day23_waterfall.png', scale=1.5)
print('Waterfall chart saved')
```
