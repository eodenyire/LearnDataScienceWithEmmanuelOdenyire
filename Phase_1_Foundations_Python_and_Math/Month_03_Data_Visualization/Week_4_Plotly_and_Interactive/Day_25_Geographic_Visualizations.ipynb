# Day 25: Geographic Visualizations with Plotly

## Phase 1 | Month 3 | Week 4 | Day 25

**🎯 Goal:** Create choropleth maps, scatter maps, and bubble maps to show data on geographic coordinates.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)

# African countries data
africa_data = pd.DataFrame({
    'country': ['Kenya','Nigeria','Ethiopia','Ghana','Tanzania','South Africa',
                 'Egypt','Algeria','Morocco','Rwanda','Uganda','Senegal',
                 'Côte d\'Ivoire','Angola','Zambia','Zimbabwe','Mozambique'],
    'iso_alpha': ['KEN','NGA','ETH','GHA','TZA','ZAF',
                   'EGY','DZA','MAR','RWA','UGA','SEN',
                   'CIV','AGO','ZMB','ZWE','MOZ'],
    'gdp_pc':   [2082, 2086, 925, 2202, 1108, 7055,
                  3500, 3901, 3356, 930, 817, 1529,
                  2286, 3511, 1246, 1476, 503],
    'population':[54, 224, 127, 34, 66, 60,
                   104, 45, 37, 14, 47, 17,
                   28, 35, 20, 16, 33],
    'internet_pct':[29, 43, 18, 59, 25, 72,
                     72, 63, 88, 30, 26, 46,
                     40, 33, 17, 35, 10],
})

# Choropleth map
fig = px.choropleth(
    africa_data,
    locations='iso_alpha',
    color='gdp_pc',
    hover_name='country',
    hover_data={'population': True, 'internet_pct': True},
    color_continuous_scale='Viridis',
    scope='africa',
    title='GDP per Capita Across Africa (USD, 2023)',
    labels={'gdp_pc': 'GDP per Capita (USD)'}
)
fig.update_layout(template='plotly_white', height=600)
fig.write_html('/tmp/day25_choropleth.html')
fig.write_image('/tmp/day25_choropleth.png', scale=1.5)
print('Choropleth map saved')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Scatter map: Kenyan cities and counties
kenyan_places = pd.DataFrame({
    'name': ['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret','Thika','Nyeri','Machakos',
              'Malindi','Garissa','Kitale','Embu','Meru','Kisii','Kakamega'],
    'lat':  [-1.28,-4.05,-0.10,-0.30,0.52,-1.03,-0.42,-1.52,
              -3.22,-0.46,1.02,-0.53,0.05,-0.68,0.28],
    'lon':  [36.82,39.67,34.76,36.07,35.27,37.07,36.95,37.26,
              40.12,42.00,34.99,37.45,37.65,34.76,34.75],
    'population':[4300, 1200, 500, 570, 475, 250, 150, 200,
                   120, 110, 155, 130, 175, 170, 195],  # thousands
    'type': ['Metro','City','City','City','City','Town','Town','Town',
              'Town','Town','Town','Town','City','Town','Town'],
})

fig = px.scatter_mapbox(
    kenyan_places,
    lat='lat', lon='lon',
    size='population', color='type',
    hover_name='name',
    hover_data={'population': True},
    zoom=5, height=600,
    center={'lat': -0.5, 'lon': 37.5},
    title='Kenyan Cities and Towns by Population',
    size_max=30,
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.update_layout(mapbox_style='open-street-map')
fig.write_html('/tmp/day25_scatter_map.html')
print('Scatter map saved to /tmp/day25_scatter_map.html')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Bubble map on Africa scope
np.random.seed(42)
africa_mobile = pd.DataFrame({
    'country': ['Kenya','Nigeria','South Africa','Egypt','Ethiopia','Ghana',
                 'Morocco','Tanzania','Algeria','Rwanda'],
    'iso_alpha': ['KEN','NGA','ZAF','EGY','ETH','GHA','MAR','TZA','DZA','RWA'],
    'lat': [-1.3, 9.1, -30.6, 26.8, 9.1, 7.9, 31.8, -6.4, 28.0, -1.9],
    'lon': [36.8, 8.7, 25.1, 30.8, 40.5, -1.0, -5.0, 34.9, 1.7, 29.9],
    'mobile_users_M': [45, 210, 52, 110, 68, 42, 47, 58, 52, 12],
    'penetration_pct': [83, 94, 87, 105, 53, 138, 128, 88, 116, 87],
})

fig = px.scatter_geo(
    africa_mobile,
    lat='lat', lon='lon',
    size='mobile_users_M',
    color='penetration_pct',
    hover_name='country',
    hover_data={'mobile_users_M': ':.0f', 'penetration_pct': ':.1f'},
    scope='africa',
    color_continuous_scale='RdYlGn',
    title='African Mobile Penetration — Bubble Map (size = users, color = penetration %)',
    labels={'mobile_users_M': 'Mobile Users (M)', 'penetration_pct': 'Penetration (%)'}
)
fig.update_layout(template='plotly_white', height=600)
fig.write_html('/tmp/day25_bubble_map.html')
fig.write_image('/tmp/day25_bubble_map.png', scale=1.5)
print('Bubble map saved')
```
