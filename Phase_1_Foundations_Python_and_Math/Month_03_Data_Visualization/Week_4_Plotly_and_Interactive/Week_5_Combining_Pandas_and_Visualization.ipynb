# Week 5: Combining Pandas and Visualization

## Learning Objectives

- Chain data manipulation with visualization
- Build complete data stories

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

sns.set_theme(style='whitegrid')
```

## 1. Pandas Plot API

```python
csv_data = """month,region,sales,units
Jan,North,12000,120
Jan,South,9500,95
Feb,North,13500,135
Feb,South,11000,110
Mar,North,15000,150
Mar,South,12500,125
Apr,North,14000,140
Apr,South,13000,130"""

df = pd.read_csv(io.StringIO(csv_data))
pivot = df.pivot_table(values='sales', index='month', columns='region', aggfunc='sum')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
pivot.plot(kind='bar', ax=axes[0], colormap='Set2', rot=0)
axes[0].set_title('Sales by Month and Region (Bar)')
axes[0].set_ylabel('Sales ($)')

pivot.plot(kind='line', ax=axes[1], marker='o')
axes[1].set_title('Sales Trend by Region')
axes[1].set_ylabel('Sales ($)')
plt.tight_layout()
plt.show()
```

## 2. Data Story: Sales Analysis

```python
rng = np.random.default_rng(42)
n = 200
dates = pd.date_range('2024-01-01', periods=n, freq='D')
products = rng.choice(['Laptop', 'Phone', 'Tablet', 'Watch'], n)
amounts = rng.integers(50, 2000, n)
regions = rng.choice(['North', 'South', 'East', 'West'], n)

df_sales = pd.DataFrame({'date': dates, 'product': products, 'amount': amounts, 'region': regions})
df_sales['month'] = df_sales['date'].dt.strftime('%b')
df_sales['week'] = df_sales['date'].dt.isocalendar().week
print(df_sales.head())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Monthly total
monthly = df_sales.groupby('month')['amount'].sum()
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul']
monthly = monthly.reindex([m for m in month_order if m in monthly.index])
monthly.plot(kind='bar', ax=axes[0,0], color='steelblue', rot=0)
axes[0,0].set_title('Monthly Revenue')
axes[0,0].set_ylabel('Revenue ($)')

# Product share
product_rev = df_sales.groupby('product')['amount'].sum()
axes[0,1].pie(product_rev, labels=product_rev.index, autopct='%1.1f%%', startangle=90)
axes[0,1].set_title('Revenue Share by Product')

# Region comparison
region_rev = df_sales.groupby('region')['amount'].sum().sort_values()
region_rev.plot(kind='barh', ax=axes[1,0], color='coral')
axes[1,0].set_title('Revenue by Region')
axes[1,0].set_xlabel('Revenue ($)')

# Distribution
axes[1,1].hist(df_sales['amount'], bins=30, color='purple', alpha=0.7, edgecolor='white')
axes[1,1].set_title('Transaction Amount Distribution')
axes[1,1].set_xlabel('Amount ($)')
axes[1,1].set_ylabel('Frequency')

plt.suptitle('Sales Dashboard', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
