# Day 24: Interactive Charts — Dropdowns, Sliders, and Range Selectors

## Phase 1 | Month 3 | Week 4 | Day 24

**🎯 Goal:** Add interactive controls (dropdowns, range sliders, buttons) to Plotly charts to let users explore data themselves.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(42)
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

stocks = {
    'SCOM': [34.2, 35.1, 33.8, 32.5, 33.0, 34.5, 36.2, 37.0, 35.8, 34.2, 35.5, 36.8],
    'EQTY': [48.5, 50.2, 47.8, 46.0, 47.5, 49.0, 51.2, 52.5, 50.0, 48.8, 50.0, 52.0],
    'KCB' : [40.1, 41.5, 39.8, 38.5, 39.2, 41.0, 43.5, 44.2, 42.8, 41.5, 43.0, 44.8],
    'EABL': [142, 145, 138, 135, 140, 148, 155, 158, 150, 145, 152, 160],
}

# Dropdown to switch between stocks
fig = go.Figure()

for name, prices in stocks.items():
    fig.add_trace(go.Scatter(
        x=months, y=prices,
        mode='lines+markers',
        name=name,
        visible=(name == 'SCOM')
    ))

# Dropdown buttons
buttons = []
for i, name in enumerate(stocks.keys()):
    visibility = [False] * len(stocks)
    visibility[i] = True
    buttons.append(dict(
        label=name,
        method='update',
        args=[{'visible': visibility},
              {'title': f'NSE Stock Price 2023 — {name}'}]
    ))

fig.update_layout(
    updatemenus=[dict(type='dropdown', direction='down',
                      active=0, x=0.01, y=1.15,
                      buttons=buttons, showactive=True)],
    title='NSE Stock Price 2023 — SCOM',
    xaxis_title='Month',
    yaxis_title='Share Price (KES)',
    template='plotly_white', height=450
)
fig.write_html('/tmp/day24_dropdown.html')
fig.write_image('/tmp/day24_dropdown.png', scale=1.5)
print('Dropdown chart saved to /tmp/day24_dropdown.html')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
# Time series with range slider
years = list(range(2015, 2024))
data_rows = []
for country, base, growth in [('Kenya',26,4.5),('Nigeria',45,3.8),('Ghana',35,5.1)]:
    for i, yr in enumerate(years):
        val = base + growth * i + np.random.normal(0, 2)
        data_rows.append({'Country': country, 'Year': yr,
                           'Mobile Internet (%)': np.clip(val, 0, 100)})
df = pd.DataFrame(data_rows)

fig = px.line(df, x='Year', y='Mobile Internet (%)', color='Country',
              markers=True,
              title='Mobile Internet Growth — Select Date Range with Slider Below',
              template='plotly_white')

# Range slider
fig.update_xaxes(rangeslider_visible=True)
fig.update_layout(height=480)
fig.write_html('/tmp/day24_rangeslider.html')
fig.write_image('/tmp/day24_rangeslider.png', scale=1.5)
print('Range slider chart saved')
```

```python
import plotly.graph_objects as go
import numpy as np

# Animation with play button — population growth
np.random.seed(42)
years = list(range(2015, 2024))
countries = ['Kenya','Nigeria','Ethiopia','Egypt','Tanzania']
base_pop  = {'Kenya':48, 'Nigeria':195, 'Ethiopia':109, 'Egypt':99, 'Tanzania':57}
growth    = {'Kenya':2.4, 'Nigeria':2.6, 'Ethiopia':2.6, 'Egypt':1.8, 'Tanzania':3.0}
base_gdp  = {'Kenya':2082, 'Nigeria':2086, 'Ethiopia':925, 'Egypt':3500, 'Tanzania':1108}
gdp_growth= {'Kenya':100, 'Nigeria':80, 'Ethiopia':120, 'Egypt':90, 'Tanzania':70}

frames = []
for yr in years:
    t = yr - 2015
    frame_data = []
    for c in countries:
        pop = base_pop[c] * (1 + growth[c]/100)**t
        gdp = base_gdp[c] + gdp_growth[c] * t
        frame_data.append({'country': c, 'pop': pop, 'gdp': gdp})
    frames.append(go.Frame(
        data=[go.Scatter(
            x=[d['gdp'] for d in frame_data],
            y=[d['pop'] for d in frame_data],
            mode='markers+text',
            marker=dict(size=[d['pop']/5 for d in frame_data], sizemode='area',
                        color=list(range(len(countries))), colorscale='Viridis'),
            text=[d['country'] for d in frame_data],
            textposition='top center',
        )],
        name=str(yr)
    ))

t0_data = [{'country':c, 'pop': base_pop[c], 'gdp': base_gdp[c]} for c in countries]
fig = go.Figure(
    data=[go.Scatter(
        x=[d['gdp'] for d in t0_data],
        y=[d['pop'] for d in t0_data],
        mode='markers+text',
        text=[d['country'] for d in t0_data],
        textposition='top center',
        marker=dict(size=[d['pop']/5 for d in t0_data], sizemode='area',
                    color=list(range(len(countries))), colorscale='Viridis')
    )],
    frames=frames
)
fig.update_layout(
    title='Population vs GDP Growth Animation 2015–2023',
    xaxis_title='GDP per Capita (USD)',
    yaxis_title='Population (Millions)',
    template='plotly_white', height=500,
    updatemenus=[dict(type='buttons', showactive=False, y=0,
                      buttons=[dict(label='▶ Play', method='animate',
                                   args=[None, {'frame':{'duration':700},'fromcurrent':True}]),
                               dict(label='⏸ Pause', method='animate',
                                   args=[[None], {'frame':{'duration':0},'mode':'immediate'}])])],
    sliders=[dict(steps=[dict(args=[[str(yr)],{'frame':{'duration':700},'mode':'immediate'}],
                              label=str(yr), method='animate') for yr in years],
                  active=0, x=0.1, len=0.8, y=0,
                  currentvalue={'prefix':'Year: '})]
)
fig.write_html('/tmp/day24_animation.html')
print('Animated chart saved to /tmp/day24_animation.html')
```
