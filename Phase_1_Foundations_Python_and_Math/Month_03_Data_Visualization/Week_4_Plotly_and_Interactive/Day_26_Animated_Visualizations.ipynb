# Day 26: Animated Visualizations with Plotly

## Phase 1 | Month 3 | Week 4 | Day 26

**🎯 Goal:** Create animated charts that show how data changes over time using Plotly's `animation_frame` parameter and `go.Frame`.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)

# ── Gapminder-style animation: African development 2015-2023 ─────────────────
years = list(range(2015, 2024))
country_info = {
    'Kenya':         (54, 2082, 66, 'East Africa'),
    'Nigeria':       (195, 2086, 53, 'West Africa'),
    'Ethiopia':      (109, 925, 65, 'East Africa'),
    'South Africa':  (60, 7055, 64, 'Southern Africa'),
    'Egypt':         (99, 3500, 72, 'North Africa'),
    'Ghana':         (33, 2202, 64, 'West Africa'),
    'Rwanda':        (13, 930, 69, 'East Africa'),
    'Tanzania':      (61, 1108, 65, 'East Africa'),
}

rows = []
for yr in years:
    t = yr - 2015
    for country, (pop0, gdp0, life0, region) in country_info.items():
        pop  = pop0 * (1 + 0.025)**t
        gdp  = gdp0 * (1 + 0.04 + np.random.normal(0, 0.01))**t
        life = min(80, life0 + 0.3 * t + np.random.normal(0, 0.3))
        rows.append({'Year': yr, 'Country': country, 'Region': region,
                     'Population_M': pop, 'GDP_pc': gdp, 'Life_Exp': life})

df = pd.DataFrame(rows)

fig = px.scatter(
    df, x='GDP_pc', y='Life_Exp',
    animation_frame='Year', animation_group='Country',
    size='Population_M', color='Region',
    hover_name='Country',
    log_x=True, size_max=45,
    range_x=[500, 12000], range_y=[48, 80],
    title='African Development 2015–2023 (Gapminder-Style Animation)',
    labels={'GDP_pc': 'GDP per Capita (USD)', 'Life_Exp': 'Life Expectancy (years)'},
    template='plotly_white'
)
fig.update_layout(height=560)
fig.write_html('/tmp/day26_gapminder.html')
print('Gapminder animation saved to /tmp/day26_gapminder.html')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
# Animated bar chart race: mobile subscribers
years = list(range(2015, 2024))
countries = ['Kenya','Nigeria','South Africa','Ethiopia','Ghana','Tanzania','Egypt','Algeria']
base = {'Kenya':40,'Nigeria':180,'South Africa':48,'Ethiopia':55,'Ghana':38,'Tanzania':50,'Egypt':95,'Algeria':45}
growth = {'Kenya':2.2,'Nigeria':2.8,'South Africa':1.2,'Ethiopia':3.5,'Ghana':2.5,'Tanzania':3.0,'Egypt':2.0,'Algeria':1.5}

rows = []
for yr in years:
    t = yr - 2015
    for c in countries:
        val = base[c] * (1 + growth[c]/100)**t + np.random.normal(0, 1)
        rows.append({'Year': str(yr), 'Country': c, 'Subscribers_M': round(val, 1)})

df = pd.DataFrame(rows)

fig = px.bar(
    df, y='Country', x='Subscribers_M',
    animation_frame='Year', animation_group='Country',
    orientation='h', color='Country',
    range_x=[0, 260],
    title='Mobile Subscribers (Millions) — Africa 2015–2023',
    labels={'Subscribers_M': 'Mobile Subscribers (Millions)'},
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.update_layout(height=520, showlegend=False)
fig.write_html('/tmp/day26_bar_race.html')
print('Bar chart race animation saved')
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
# Animated choropleth: internet penetration growth
years = list(range(2015, 2024))
countries_data = [
    ('Kenya','KEN',26), ('Nigeria','NGA',38), ('Ethiopia','ETH',10),
    ('South Africa','ZAF',58), ('Egypt','EGY',55), ('Ghana','GHA',38),
    ('Morocco','MAR',72), ('Tanzania','TZA',18), ('Algeria','DZA',50), ('Rwanda','RWA',22)
]

rows = []
for yr in years:
    t = yr - 2015
    for name, iso, base in countries_data:
        pct = min(95, base + 4.5*t + np.random.normal(0, 1.5))
        rows.append({'Year': str(yr), 'Country': name, 'ISO': iso, 'Internet_Pct': round(pct, 1)})

df = pd.DataFrame(rows)

fig = px.choropleth(
    df, locations='ISO',
    color='Internet_Pct', animation_frame='Year',
    hover_name='Country',
    color_continuous_scale='Viridis',
    scope='africa', range_color=[0, 100],
    title='African Internet Penetration Growth 2015–2023 (Animated)',
    labels={'Internet_Pct': 'Internet Penetration (%)'}
)
fig.update_layout(template='plotly_white', height=580)
fig.write_html('/tmp/day26_animated_map.html')
print('Animated choropleth saved to /tmp/day26_animated_map.html')
```
