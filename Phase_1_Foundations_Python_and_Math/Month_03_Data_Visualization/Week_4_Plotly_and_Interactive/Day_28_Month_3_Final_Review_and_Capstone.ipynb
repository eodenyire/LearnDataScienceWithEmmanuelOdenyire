### 🌍 Visualisation Decision Framework

| Data Story | Best Chart | Library |
|------------|-----------|----------|
| Trend over time | Line chart | Matplotlib or Plotly |
| Compare categories | Bar chart | Matplotlib or Seaborn |
| Distribution shape | Histogram/KDE | Seaborn |
| Two variables | Scatter | Seaborn (`regplot`) |
| Many variables | Pair plot | Seaborn |
| Part of whole | Pie/Donut | Matplotlib |
| Geographic data | Choropleth/Bubble map | Plotly |
| Correlation matrix | Heatmap | Seaborn |
| Interactive exploration | Any chart | Plotly |
| Dashboard | Multi-panel | Dash or Streamlit |

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid', palette='Set2', font_scale=1.05)
np.random.seed(2024)

# ════════════════════════════════════════════════════════════════════
# CAPSTONE PART 1: Static Matplotlib/Seaborn Dashboard
# Kenya Digital Economy Snapshot 2023
# ════════════════════════════════════════════════════════════════════

years = list(range(2015, 2024))
mpesa_users    = [23.6, 26.8, 28.4, 30.0, 31.5, 32.0, 34.2, 36.5, 38.1]
mpesa_value    = [3.7, 4.0, 4.4, 4.9, 5.4, 5.9, 6.5, 7.2, 8.0]  # trillion KES/year
internet_pct   = [16, 18, 22, 27, 30, 34, 38, 44, 52]  # %
smartphone_pct = [20, 25, 30, 36, 42, 48, 54, 61, 68]  # %
ecommerce_bn   = [0.5, 0.8, 1.2, 1.8, 2.5, 2.9, 3.8, 5.2, 7.1]  # KES billion

# County digital divide (simulated)
county_internet = {
    'Nairobi':88, 'Mombasa':65, 'Kisumu':48, 'Nakuru':42, 'Kisii':31,
    'Meru':36, 'Nyeri':39, 'Kakamega':28, 'Bungoma':25, 'Garissa':18,
    'Turkana':12, 'Mandera':9, 'Wajir':11, 'Marsabit':14, 'Samburu':16
}

fig = plt.figure(figsize=(16, 12), constrained_layout=True)
fig.suptitle('Kenya Digital Economy — Data Story 2015–2023',
             fontsize=16, fontweight='bold')
gs = gridspec.GridSpec(3, 3, figure=fig)

# 1. M-Pesa users growth
ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(years, mpesa_users, 'o-', color='#00A651', linewidth=2.5, markersize=7)
ax1.fill_between(years, mpesa_users, alpha=0.15, color='#00A651')
for y, v in zip(years, mpesa_users):
    ax1.text(y, v + 0.3, f'{v:.1f}M', ha='center', fontsize=8)
ax1.set_title('M-Pesa Registered Users', fontweight='bold')
ax1.set_ylabel('Users (Millions)')
ax1.set_ylim(0, 45)
ax1.grid(True, alpha=0.3)

# 2. Digital adoption donut
ax2 = fig.add_subplot(gs[0, 2])
adoption = {'M-Pesa Users': 38.1, 'Smartphone': 68.0, 'Internet': 52.0,
             'e-Commerce': 25.0, 'Digital Banking': 35.0}
dont_have = {k: 100-v for k, v in adoption.items()}
colors2 = ['#00A651','#2196F3','#FF9800','#9C27B0','#E91E63']
ax2.barh(list(adoption.keys()), list(adoption.values()), color=colors2, alpha=0.85)
ax2.barh(list(adoption.keys()), list(dont_have.values()),
          left=list(adoption.values()), color='lightgrey', alpha=0.5)
ax2.set_xlim(0, 100)
ax2.set_title('Digital Adoption Rate (%) 2023', fontweight='bold')
ax2.set_xlabel('Percentage of Population')
for i, (k, v) in enumerate(adoption.items()):
    ax2.text(v + 2, i, f'{v:.0f}%', va='center', fontsize=9, fontweight='bold')

# 3. Internet vs Smartphone — dual line
ax3 = fig.add_subplot(gs[1, :2])
ax3.plot(years, internet_pct, 's-', color='#2196F3', linewidth=2, label='Internet Penetration')
ax3.plot(years, smartphone_pct, 'o-', color='#FF5722', linewidth=2, label='Smartphone Ownership')
ax3.fill_between(years, internet_pct, alpha=0.1, color='#2196F3')
ax3.fill_between(years, smartphone_pct, alpha=0.1, color='#FF5722')
ax3.set_title('Internet & Smartphone Penetration (%)', fontweight='bold')
ax3.set_ylabel('%')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. E-commerce growth bar
ax4 = fig.add_subplot(gs[1, 2])
ax4.bar(years, ecommerce_bn, color=plt.cm.Greens(np.linspace(0.3, 0.9, len(years))))
ax4.set_title('E-Commerce Revenue (KES B)', fontweight='bold')
ax4.set_ylabel('KES Billion')
ax4.tick_params(axis='x', rotation=45)

# 5. County digital divide — horizontal bar
ax5 = fig.add_subplot(gs[2, :])
counties_sorted = sorted(county_internet.items(), key=lambda x: x[1], reverse=True)
county_names = [c[0] for c in counties_sorted]
county_vals  = [c[1] for c in counties_sorted]
colors5 = ['#27AE60' if v > 50 else ('#F39C12' if v > 30 else '#E74C3C') for v in county_vals]
ax5.barh(county_names, county_vals, color=colors5, edgecolor='white')
ax5.axvline(50, color='black', linestyle='--', alpha=0.5, label='50% threshold')
ax5.set_title('Internet Penetration by County — Digital Divide', fontweight='bold')
ax5.set_xlabel('Internet Penetration (%)')
ax5.legend()
for i, v in enumerate(county_vals):
    ax5.text(v + 1, i, f'{v}%', va='center', fontsize=9)

fig.savefig('/tmp/day28_capstone_static.png', dpi=130, bbox_inches='tight')
print('Static capstone dashboard saved to /tmp/day28_capstone_static.png')
plt.close()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(2024)
years = list(range(2015, 2024))
mpesa_users    = [23.6, 26.8, 28.4, 30.0, 31.5, 32.0, 34.2, 36.5, 38.1]
internet_pct   = [16, 18, 22, 27, 30, 34, 38, 44, 52]
ecommerce_bn   = [0.5, 0.8, 1.2, 1.8, 2.5, 2.9, 3.8, 5.2, 7.1]

# ════════════════════════════════════════════════════════════════════
# CAPSTONE PART 2: Interactive Plotly Dashboard
# ════════════════════════════════════════════════════════════════════
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=[
        'M-Pesa Users Growth', 'Internet Penetration (%)',
        'E-Commerce Revenue (KES B)', 'Digital KPIs Summary'
    ],
    specs=[[{'type':'scatter'}, {'type':'bar'}],
           [{'type':'scatter'}, {'type':'indicator'}]]
)

# Row 1, Col 1: M-Pesa line
fig.add_trace(go.Scatter(
    x=years, y=mpesa_users, mode='lines+markers',
    line=dict(color='#00A651', width=2.5),
    fill='tozeroy', fillcolor='rgba(0,166,81,0.1)',
    hovertemplate='%{x}: %{y:.1f}M users<extra></extra>',
    name='M-Pesa Users'
), row=1, col=1)

# Row 1, Col 2: Internet bar
fig.add_trace(go.Bar(
    x=years, y=internet_pct,
    marker_color=internet_pct,
    marker_colorscale='Blues',
    hovertemplate='%{x}: %{y}%<extra></extra>',
    name='Internet %'
), row=1, col=2)

# Row 2, Col 1: E-commerce line
fig.add_trace(go.Scatter(
    x=years, y=ecommerce_bn, mode='lines+markers',
    line=dict(color='#9C27B0', width=2.5),
    fill='tonexty' if False else 'tozeroy',
    fillcolor='rgba(156,39,176,0.1)',
    hovertemplate='%{x}: KES %{y:.1f}B<extra></extra>',
    name='E-Commerce'
), row=2, col=1)

# Row 2, Col 2: KPI Indicator
fig.add_trace(go.Indicator(
    mode='number+delta+gauge',
    value=38.1,
    delta={'reference': 23.6, 'valueformat': '.1f'},
    gauge={'axis': {'range': [0, 60]},
           'bar': {'color': '#00A651'},
           'steps': [{'range': [0, 20], 'color': '#FFCCCC'},
                     {'range': [20, 40], 'color': '#FFFFCC'},
                     {'range': [40, 60], 'color': '#CCFFCC'}]},
    title={'text': 'M-Pesa Users (M)<br>2023 vs 2015'}
), row=2, col=2)

fig.update_layout(
    title_text='Kenya Digital Economy Interactive Dashboard 2015–2023',
    title_font_size=15,
    template='plotly_white',
    height=700,
    showlegend=False
)
fig.write_html('/tmp/day28_interactive_dashboard.html')
fig.write_image('/tmp/day28_interactive_dashboard.png', scale=1.5)
print('Interactive dashboard saved to /tmp/day28_interactive_dashboard.html')
```
