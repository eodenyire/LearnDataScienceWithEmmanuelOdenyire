# Day 12: Images, Heatmaps and Colormesh

## Phase 1 | Month 3 | Week 2 | Day 12

**🎯 Goal:** Display image data, create heatmaps with `imshow`/`pcolormesh`, and add colorbars.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Heatmap: Correlation matrix ───────────────────────────────────────────────
np.random.seed(42)
features = ['GDP', 'Inflation', 'Unemployment', 'FDI', 'Exports', 'PopGrowth', 'Literacy']
n = len(features)

# Synthetic correlation matrix
corr_data = np.array([
    [ 1.00, -0.65, -0.72,  0.81,  0.78, -0.20,  0.69],
    [-0.65,  1.00,  0.55, -0.58, -0.52,  0.15, -0.48],
    [-0.72,  0.55,  1.00, -0.64, -0.60,  0.08, -0.55],
    [ 0.81, -0.58, -0.64,  1.00,  0.72, -0.18,  0.61],
    [ 0.78, -0.52, -0.60,  0.72,  1.00, -0.12,  0.58],
    [-0.20,  0.15,  0.08, -0.18, -0.12,  1.00, -0.09],
    [ 0.69, -0.48, -0.55,  0.61,  0.58, -0.09,  1.00],
])

fig, ax = plt.subplots(figsize=(8, 7))
im = ax.imshow(corr_data, cmap='RdBu_r', vmin=-1, vmax=1, aspect='auto')

for i in range(n):
    for j in range(n):
        ax.text(j, i, f'{corr_data[i,j]:.2f}', ha='center', va='center',
                fontsize=9, color='white' if abs(corr_data[i,j]) > 0.6 else 'black')

ax.set_xticks(range(n)); ax.set_xticklabels(features, rotation=45, ha='right')
ax.set_yticks(range(n)); ax.set_yticklabels(features)
ax.set_title('Correlation Matrix: African Economic Indicators', fontsize=12, fontweight='bold')
plt.colorbar(im, ax=ax, label='Pearson r', fraction=0.046, pad=0.04)
fig.tight_layout()
fig.savefig('/tmp/day12_heatmap.png', dpi=120)
print('Correlation heatmap saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── pcolormesh: Rainfall calendar heatmap ────────────────────────────────────
np.random.seed(7)
weeks = 53
days_per_week = 7
rainfall = np.random.exponential(scale=8, size=(days_per_week, weeks))
rainfall[rainfall > 50] = 0  # no extreme outliers

fig, ax = plt.subplots(figsize=(15, 4))
im = ax.pcolormesh(rainfall, cmap='Blues', edgecolors='white', linewidth=0.3)
ax.set_yticks([0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])
ax.set_yticklabels(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
ax.set_title('Simulated Daily Rainfall (mm) — Nairobi 2023', fontsize=12, fontweight='bold')
ax.set_xlabel('Week of Year')
plt.colorbar(im, ax=ax, label='Rainfall (mm)')
fig.tight_layout()
fig.savefig('/tmp/day12_calendar.png', dpi=120)
print('Calendar heatmap saved')
plt.close()
```
