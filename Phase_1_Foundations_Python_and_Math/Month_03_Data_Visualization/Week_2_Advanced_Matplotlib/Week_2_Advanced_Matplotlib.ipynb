```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

## 1. Plot Styles

```python
# Available styles
print('Available styles:', plt.style.available[:8])

with plt.style.context('seaborn-v0_8-whitegrid'):
    fig, ax = plt.subplots(figsize=(7, 4))
    x = np.linspace(0, 10, 100)
    ax.plot(x, np.sin(x), lw=2, label='sin')
    ax.plot(x, np.cos(x), lw=2, label='cos')
    ax.legend()
    ax.set_title('seaborn-whitegrid style')
    plt.tight_layout()
    plt.show()
```

## 2. Annotations and Text

```python
rng = np.random.default_rng(0)
x = np.arange(1, 13)
sales = rng.integers(50, 200, 12)
peak_idx = sales.argmax()

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x, sales, marker='o', color='steelblue', linewidth=2)
ax.fill_between(x, sales, alpha=0.15, color='steelblue')

# Annotate peak
ax.annotate(f'Peak: {sales[peak_idx]}',
            xy=(x[peak_idx], sales[peak_idx]),
            xytext=(x[peak_idx]+1, sales[peak_idx]+20),
            arrowprops=dict(arrowstyle='->', color='red'),
            color='red', fontsize=11)

ax.set_xticks(x)
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
ax.set_title('Monthly Sales with Peak Annotation')
ax.set_ylabel('Units Sold')
plt.tight_layout()
plt.show()
```

## 3. Twin Axes

```python
months = np.arange(1, 13)
revenue = rng.integers(100, 500, 12)
customers = rng.integers(50, 200, 12)

fig, ax1 = plt.subplots(figsize=(10, 5))
ax2 = ax1.twinx()

ax1.bar(months, revenue, color='steelblue', alpha=0.6, label='Revenue ($K)')
ax2.plot(months, customers, color='orange', marker='o', linewidth=2, label='Customers')

ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue ($K)', color='steelblue')
ax2.set_ylabel('Customers', color='orange')
ax1.set_title('Revenue and Customers (Dual Axis)')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
plt.tight_layout()
plt.show()
```

## 4. Heatmaps with Matplotlib

```python
data = rng.integers(0, 100, (5, 5))
labels = ['Math', 'Science', 'English', 'History', 'Art']

fig, ax = plt.subplots(figsize=(7, 5))
im = ax.imshow(data, cmap='YlOrRd')
plt.colorbar(im, ax=ax)

ax.set_xticks(range(5)); ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticks(range(5)); ax.set_yticklabels(labels)
ax.set_title('Score Heatmap')

for i in range(5):
    for j in range(5):
        ax.text(j, i, data[i, j], ha='center', va='center', fontsize=10)
plt.tight_layout()
plt.show()
```
