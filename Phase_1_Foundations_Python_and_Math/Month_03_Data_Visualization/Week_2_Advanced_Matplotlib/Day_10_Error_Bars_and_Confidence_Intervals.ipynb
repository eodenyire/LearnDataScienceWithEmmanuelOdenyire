# Day 10: Error Bars and Confidence Intervals

## Phase 1 | Month 3 | Week 2 | Day 10

**🎯 Goal:** Show data uncertainty using error bars, confidence bands, and box plots.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

np.random.seed(42)

# ── Symmetric error bars (mean ± std) ─────────────────────────────────────────
counties  = ['Nairobi', 'Mombasa', 'Kisumu', 'Nakuru', 'Eldoret', 'Thika']
maize_yields = [2800, 1900, 2200, 3100, 2700, 3000]  # kg/ha mean
errors       = [320,   280,  410,  380,  295,  350]  # std deviation

fig, ax = plt.subplots(figsize=(9, 5))
ax.errorbar(range(len(counties)), maize_yields, yerr=errors,
            fmt='o', color='steelblue', ecolor='tomato', elinewidth=2,
            capsize=6, capthick=2, markersize=8, linewidth=0)
ax.set_xticks(range(len(counties)))
ax.set_xticklabels(counties)
ax.set_title('Maize Yield by County (Mean ± Std Dev) — Kenya 2023', fontsize=13, fontweight='bold')
ax.set_ylabel('Yield (kg/ha)')
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day10_errorbars.png', dpi=120)
print('Error bar chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

np.random.seed(0)

# ── Confidence band on a trend line ──────────────────────────────────────────
n_samples = 60
studyhours = np.random.uniform(1, 12, n_samples)
scores = 35 + 4.8 * studyhours + np.random.normal(0, 10, n_samples)

# Bootstrap confidence interval
x_range = np.linspace(1, 12, 100)
lower, upper = [], []
for x_val in x_range:
    nearby = scores[(studyhours > x_val - 1.5) & (studyhours < x_val + 1.5)]
    if len(nearby) >= 5:
        ci = stats.t.interval(0.95, df=len(nearby)-1, loc=nearby.mean(),
                              scale=stats.sem(nearby))
        lower.append(ci[0]); upper.append(ci[1])
    else:
        m, b = np.polyfit(studyhours, scores, 1)
        lower.append(m*x_val+b - 10); upper.append(m*x_val+b + 10)

m, b = np.polyfit(studyhours, scores, 1)
y_trend = m * x_range + b

fig, ax = plt.subplots(figsize=(9, 6))
ax.scatter(studyhours, scores, alpha=0.5, color='steelblue', s=40, label='Students')
ax.plot(x_range, y_trend, 'r-', linewidth=2, label=f'Trend: y={m:.1f}x+{b:.1f}')
ax.fill_between(x_range, lower, upper, alpha=0.2, color='red', label='95% CI band')
ax.set_title('Study Hours vs Exam Score with 95% Confidence Band', fontsize=12, fontweight='bold')
ax.set_xlabel('Daily Study Hours')
ax.set_ylabel('Exam Score (%)')
ax.legend()
ax.grid(True, alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day10_confband.png', dpi=120)
print('Confidence band chart saved')
plt.close()
```
