```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.backends.backend_pdf as pdf_backend
import numpy as np

# ════════════════════════════════════════════════════════════════════
# WEEK 2 PROJECT: NSE Annual Market Performance Report 2023
# ════════════════════════════════════════════════════════════════════
np.random.seed(42)

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
# NSE 20 Share Index
nse20  = [1750, 1780, 1730, 1690, 1660, 1640, 1625, 1610, 1590, 1570, 1585, 1600]
# Turnover (KES Billion)
turnover = [12.5, 14.2, 11.8, 10.5, 9.8, 11.2, 10.9, 12.4, 11.7, 10.8, 13.5, 15.2]
# Top 5 stocks performance %
stocks = {'SCOM': [2.1,-1.5, 3.2,-0.8, 1.5,-2.1, 0.8,-1.2, 2.5,-0.5, 1.8, 2.9],
          'EQTY': [3.5, 2.1,-2.5, 1.8,-1.2, 2.8,-0.5, 1.5,-2.8, 3.2, 1.1, 2.5],
          'KCB' : [1.8,-0.9, 2.5, 0.5,-2.1, 1.8, 1.2,-0.8, 3.1,-1.5, 0.9, 1.8],
          'EABL': [-1.2, 2.5, 1.8,-1.5, 0.9, 2.1,-1.8, 0.5, 1.9,-2.1, 2.2, 1.5],
          'BAT' : [ 0.8, 1.5,-1.2, 2.8,-0.5, 1.2, 0.9,-2.1, 0.8, 1.5,-1.8, 2.1]}

with plt.style.context('seaborn-v0_8-whitegrid'):
    fig = plt.figure(figsize=(16, 11), constrained_layout=True)
    fig.suptitle('Nairobi Securities Exchange — Annual Report 2023',
                 fontsize=16, fontweight='bold')
    gs = gridspec.GridSpec(2, 3, figure=fig)

    # 1. NSE 20 trend
    ax1 = fig.add_subplot(gs[0, :2])
    ax1.plot(months, nse20, 'o-', color='navy', linewidth=2.5, markersize=7)
    ax1.fill_between(months, nse20, 1500, alpha=0.1, color='navy')
    ax1.axhline(np.mean(nse20), color='red', linestyle='--', alpha=0.7, label=f'Mean: {np.mean(nse20):.0f}')
    ax1.set_title('NSE 20 Share Index 2023', fontweight='bold')
    ax1.set_ylabel('Index Points')
    ax1.legend()
    ax1.set_ylim(1450, 1900)

    # 2. Monthly turnover bar
    ax2 = fig.add_subplot(gs[0, 2])
    colors2 = ['#27AE60' if t > np.mean(turnover) else '#E74C3C' for t in turnover]
    ax2.bar(months, turnover, color=colors2, alpha=0.85, edgecolor='white')
    ax2.axhline(np.mean(turnover), color='black', linestyle='--', alpha=0.5, linewidth=1.5)
    ax2.set_title('Monthly Turnover (KES B)', fontweight='bold')
    ax2.set_ylabel('KES Billion')
    ax2.tick_params(axis='x', rotation=45)

    # 3. Stock performance heatmap
    ax3 = fig.add_subplot(gs[1, :2])
    stock_matrix = np.array([v for v in stocks.values()])
    im = ax3.imshow(stock_matrix, cmap='RdYlGn', aspect='auto', vmin=-4, vmax=4)
    ax3.set_xticks(range(12)); ax3.set_xticklabels(months)
    ax3.set_yticks(range(5));  ax3.set_yticklabels(list(stocks.keys()), fontweight='bold')
    for i in range(5):
        for j in range(12):
            ax3.text(j, i, f'{stock_matrix[i,j]:+.1f}', ha='center', va='center', fontsize=8)
    ax3.set_title('Monthly Stock Returns (%) — Top 5 Stocks', fontweight='bold')
    plt.colorbar(im, ax=ax3, label='Return (%)', orientation='vertical')

    # 4. Sector allocation donut
    ax4 = fig.add_subplot(gs[1, 2])
    sectors = ['Banking', 'Telecom', 'Manufacturing', 'Energy', 'Agriculture']
    shares  = [38, 22, 18, 12, 10]
    colors4 = ['#2980B9','#00A651','#E67E22','#E74C3C','#27AE60']
    ax4.pie(shares, labels=sectors, autopct='%1.0f%%', colors=colors4,
            wedgeprops={'width':0.55, 'edgecolor':'white'},
            textprops={'fontsize':9})
    ax4.set_title('Market Capitalisation by Sector', fontweight='bold')

    fig.savefig('/tmp/day14_nse_report.png', dpi=130, bbox_inches='tight')
    print('NSE Annual Report dashboard saved to /tmp/day14_nse_report.png')
    plt.close()
```
