# Day 9: Annotations and Text in Plots

## Phase 1 | Month 3 | Week 2 | Day 9

**🎯 Goal:** Add meaningful annotations, callouts, arrows, and text boxes to communicate insights directly on charts.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── ax.annotate() fundamentals ────────────────────────────────────────────────
years = list(range(2010, 2024))
mpesa_agents = [15000, 28000, 52000, 89000, 110000, 122000, 133000, 144000,
                160000, 176000, 191000, 200000, 213000, 228000]

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(years, mpesa_agents, 'o-', color='#00A651', linewidth=2.5, markersize=6)
ax.fill_between(years, mpesa_agents, alpha=0.1, color='#00A651')

# Key events annotation
events = [
    (2012, 52000, 'M-Pesa\nlaunches in\nTanzania', (2013, 75000)),
    (2014, 110000, 'M-Shwari\nlaunched', (2015, 90000)),
    (2020, 200000, 'COVID:\ndigi-push', (2021, 185000)),
]

for x, y, label, xytext in events:
    ax.annotate(label, xy=(x, y), xytext=xytext,
                arrowprops=dict(arrowstyle='->', color='darkred', lw=1.5),
                fontsize=9, color='darkred',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow', edgecolor='darkred', alpha=0.8))

ax.set_title('M-Pesa Agent Network Growth 2010–2023', fontsize=13, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Agents')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
ax.grid(True, alpha=0.3)
fig.tight_layout()
fig.savefig('/tmp/day09_annotations.png', dpi=120)
print('Annotated chart saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── Text boxes and span annotations ──────────────────────────────────────────
np.random.seed(42)
days = np.arange(1, 366)
covid_index = np.concatenate([
    np.random.normal(100, 3, 70),    # Pre-COVID normal
    np.linspace(100, 45, 80),        # March 2020: crash
    np.linspace(45, 65, 60),         # Partial recovery
    np.linspace(65, 82, 155),        # Gradual recovery
])

fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(days, covid_index, color='steelblue', linewidth=1.5)

# Span to highlight lockdown period
ax.axvspan(70, 150, alpha=0.15, color='red', label='Lockdown Period')
ax.axhline(100, color='green', linestyle='--', alpha=0.6, label='Pre-COVID baseline')
ax.axhline(covid_index.min(), color='tomato', linestyle=':', alpha=0.6, label=f'Lowest ({covid_index.min():.0f})')

# Text box
ax.text(0.02, 0.95, 'NSE Performance Index\n2020 (simulated)', transform=ax.transAxes,
        fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

ax.set_title('Kenya NSE 20 Index Performance — 2020 COVID Impact', fontsize=13, fontweight='bold')
ax.set_xlabel('Day of Year 2020')
ax.set_ylabel('Index Level')
ax.legend()
ax.grid(True, alpha=0.25)
fig.tight_layout()
fig.savefig('/tmp/day09_span.png', dpi=120)
print('Span annotation chart saved')
plt.close()
```
