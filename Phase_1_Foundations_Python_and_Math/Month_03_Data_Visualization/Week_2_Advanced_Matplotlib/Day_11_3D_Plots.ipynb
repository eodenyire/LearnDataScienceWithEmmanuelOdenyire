# Day 11: 3D Plots with Matplotlib

## Phase 1 | Month 3 | Week 2 | Day 11

**🎯 Goal:** Create 3D surface, wireframe, scatter, and bar charts using `mpl_toolkits.mplot3d`.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# ── 3D Surface plot: price surface ────────────────────────────────────────────
x = np.linspace(-3, 3, 60)
y = np.linspace(-3, 3, 60)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)
fig.colorbar(surf, ax=ax, shrink=0.5, label='Amplitude')
ax.set_title('3D Surface: sin(√(x²+y²))', fontsize=13, fontweight='bold')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.savefig('/tmp/day11_surface.png', dpi=120)
print('3D surface saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# ── 3D Scatter: African cities (lat, lon, population) ─────────────────────────
np.random.seed(5)
city_lat  = [-1.28, -4.04, -17.82, 6.37, 5.55, 9.05, 30.05, -25.74, 11.55, -18.91]
city_lon  = [36.82, 39.67,  31.05, 2.37,-0.20, 7.49, 31.23,  28.18, 43.15,  47.53]
city_pop  = [4.3,   1.2,    1.5,   2.7,  2.5,  3.5,  10.4,   4.6,   0.98,  1.1]  # millions
city_gdp  = [28, 12, 8, 7, 6, 9, 110, 12, 4, 5]  # USD billion approx

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(city_lon, city_lat, city_pop,
                s=[g*15 for g in city_gdp],
                c=city_gdp, cmap='plasma',
                alpha=0.85, edgecolors='white', linewidth=0.5)
plt.colorbar(sc, ax=ax, label='City GDP (USD B)', shrink=0.6)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Population (M)')
ax.set_title('African Major Cities: Location vs Population vs GDP', fontsize=11, fontweight='bold')
fig.savefig('/tmp/day11_3d_scatter.png', dpi=120)
print('3D scatter saved')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# ── 3D Bar chart: quarterly revenue by county ─────────────────────────────────
counties   = ['Nairobi', 'Mombasa', 'Kisumu', 'Nakuru']
quarters   = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = np.array([[42, 45, 48, 50],
                     [15, 16, 17, 18],
                     [ 8,  9, 10, 11],
                     [12, 13, 14, 15]])

fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')
colors = ['#3498DB', '#E74C3C', '#27AE60', '#F39C12']

for i, (county, color) in enumerate(zip(counties, colors)):
    ax.bar(np.arange(4), revenue[i], zs=i, zdir='y',
           color=color, alpha=0.8, width=0.7)

ax.set_xticks([0,1,2,3])
ax.set_xticklabels(quarters)
ax.set_yticks([0,1,2,3])
ax.set_yticklabels(counties)
ax.set_zlabel('Revenue (KES B)')
ax.set_title('Quarterly Revenue by County (KES Billion)', fontsize=12, fontweight='bold')
fig.savefig('/tmp/day11_3d_bar.png', dpi=120)
print('3D bar chart saved')
plt.close()
```
