# Day 13: Saving and Exporting Figures

## Phase 1 | Month 3 | Week 2 | Day 13

**🎯 Goal:** Save Matplotlib figures in PNG, SVG, PDF, and other formats at the correct resolution for reports and presentations.
**⏱️ Study Session:** ~2 hours of this 15-hour day

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
import os

# ── savefig() parameters explained ───────────────────────────────────────────
print('Key savefig() parameters:')
print('  dpi=300   → print quality (300 DPI)')
print('  dpi=150   → screen / presentation quality')
print('  dpi=72    → web / low resolution')
print('  format=   → png, svg, pdf, eps, jpeg, tiff')
print('  bbox_inches="tight" → trim whitespace')
print('  transparent=True    → transparent background')
print('  facecolor="white"   → explicit background color')

x = np.linspace(0, 2*np.pi, 100)
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, np.sin(x), 'b-', linewidth=2, label='sin(x)')
ax.plot(x, np.cos(x), 'r--', linewidth=2, label='cos(x)')
ax.set_title('Export Format Comparison', fontsize=13, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Save in multiple formats
formats = ['png', 'svg', 'pdf']
for fmt in formats:
    path = f'/tmp/day13_output.{fmt}'
    fig.savefig(path, dpi=150, bbox_inches='tight', format=fmt)
    size = os.path.getsize(path)
    print(f'Saved {fmt.upper():4}: {path}  ({size:,} bytes)')
plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── DPI comparison ────────────────────────────────────────────────────────────
np.random.seed(10)
data = np.random.randn(200)

for dpi, label in [(72, 'web'), (150, 'screen'), (300, 'print')]:
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.hist(data, bins=25, color='steelblue', edgecolor='white', alpha=0.85)
    ax.set_title(f'DPI={dpi} ({label} quality)', fontsize=12, fontweight='bold')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    path = f'/tmp/day13_dpi_{dpi}.png'
    fig.savefig(path, dpi=dpi, bbox_inches='tight')
    import os
    print(f'DPI={dpi:3} | File size: {os.path.getsize(path):>8,} bytes | Path: {path}')
    plt.close()
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf_backend
import numpy as np

# ── Multi-page PDF: quarterly report ─────────────────────────────────────────
fig1, ax1 = plt.subplots(figsize=(10, 6))
months = ['Jan','Feb','Mar','Apr','May','Jun']
ax1.bar(months, [45, 48, 43, 52, 50, 55], color='steelblue')
ax1.set_title('H1 Revenue (KES Billion)', fontweight='bold')
ax1.set_ylabel('KES Billion')

fig2, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(months, [9.0, 9.2, 9.9, 9.5, 8.2, 7.9], 'ro-', linewidth=2)
ax2.set_title('H1 Inflation (%)', fontweight='bold')
ax2.set_ylabel('%')
ax2.grid(True, alpha=0.3)

with pdf_backend.PdfPages('/tmp/day13_quarterly_report.pdf') as pdf:
    pdf.savefig(fig1, bbox_inches='tight')
    pdf.savefig(fig2, bbox_inches='tight')
    d = pdf.infodict()
    d['Title'] = 'Kenya Quarterly Economic Report'
    d['Author'] = 'Emmanuel Odenyire'
    d['Subject'] = 'Data Science Bootcamp'

plt.close('all')
import os
print(f'Multi-page PDF saved: {os.path.getsize("/tmp/day13_quarterly_report.pdf"):,} bytes')
```
