# Day 17: Hypothesis Testing Fundamentals

## Phase 1 | Month 4 | Week 3 | Day 17

**🎯 Goal:** Understand the hypothesis testing framework — null hypothesis, p-values, Type I/II errors.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats

# Example: Safaricom claims average call drop rate ≤ 2%
# Regulator samples 200 calls, finds 7 dropped
n = 200
x_dropped = 7
p_hat = x_dropped / n
p_null = 0.02   # claimed rate (H0: p ≤ 0.02)

# One-proportion z-test
z_stat = (p_hat - p_null) / np.sqrt(p_null*(1-p_null)/n)
p_value = 1 - stats.norm.cdf(z_stat)   # one-tailed (right)

print('=== One-Proportion Z-Test ===')
print(f'H0: p ≤ 0.02 (Safaricom claim)')
print(f'H1: p > 0.02 (Regulator concern)')
print(f'Observed drop rate: {p_hat:.4f} ({p_hat*100:.1f}%)')
print(f'Z-statistic: {z_stat:.4f}')
print(f'P-value: {p_value:.4f}')
print()
alpha = 0.05
if p_value < alpha:
    print(f'p={p_value:.4f} < α={alpha} → REJECT H0. Evidence of higher drop rate!')
else:
    print(f'p={p_value:.4f} ≥ α={alpha} → Fail to reject H0.')
```

## The Hypothesis Testing Workflow

1. **State H₀ and H₁**
2. **Choose α** (significance level, typically 0.05)
3. **Collect data and compute test statistic**
4. **Calculate p-value**
5. **Decision**: if p < α → reject H₀
6. **Interpret** in context

## Error Types

| Decision | H₀ True | H₀ False |
|----------|---------|----------|
| Reject H₀ | **Type I Error** (α) | Correct |
| Fail to Reject H₀ | Correct | **Type II Error** (β) |

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Visualise p-value and rejection region
z_stat = 1.96   # example statistic
x = np.linspace(-4, 4, 400)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, stats.norm.pdf(x), 'navy', lw=2.5)
ax.fill_between(x, stats.norm.pdf(x), where=(x >= 1.645), color='tomato', alpha=0.5, label='Rejection region (α=0.05, one-tail)')
ax.axvline(z_stat, color='green', lw=2, ls='--', label=f'Test statistic z={z_stat}')
ax.set_title('Hypothesis Test Rejection Region', fontweight='bold')
ax.set_xlabel('Z-score')
ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day17_hypothesis.png', dpi=120)
print('Saved /tmp/day17_hypothesis.png')
plt.close()
```
