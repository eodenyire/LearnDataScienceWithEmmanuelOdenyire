# Day 16: Confidence Intervals

## Phase 1 | Month 4 | Week 3 | Day 16

**🎯 Goal:** Build and interpret confidence intervals for means and proportions.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(10)
# Sample: wait times (minutes) at 35 Equity Bank branches
wait_times = np.random.normal(loc=12.5, scale=4.2, size=35)
n = len(wait_times)
x_bar = wait_times.mean()
s     = wait_times.std(ddof=1)
se    = s / np.sqrt(n)

for conf in [0.90, 0.95, 0.99]:
    ci = stats.t.interval(conf, df=n-1, loc=x_bar, scale=se)
    t_crit = stats.t.ppf((1+conf)/2, df=n-1)
    print(f'{int(conf*100)}% CI: ({ci[0]:.2f}, {ci[1]:.2f}) min  |  t*={t_crit:.3f}  |  width={ci[1]-ci[0]:.2f} min')
```

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Visualise CI across 50 repeated samples
np.random.seed(0)
true_mean = 12.5
samples_cover = 0
fig, ax = plt.subplots(figsize=(14, 8))

for i in range(50):
    s = np.random.normal(true_mean, 4.2, 35)
    ci = stats.t.interval(0.95, df=34, loc=s.mean(), scale=s.std(ddof=1)/np.sqrt(35))
    color = 'steelblue' if ci[0] <= true_mean <= ci[1] else 'tomato'
    ax.plot([ci[0], ci[1]], [i, i], color=color, lw=2)
    ax.plot(s.mean(), i, 'o', color=color, markersize=4)
    if ci[0] <= true_mean <= ci[1]: samples_cover += 1

ax.axvline(true_mean, color='black', lw=2, ls='--', label='True mean (12.5 min)')
ax.set_title(f'50 Repeated 95% CIs — {samples_cover} contain true mean', fontweight='bold')
ax.set_xlabel('Wait Time (minutes)')
ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day16_ci.png', dpi=120)
print(f'Saved. {samples_cover}/50 ({samples_cover*2}%) CIs contain true mean.')
plt.close()
```

## CI for a Proportion

Used for survey results: e.g., '68% of Kenyans use mobile money' from a poll of n=1200.

```python
import numpy as np
from scipy import stats

# AfroBarometer-style poll: 68% of 1200 respondents use mobile money
n, p_hat = 1200, 0.68
se_prop = np.sqrt(p_hat*(1-p_hat)/n)
ci = stats.norm.interval(0.95, loc=p_hat, scale=se_prop)
print(f'Sample proportion: {p_hat:.2%}')
print(f'95% CI for proportion: ({ci[0]:.4f}, {ci[1]:.4f})')
print(f'Margin of Error: ±{(ci[1]-ci[0])/2*100:.2f} percentage points')
```

## 📋 Summary

- ✓ CI = point estimate ± (critical value × SE)
- ✓ Use t-distribution when σ unknown (almost always)
- ✓ Wider CI = higher confidence, less precision
- ✓ 95% CI: 95% of such intervals will contain the true parameter
