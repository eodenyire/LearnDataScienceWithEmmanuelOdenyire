# Day 19: Chi-Square Tests and ANOVA

## Phase 1 | Month 4 | Week 3 | Day 19

**🎯 Goal:** Test categorical associations with chi-square and compare more than two group means with ANOVA.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from scipy import stats

# CHI-SQUARE TEST OF INDEPENDENCE
# Does preferred payment method differ by county?
# Observed frequencies
contingency = np.array([
    # M-Pesa  Card  Cash
    [  350,    85,   65],   # Nairobi
    [  180,    30,   90],   # Mombasa
    [  120,    15,   65],   # Kisumu
])

chi2, p, dof, expected = stats.chi2_contingency(contingency)
print('=== Chi-Square Test: Payment Method vs County ===')
print(f'Chi2 = {chi2:.4f},  df = {dof},  p = {p:.6f}')
if p < 0.05:
    print('Reject H0: payment preference DOES differ by county.')
else:
    print('Fail to reject H0.')
```

```python
import numpy as np
from scipy import stats

# ONE-WAY ANOVA
# Do average M-Pesa transaction amounts differ across 4 age groups?
np.random.seed(5)
youth    = np.random.normal(1500,  400, 80)   # 18-25
Adult    = np.random.normal(3800,  900, 80)   # 26-40
Middle   = np.random.normal(6000, 1200, 80)   # 41-55
Senior   = np.random.normal(4200,  800, 80)   # 55+

f_stat, p_anova = stats.f_oneway(youth, Adult, Middle, Senior)
print('=== One-Way ANOVA: Transaction Amount by Age Group ===')
print(f'F = {f_stat:.4f},  p = {p_anova:.8f}')
if p_anova < 0.05:
    print('Reject H0: at least one group mean differs significantly.')

# Post-hoc: Tukey HSD
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
all_vals = np.concatenate([youth, Adult, Middle, Senior])
groups   = ['Youth']*80 + ['Adult']*80 + ['Middle']*80 + ['Senior']*80
tukey = pairwise_tukeyhsd(all_vals, groups, alpha=0.05)
print('\nTukey HSD results:')
print(tukey)
```
