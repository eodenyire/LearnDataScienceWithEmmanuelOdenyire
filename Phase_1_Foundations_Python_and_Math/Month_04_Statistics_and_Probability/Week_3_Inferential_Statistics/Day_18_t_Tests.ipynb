# Day 18: t-Tests (One-Sample, Two-Sample, Paired)

## Phase 1 | Month 4 | Week 3 | Day 18

**🎯 Goal:** Apply one-sample, independent two-sample, and paired t-tests to business problems.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats

np.random.seed(42)

# ONE-SAMPLE t-TEST
# Is average Nairobi commute time > 45 minutes? (CBK employee survey)
commute = np.random.normal(49, 12, 40)
t_stat, p_val = stats.ttest_1samp(commute, popmean=45, alternative='greater')
print('=== One-Sample t-Test: Commute Time ===')
print(f'Sample mean: {commute.mean():.2f} min (H0: μ=45)')
print(f't = {t_stat:.4f},  p = {p_val:.4f}')
print(f'Decision: {"Reject H0" if p_val < 0.05 else "Fail to reject H0"} at α=0.05\n')

# INDEPENDENT TWO-SAMPLE t-TEST
# Do Nairobi and Mombasa M-Pesa agents have different average float balances?
nai_float = np.random.normal(25000, 8000, 45)
msa_float = np.random.normal(21000, 7500, 38)
t_stat2, p_val2 = stats.ttest_ind(nai_float, msa_float, equal_var=False)  # Welch
print('=== Two-Sample t-Test (Welch): Float Balance ===')
print(f'Nairobi mean: KES {nai_float.mean():,.0f}  |  Mombasa mean: KES {msa_float.mean():,.0f}')
print(f't = {t_stat2:.4f},  p = {p_val2:.4f}')
print(f'Decision: {"Reject H0" if p_val2 < 0.05 else "Fail to reject H0"} at α=0.05\n')

# PAIRED t-TEST
# Did a USSD menu redesign reduce customer task time?
before = np.random.normal(75, 15, 30)
after  = before - np.random.normal(8, 5, 30)   # redesign improved
t_stat3, p_val3 = stats.ttest_rel(before, after)
print('=== Paired t-Test: USSD Redesign ===')
print(f'Before mean: {before.mean():.2f}s  |  After mean: {after.mean():.2f}s')
print(f't = {t_stat3:.4f},  p = {p_val3:.6f}')
print(f'Decision: {"Reject H0" if p_val3 < 0.05 else "Fail to reject H0"} at α=0.05')
```

## Choosing the Right t-Test

| Situation | Test |
|-----------|------|
| One sample vs known value | One-sample t-test |
| Two independent groups | Two-sample t-test (Welch) |
| Same subjects before/after | Paired t-test |

**Assumptions:** 
- Approximately normal (or n ≥ 30)
- Independent observations
- Welch's t-test: no equal-variance assumption needed
