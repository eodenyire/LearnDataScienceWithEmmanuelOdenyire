# Day 15: Introduction to Statistical Inference

## Phase 1 | Month 4 | Week 3 | Day 15

**🎯 Goal:** Understand the goals of statistical inference — estimation and hypothesis testing.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
# Population: daily sales at all Nairobi supermarkets
population = np.random.normal(loc=85_000, scale=22_000, size=500_000)

# KNBS-style: take many samples and study the sampling distribution
sample_means = [np.random.choice(population, 50).mean() for _ in range(3000)]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
axes[0].hist(population, bins=80, color='orange', alpha=0.6, edgecolor='white')
axes[0].axvline(population.mean(), color='red', lw=2)
axes[0].set_title('Population Distribution', fontweight='bold')
axes[0].set_xlabel('Daily Sales (KES)')

axes[1].hist(sample_means, bins=60, color='steelblue', alpha=0.8, edgecolor='white')
axes[1].axvline(population.mean(), color='red', lw=2, label='Population mean')
axes[1].set_title('Sampling Distribution of X̄ (n=50)', fontweight='bold')
axes[1].set_xlabel('Sample Mean (KES)')
axes[1].legend()

plt.tight_layout()
fig.savefig('/tmp/day15_inference.png', dpi=120)
print(f'Population μ: KES {population.mean():,.0f}')
print(f'Mean of sample means: KES {np.mean(sample_means):,.0f}')
print(f'SE (theoretical): KES {population.std()/np.sqrt(50):,.0f}')
print(f'SE (empirical):   KES {np.std(sample_means):,.0f}')
plt.close()
```

## Two Main Tasks of Statistical Inference

1. **Estimation** — use sample statistics to estimate population parameters
   - Point estimate: a single value (e.g., x̄ = 85,200)
   - Interval estimate: a range with confidence level (e.g., 95% CI: [83,100; 87,300])

2. **Hypothesis Testing** — use data to evaluate competing claims
   - H₀ (null): status quo / no effect
   - H₁ (alternative): what you want to prove

```python
import numpy as np
from scipy import stats

# Point vs interval estimation
np.random.seed(7)
# Sample of 40 M-Pesa daily transaction amounts
sample = np.random.lognormal(mean=8.0, sigma=1.1, size=40)

point_est = sample.mean()
se = sample.std(ddof=1) / np.sqrt(len(sample))
ci_95 = stats.t.interval(0.95, df=len(sample)-1, loc=point_est, scale=se)

print(f'Sample mean (point estimate): KES {point_est:,.2f}')
print(f'95% Confidence Interval: KES ({ci_95[0]:,.2f}, {ci_95[1]:,.2f})')
print(f'Width of CI: KES {ci_95[1]-ci_95[0]:,.2f}')
```
