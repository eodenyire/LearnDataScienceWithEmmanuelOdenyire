```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(2024)
N = 1000
df = pd.DataFrame({
    'branch':       np.random.choice(['Nairobi CBD','Westlands','Mombasa','Kisumu','Nakuru'], N),
    'account_type': np.random.choice(['Savings','Current','Fixed Deposit'], N, p=[0.5,0.35,0.15]),
    'balance_kes':  np.random.lognormal(10.5, 1.1, N).round(0),
    'loan_default': np.random.binomial(1, 0.05, N),
    'satisfaction': np.random.choice(['Low','Med','High'], N, p=[0.2,0.45,0.35]),
})

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1 – Balance by branch (ANOVA insight)
branch_means = df.groupby('branch')['balance_kes'].mean().sort_values(ascending=False)
axes[0,0].bar(branch_means.index, branch_means.values/1000, color='steelblue')
axes[0,0].set_title('Mean Balance by Branch', fontweight='bold')
axes[0,0].set_ylabel('Balance (KES thousands)')
axes[0,0].tick_params(axis='x', rotation=25)

# 2 – Default rate by account type (chi-square insight)
default_rate = df.groupby('account_type')['loan_default'].mean()
axes[0,1].bar(default_rate.index, default_rate.values*100, color='tomato')
axes[0,1].set_title('Default Rate by Account Type', fontweight='bold')
axes[0,1].set_ylabel('Default Rate (%)')

# 3 – Balance distribution
axes[1,0].hist(df['balance_kes']/1000, bins=50, color='green', alpha=0.7, edgecolor='white')
axes[1,0].set_title('Account Balance Distribution', fontweight='bold')
axes[1,0].set_xlabel('Balance (KES thousands)')

# 4 – Satisfaction counts
sat = df['satisfaction'].value_counts()
axes[1,1].bar(sat.index, sat.values, color='purple', alpha=0.7)
axes[1,1].set_title('Customer Satisfaction', fontweight='bold')

plt.tight_layout()
fig.savefig('/tmp/day21_project.png', dpi=120)
print('Dashboard saved')
plt.close()
```

## 📋 Week 3 Summary

| Test | When to Use | Python |
|------|-------------|--------|
| One-sample t | Mean vs known value | `stats.ttest_1samp` |
| Two-sample t (Welch) | Compare 2 group means | `stats.ttest_ind` |
| Paired t | Before/after same subjects | `stats.ttest_rel` |
| Chi-square | Categorical independence | `stats.chi2_contingency` |
| One-way ANOVA | 3+ group means | `stats.f_oneway` |
| Pearson/Spearman | Correlation | `stats.pearsonr/spearmanr` |
