# Day 20: Correlation Analysis

## Phase 1 | Month 4 | Week 3 | Day 20

**🎯 Goal:** Measure and visualise linear and monotonic relationships between variables.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(7)
n = 200
# Nairobi household data
income = np.random.normal(60000, 20000, n)
rent   = income * 0.30 + np.random.normal(0, 3000, n)
food   = income * 0.25 + np.random.normal(0, 2000, n)
saving = income * 0.15 - np.random.normal(0, 2500, n)

df = pd.DataFrame({'Income': income, 'Rent': rent, 'Food': food, 'Savings': saving})

# Pearson correlation
print('Pearson Correlation Matrix:')
print(df.corr(method='pearson').round(3))

# Individual correlation tests
r, p = stats.pearsonr(income, rent)
print(f'\nIncome-Rent: r={r:.4f}, p={p:.4f}')
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(7)
n = 200
income = np.random.normal(60000, 20000, n)
rent   = income * 0.30 + np.random.normal(0, 3000, n)
food   = income * 0.25 + np.random.normal(0, 2000, n)
saving = income * 0.15 - np.random.normal(0, 2500, n)
df = pd.DataFrame({'Income':income,'Rent':rent,'Food':food,'Savings':saving})

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col, color in zip(axes, ['Rent','Food','Savings'], ['steelblue','tomato','green']):
    ax.scatter(income, df[col], alpha=0.4, color=color, s=20)
    m, b, r, p, _ = stats.linregress(income, df[col])
    x_line = np.linspace(income.min(), income.max(), 100)
    ax.plot(x_line, m*x_line+b, 'k-', lw=2)
    ax.set_title(f'Income vs {col}\nr={r:.3f}', fontweight='bold')
    ax.set_xlabel('Income (KES)')
    ax.set_ylabel(col)
plt.tight_layout()
fig.savefig('/tmp/day20_correlation.png', dpi=120)
print('Saved /tmp/day20_correlation.png')
plt.close()
```

## Pearson vs Spearman Correlation

| Method | Measures | Use When |
|--------|----------|----------|
| **Pearson r** | Linear relationship | Both variables are continuous, normally distributed |
| **Spearman ρ** | Monotonic relationship | Ordinal data or non-linear relationship |
| **Kendall τ** | Concordance | Small samples, many ties |

⚠️ **Correlation ≠ Causation** — always think critically!

```python
import numpy as np
from scipy import stats

# Spearman for skewed data (e.g., M-Pesa amounts and credit score)
np.random.seed(3)
amount = np.random.lognormal(8, 1.2, 100)
credit = np.log(amount) * 10 + np.random.normal(0, 5, 100)
r_p, p_p = stats.pearsonr(amount, credit)
r_s, p_s = stats.spearmanr(amount, credit)
print(f'Pearson  r={r_p:.4f}  p={p_p:.4f}')
print(f'Spearman ρ={r_s:.4f}  p={p_s:.4f}')
```
