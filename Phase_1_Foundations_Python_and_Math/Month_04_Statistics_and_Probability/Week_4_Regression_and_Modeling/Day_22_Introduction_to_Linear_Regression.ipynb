# Day 22: Introduction to Linear Regression

## Phase 1 | Month 4 | Week 4 | Day 22

**🎯 Goal:** Understand simple linear regression — fitting a line to data and interpreting coefficients.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
n = 150
# Kenya: household income (KES thousands) vs monthly expenditure
income = np.random.normal(65, 22, n).clip(20, 200)
noise  = np.random.normal(0, 5, n)
expenditure = 0.75 * income + 8 + noise

# Simple Linear Regression via scipy
slope, intercept, r, p, se = stats.linregress(income, expenditure)

print(f'Regression: Expenditure = {intercept:.2f} + {slope:.4f} × Income')
print(f'R² = {r**2:.4f}  (explains {r**2*100:.1f}% of variance)')
print(f'p-value for slope: {p:.6f}')
print(f'SE of slope: {se:.4f}')
print(f'\nInterpretation: For every KES 1k increase in income, expenditure rises by KES {slope*1000:.0f}')
```

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
n = 150
income = np.random.normal(65, 22, n).clip(20, 200)
expenditure = 0.75*income + 8 + np.random.normal(0,5,n)
slope, intercept, r, p, se = stats.linregress(income, expenditure)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Scatter + regression line
axes[0].scatter(income, expenditure, alpha=0.5, color='steelblue', s=30)
x_line = np.linspace(income.min(), income.max(), 100)
axes[0].plot(x_line, intercept+slope*x_line, 'tomato', lw=2.5,
             label=f'ŷ = {intercept:.1f} + {slope:.3f}x\nR²={r**2:.3f}')
axes[0].set_title('Income vs Expenditure (Kenya)', fontweight='bold')
axes[0].set_xlabel('Monthly Income (KES thousands)')
axes[0].set_ylabel('Monthly Expenditure (KES thousands)')
axes[0].legend()

# Residuals
fitted = intercept + slope*income
residuals = expenditure - fitted
axes[1].scatter(fitted, residuals, alpha=0.5, color='green', s=30)
axes[1].axhline(0, color='red', lw=2, ls='--')
axes[1].set_title('Residual Plot', fontweight='bold')
axes[1].set_xlabel('Fitted Values')
axes[1].set_ylabel('Residuals')

plt.tight_layout()
fig.savefig('/tmp/day22_regression.png', dpi=120)
print('Saved /tmp/day22_regression.png')
plt.close()
```

## Assumptions of Simple Linear Regression

1. **Linearity**: relationship is linear
2. **Independence**: observations are independent
3. **Homoscedasticity**: constant residual variance
4. **Normality**: residuals are approximately normal
