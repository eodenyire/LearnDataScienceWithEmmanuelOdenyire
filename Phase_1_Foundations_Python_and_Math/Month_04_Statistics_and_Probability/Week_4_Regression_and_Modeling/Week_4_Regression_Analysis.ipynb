# Week 4: Regression Analysis

## Learning Objectives
- Understand simple and multiple linear regression
- Interpret regression coefficients
- Evaluate regression models
- Diagnose regression assumptions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Simple Linear Regression

```python
n = 100
experience = rng.uniform(0, 15, n)
salary = 40000 + 4500 * experience + rng.normal(0, 6000, n)

df = pd.DataFrame({'experience': experience, 'salary': salary})
X = df[['experience']]
y = df['salary']

model = LinearRegression()
model.fit(X, y)
print(f'Intercept: ${model.intercept_:,.0f}')
print(f'Coefficient: ${model.coef_[0]:,.0f} per year of experience')
print(f'R²: {r2_score(y, model.predict(X)):.4f}')
```

```python
x_line = np.linspace(0, 15, 100).reshape(-1, 1)
y_line = model.predict(x_line)

plt.figure(figsize=(8, 5))
plt.scatter(experience, salary, alpha=0.5, label='Data')
plt.plot(x_line, y_line, color='red', linewidth=2, label='Regression line')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
plt.legend()
plt.tight_layout()
plt.show()
```

## 2. Multiple Linear Regression

```python
n = 200
exp = rng.uniform(0, 15, n)
edu = rng.integers(0, 4, n)  # 0=HS, 1=BS, 2=MS, 3=PhD
perf = rng.uniform(60, 100, n)
salary_m = 35000 + 4000*exp + 8000*edu + 200*perf + rng.normal(0, 5000, n)

df_m = pd.DataFrame({'experience': exp, 'education': edu, 'performance': perf, 'salary': salary_m})
X_m = df_m.drop('salary', axis=1)
y_m = df_m['salary']
X_train, X_test, y_train, y_test = train_test_split(X_m, y_m, test_size=0.2, random_state=42)

model_m = LinearRegression().fit(X_train, y_train)
y_pred = model_m.predict(X_test)
print('Multiple Regression Results:')
print(f'R² (test):  {r2_score(y_test, y_pred):.4f}')
print(f'RMSE:       ${np.sqrt(mean_squared_error(y_test, y_pred)):,.0f}')
for feat, coef in zip(X_m.columns, model_m.coef_):
    print(f'  {feat}: {coef:+.2f}')
```

## 3. Residual Analysis

```python
residuals = y_test.values - y_pred

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(y_pred, residuals, alpha=0.5)
axes[0].axhline(0, color='red', linewidth=1.5)
axes[0].set_title('Residuals vs Fitted')
axes[0].set_xlabel('Fitted Values'); axes[0].set_ylabel('Residuals')

stats.probplot(residuals, plot=axes[1])
axes[1].set_title('Q-Q Plot of Residuals')
plt.tight_layout(); plt.show()
```
