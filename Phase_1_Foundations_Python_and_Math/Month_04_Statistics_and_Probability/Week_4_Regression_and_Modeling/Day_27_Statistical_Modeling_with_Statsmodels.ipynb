```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 500
df = pd.DataFrame({
    'income_k':   np.random.normal(60, 25, N).clip(10, 300),
    'age':        np.random.normal(35, 10, N).clip(18, 70),
    'education':  np.random.choice(['Primary','Secondary','Tertiary'], N, p=[0.2,0.5,0.3]),
    'urban':      np.random.binomial(1, 0.65, N),
})
df['expenditure_k'] = (0.6 * df['income_k']
                       + 0.1 * df['age']
                       + 5   * (df['education']=='Tertiary').astype(int)
                       + 3   * df['urban']
                       + np.random.normal(0, 4, N))

# Formula API — treats categorical variables automatically
model = smf.ols('expenditure_k ~ income_k + age + C(education) + urban', data=df).fit()
print(model.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 500
df = pd.DataFrame({
    'income_k':  np.random.normal(60,25,N).clip(10,300),
    'age':       np.random.normal(35,10,N).clip(18,70),
    'education': np.random.choice(['Primary','Secondary','Tertiary'],N,p=[0.2,0.5,0.3]),
    'urban':     np.random.binomial(1,0.65,N),
})
df['expenditure_k'] = (0.6*df['income_k']+0.1*df['age']
                       +5*(df['education']=='Tertiary')+3*df['urban']
                       +np.random.normal(0,4,N))
model = smf.ols('expenditure_k ~ income_k + age + C(education) + urban', data=df).fit()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Coefficient plot
coef = model.params[1:]
err  = model.bse[1:]
axes[0].barh(coef.index, coef.values, xerr=err.values*1.96,
             color='steelblue', alpha=0.8, ecolor='black', capsize=4)
axes[0].axvline(0, color='red', lw=2, ls='--')
axes[0].set_title('Regression Coefficients (95% CI)', fontweight='bold')
axes[0].set_xlabel('Effect on Expenditure (KES thousands)')

# Actual vs Predicted
axes[1].scatter(df['expenditure_k'], model.fittedvalues, alpha=0.3, s=15, color='navy')
lims = [df['expenditure_k'].min(), df['expenditure_k'].max()]
axes[1].plot(lims, lims, 'r--', lw=2, label='Perfect prediction')
axes[1].set_title('Actual vs Predicted Expenditure', fontweight='bold')
axes[1].set_xlabel('Actual'); axes[1].set_ylabel('Predicted')
axes[1].legend()

plt.tight_layout()
fig.savefig('/tmp/day27_statsmodels.png', dpi=120)
print(f'Adj. R² = {model.rsquared_adj:.4f}')
plt.close()
```
