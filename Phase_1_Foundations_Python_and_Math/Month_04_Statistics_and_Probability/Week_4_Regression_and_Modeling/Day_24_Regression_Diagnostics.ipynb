# Day 24: Regression Diagnostics and Assumptions

## Phase 1 | Month 4 | Week 4 | Day 24

**🎯 Goal:** Validate regression assumptions using diagnostic plots and statistical tests.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
n = 200
x = np.random.uniform(10, 100, n)
y = 3 + 1.5*x + np.random.normal(0, 8, n)
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
residuals = model.resid
fitted    = model.fittedvalues

fig, axes = plt.subplots(2, 2, figsize=(13, 10))

# 1 – Residuals vs Fitted
axes[0,0].scatter(fitted, residuals, alpha=0.5, color='steelblue', s=20)
axes[0,0].axhline(0, color='red', lw=2)
axes[0,0].set_title('Residuals vs Fitted', fontweight='bold')
axes[0,0].set_xlabel('Fitted values'); axes[0,0].set_ylabel('Residuals')

# 2 – Q-Q plot
sm.qqplot(residuals, line='s', ax=axes[0,1], alpha=0.5)
axes[0,1].set_title('Normal Q-Q Plot', fontweight='bold')

# 3 – Scale-Location
axes[1,0].scatter(fitted, np.sqrt(np.abs(residuals)), alpha=0.5, color='green', s=20)
axes[1,0].set_title('Scale-Location (Homoscedasticity)', fontweight='bold')
axes[1,0].set_xlabel('Fitted values'); axes[1,0].set_ylabel('√|Residuals|')

# 4 – Histogram of residuals
axes[1,1].hist(residuals, bins=30, color='purple', alpha=0.7, edgecolor='white')
axes[1,1].set_title('Residual Distribution', fontweight='bold')

plt.tight_layout()
fig.savefig('/tmp/day24_diagnostics.png', dpi=120)
print('Saved /tmp/day24_diagnostics.png')

# Shapiro-Wilk test for normality of residuals
stat, p = stats.shapiro(residuals)
print(f'Shapiro-Wilk: W={stat:.4f}, p={p:.4f}')
plt.close()
```

## The Four Diagnostic Plots

| Plot | What It Shows | What You Want |
|------|--------------|---------------|
| Residuals vs Fitted | Linearity, homoscedasticity | Random scatter around 0 |
| Normal Q-Q | Normality of residuals | Points on diagonal |
| Scale-Location | Homoscedasticity | Flat horizontal band |
| Residuals vs Leverage | Influential observations | No points beyond Cook's distance |
