```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
N = 600

# Predicting loan default for KCB customers
age        = np.random.normal(35, 10, N).clip(18, 70)
income_k   = np.random.normal(60, 25, N).clip(10, 300)
debt_ratio = np.random.uniform(0.1, 0.8, N)
credit_scr = np.random.normal(650, 80, N).clip(300, 850)

logit  = -8 + 0.02*age - 0.01*income_k + 3*debt_ratio - 0.005*credit_scr
prob   = 1 / (1 + np.exp(-logit))
default = (np.random.rand(N) < prob).astype(int)

df = pd.DataFrame({'default':default,'age':age,'income_k':income_k,
                   'debt_ratio':debt_ratio,'credit_scr':credit_scr})

X = sm.add_constant(df[['age','income_k','debt_ratio','credit_scr']])
model = sm.Logit(df['default'], X).fit()
print(model.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve

np.random.seed(42)
N = 600
age = np.random.normal(35,10,N).clip(18,70)
income_k = np.random.normal(60,25,N).clip(10,300)
debt_ratio = np.random.uniform(0.1,0.8,N)
credit_scr = np.random.normal(650,80,N).clip(300,850)
logit = -8+0.02*age-0.01*income_k+3*debt_ratio-0.005*credit_scr
prob = 1/(1+np.exp(-logit))
default = (np.random.rand(N)<prob).astype(int)
df = pd.DataFrame({'default':default,'age':age,'income_k':income_k,'debt_ratio':debt_ratio,'credit_scr':credit_scr})
X = sm.add_constant(df[['age','income_k','debt_ratio','credit_scr']])
model = sm.Logit(df['default'], X).fit(disp=0)
y_pred_prob = model.predict(X)
y_pred = (y_pred_prob >= 0.5).astype(int)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# S-curve (logistic function)
x_cs = np.linspace(300, 850, 300)
logit_cs = -8+0.02*35-0.01*60+3*0.35-0.005*x_cs
p_cs = 1/(1+np.exp(-logit_cs))
axes[0].plot(x_cs, p_cs, 'steelblue', lw=2.5)
axes[0].set_title('P(Default) vs Credit Score', fontweight='bold')
axes[0].set_xlabel('Credit Score'); axes[0].set_ylabel('P(Default)')
axes[0].axhline(0.5, color='red', ls='--')

# ROC curve
fpr, tpr, _ = roc_curve(default, y_pred_prob)
auc = roc_auc_score(default, y_pred_prob)
axes[1].plot(fpr, tpr, 'navy', lw=2, label=f'AUC={auc:.3f}')
axes[1].plot([0,1],[0,1],'k--')
axes[1].set_title('ROC Curve', fontweight='bold')
axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

plt.tight_layout()
fig.savefig('/tmp/day25_logistic.png', dpi=120)
print(f'AUC: {auc:.4f}')
plt.close()
```
