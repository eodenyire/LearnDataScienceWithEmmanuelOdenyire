# Day 13: Sampling Distributions and the Central Limit Theorem

## Phase 1 | Month 4 | Week 2 | Day 13

**🎯 Goal:** Understand sampling variability and the Central Limit Theorem — the most important theorem in statistics.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Demonstrate CLT: population is highly skewed (M-Pesa transactions)
np.random.seed(42)
population = np.random.exponential(scale=5000, size=100_000)  # right-skewed

sample_sizes = [5, 30, 100]
n_samples = 2000

fig, axes = plt.subplots(1, 4, figsize=(18, 4))

# Population
axes[0].hist(population, bins=80, color='orange', edgecolor='white', alpha=0.7)
axes[0].set_title('Population\n(Exponential, skewed)', fontweight='bold')
axes[0].set_xlabel('Transaction (KES)')

for ax, n in zip(axes[1:], sample_sizes):
    sample_means = [np.random.choice(population, n, replace=False).mean() for _ in range(n_samples)]
    ax.hist(sample_means, bins=50, color='steelblue', edgecolor='white', alpha=0.7)
    ax.set_title(f'Sample Means (n={n})\nApprox Normal!', fontweight='bold')
    ax.set_xlabel('Sample Mean (KES)')

plt.tight_layout()
fig.savefig('/tmp/day13_clt.png', dpi=120)
print('Saved /tmp/day13_clt.png')
plt.close()
```

## The Central Limit Theorem

**Statement:** For a sufficiently large sample size (n ≥ 30), the sampling distribution of the sample mean is approximately Normal, regardless of the population's shape:

$$\bar{X} \sim N\left(\mu, \frac{\sigma^2}{n}\right)$$

**Standard Error (SE):** $SE = \sigma / \sqrt{n}$ — measures variability of the sample mean.

```python
import numpy as np
from scipy import stats

# Practical implication: KNBS estimates national mean income
# Population: μ=45000, σ=18000 (monthly income KES)
mu, sigma = 45_000, 18_000

for n in [10, 30, 100, 500, 1000]:
    se = sigma / np.sqrt(n)
    margin_of_error_95 = 1.96 * se
    print(f'n={n:4d}  SE=KES {se:6.0f}  95% MoE=±KES {margin_of_error_95:6.0f}')
```
