```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Simulate rolling a die 10,000 times (like Nairobi board game café data)
np.random.seed(42)
rolls = np.random.randint(1, 7, size=10_000)

for face in range(1, 7):
    empirical = (rolls == face).mean()
    print(f'P(face={face}): theoretical=1/6={1/6:.4f}  empirical={empirical:.4f}')
```

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Law of Large Numbers: empirical probability converges to theoretical
np.random.seed(0)
tosses = np.random.randint(0, 2, size=10_000)   # 0=Tails, 1=Heads
cumulative_prob = np.cumsum(tosses) / np.arange(1, len(tosses)+1)

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(cumulative_prob, color='steelblue', lw=1.5, label='Empirical P(Heads)')
ax.axhline(0.5, color='red', lw=2, ls='--', label='Theoretical P=0.5')
ax.set_xscale('log')
ax.set_title('Law of Large Numbers: Coin Flip', fontweight='bold')
ax.set_xlabel('Number of Tosses (log scale)')
ax.set_ylabel('P(Heads)')
ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day08_prob.png', dpi=120)
print('Saved /tmp/day08_prob.png')
plt.close()
```

## Probability Rules

| Rule | Formula | Example |
|------|---------|--------|
| Complement | P(Aᶜ) = 1 - P(A) | P(not rain) = 1 - P(rain) |
| Addition | P(A∪B) = P(A)+P(B)-P(A∩B) | P(rain or traffic) |
| Multiplication (independent) | P(A∩B) = P(A)×P(B) | P(heads AND six) |
| Multiplication (dependent) | P(A∩B) = P(A)×P(B|A) | Cards without replacement |

```python
# Real-world example: Safaricom SIM card quality control
# P(defective) = 0.02, P(software_bug) = 0.03, independent events
P_defective = 0.02
P_software  = 0.03

P_both   = P_defective * P_software
P_either = P_defective + P_software - P_both
P_neither = (1 - P_defective) * (1 - P_software)

print(f'P(defective AND software bug) = {P_both:.6f}')
print(f'P(defective OR  software bug) = {P_either:.6f}')
print(f'P(neither fault)              = {P_neither:.6f}')
```
