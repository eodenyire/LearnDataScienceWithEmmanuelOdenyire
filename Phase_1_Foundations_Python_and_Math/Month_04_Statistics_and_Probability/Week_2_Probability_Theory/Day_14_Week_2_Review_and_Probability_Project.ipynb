## 📊 Mini-Project: Safaricom Customer Churn Model

You will build a probability-based framework for predicting mobile subscriber churn.

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(2024)
N = 5000

# Simulate subscriber features
df = pd.DataFrame({
    'subscriber_id':     range(1, N+1),
    'monthly_spend_kes': np.random.lognormal(8.5, 0.8, N).round(0),
    'calls_per_month':   np.random.poisson(45, N),
    'data_gb_per_month': np.random.gamma(3, 2, N).round(2),
    'months_on_network': np.random.randint(1, 120, N),
    'complaints_90d':    np.random.poisson(0.3, N),
})
# Churn probability increases with complaints, decreases with tenure
logit = (-0.5
         + 0.8 * df['complaints_90d']
         - 0.01 * df['months_on_network']
         - 0.0001 * df['monthly_spend_kes'])
df['churn_prob'] = 1 / (1 + np.exp(-logit))
df['churned']    = (np.random.rand(N) < df['churn_prob']).astype(int)

print(df.head())
print(f'\nChurn rate: {df.churned.mean():.3f} ({df.churned.mean()*100:.1f}%)')
```

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(2024)
N = 5000
df = pd.DataFrame({
    'monthly_spend_kes': np.random.lognormal(8.5, 0.8, N).round(0),
    'calls_per_month':   np.random.poisson(45, N),
    'months_on_network': np.random.randint(1, 120, N),
    'complaints_90d':    np.random.poisson(0.3, N),
})
logit = (-0.5 + 0.8*df['complaints_90d'] - 0.01*df['months_on_network'] - 0.0001*df['monthly_spend_kes'])
df['churn_prob'] = 1/(1+np.exp(-logit))
df['churned'] = (np.random.rand(N) < df['churn_prob']).astype(int)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1 – Spend distribution by churn status
for status, color in [(0,'steelblue'),(1,'tomato')]:
    d = df[df.churned==status]['monthly_spend_kes']
    axes[0,0].hist(d, bins=50, alpha=0.6, color=color,
                   label=f'{"Churned" if status else "Retained"} (n={len(d)})')
axes[0,0].set_title('Monthly Spend by Churn Status', fontweight='bold')
axes[0,0].legend()
axes[0,0].set_xlabel('Spend (KES)')

# 2 – Churn rate by complaint bucket
df['complaint_bucket'] = pd.cut(df['complaints_90d'], bins=[-1,0,1,2,10], labels=['0','1','2','3+'])
churn_by_complaint = df.groupby('complaint_bucket')['churned'].mean()
axes[0,1].bar(churn_by_complaint.index, churn_by_complaint.values, color='tomato')
axes[0,1].set_title('Churn Rate by Complaints', fontweight='bold')
axes[0,1].set_ylabel('Churn Rate')

# 3 – Tenure distribution
axes[1,0].hist(df['months_on_network'], bins=30, color='green', alpha=0.7)
axes[1,0].set_title('Subscriber Tenure Distribution', fontweight='bold')
axes[1,0].set_xlabel('Months on Network')

# 4 – Churn probability distribution
axes[1,1].hist(df['churn_prob'], bins=50, color='purple', alpha=0.7)
axes[1,1].axvline(0.5, color='red', lw=2, ls='--', label='Decision boundary')
axes[1,1].set_title('Predicted Churn Probability', fontweight='bold')
axes[1,1].set_xlabel('P(churn)')
axes[1,1].legend()

plt.tight_layout()
fig.savefig('/tmp/day14_project.png', dpi=120)
print('Dashboard saved')
plt.close()
```
