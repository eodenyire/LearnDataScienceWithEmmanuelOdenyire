```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Discrete RV: number of M-Pesa transactions per customer per day
# Modelled as Poisson(λ=3)
lambda_ = 3
x = np.arange(0, 15)
pmf = stats.poisson.pmf(x, lambda_)
cdf = stats.poisson.cdf(x, lambda_)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(x, pmf, color='steelblue', edgecolor='white')
axes[0].set_title('PMF: Poisson(λ=3) — Daily Transactions', fontweight='bold')
axes[0].set_xlabel('Number of transactions')
axes[0].set_ylabel('P(X = x)')

axes[1].step(x, cdf, where='post', color='tomato', lw=2)
axes[1].set_title('CDF: Poisson(λ=3)', fontweight='bold')
axes[1].set_xlabel('Number of transactions')
axes[1].set_ylabel('P(X ≤ x)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
fig.savefig('/tmp/day10_rv.png', dpi=120)
print('Saved /tmp/day10_rv.png')
plt.close()
print(f'P(X ≤ 5) = {stats.poisson.cdf(5, lambda_):.4f}')
print(f'P(X > 5) = {1 - stats.poisson.cdf(5, lambda_):.4f}')
```

## Expected Value and Variance of a Random Variable

For a discrete RV: E[X] = Σ x·P(X=x)

For a continuous RV: E[X] = ∫ x·f(x) dx

**Key insight:** E[X] is the long-run average, not necessarily an achievable value.

```python
import numpy as np
from scipy import stats

# Continuous RV: transaction amount (KES) follows LogNormal
mu, sigma = 7.5, 1.2   # log-scale parameters
dist = stats.lognorm(s=sigma, scale=np.exp(mu))

print('LogNormal distribution: M-Pesa transaction amounts')
print(f'  Mean (E[X]):   KES {dist.mean():,.0f}')
print(f'  Median:        KES {dist.median():,.0f}')
print(f'  Std Dev:       KES {dist.std():,.0f}')
print(f'  P(X ≤ 1000):  {dist.cdf(1000):.4f}')
print(f'  P(X > 10000): {1-dist.cdf(10000):.4f}')
print(f'  90th percentile: KES {dist.ppf(0.90):,.0f}')
```
