# Day 9: Conditional Probability and Bayes' Theorem

## Phase 1 | Month 4 | Week 2 | Day 9

**🎯 Goal:** Understand conditional probability and apply Bayes' theorem to real-world classification problems.
**⏱️ Study Session:** ~2 hours

```python
# Conditional probability: P(A|B) = P(A∩B) / P(B)
# Medical test example — relevant to Kenya's healthcare data analytics

# COVID rapid test performance
P_covid_positive  = 0.10   # prevalence (10% of population has it)
P_positive_given_covid    = 0.95   # sensitivity
P_positive_given_no_covid = 0.05   # 1 - specificity (false positive rate)

P_no_covid = 1 - P_covid_positive

# P(positive test) using Law of Total Probability
P_positive_test = (P_positive_given_covid * P_covid_positive +
                   P_positive_given_no_covid * P_no_covid)

# Bayes' Theorem: P(covid | positive test)
P_covid_given_positive = (P_positive_given_covid * P_covid_positive) / P_positive_test

print('=== Bayes Theorem: COVID Test ===')
print(f'Prevalence (P(COVID)):         {P_covid_positive:.2f}')
print(f'Sensitivity (P(+|COVID)):      {P_positive_given_covid:.2f}')
print(f'False Positive Rate:           {P_positive_given_no_covid:.2f}')
print(f'P(positive test):              {P_positive_test:.4f}')
print(f'P(COVID | positive test):      {P_covid_given_positive:.4f}')
print(f'\nOnly {P_covid_given_positive*100:.1f}% chance you actually have COVID given a positive test!')
print('(This is why low-prevalence screening is tricky)')
```

## Visualising Bayes' Theorem: Frequency Grid

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Frequency visualisation with 1000 hypothetical people
N = 1000
covid     = int(N * 0.10)
no_covid  = N - covid

true_pos  = int(covid * 0.95)
false_neg = covid - true_pos
false_pos = int(no_covid * 0.05)
true_neg  = no_covid - false_pos

fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')
table_data = [
    ['', 'Test +', 'Test -', 'Total'],
    ['Has COVID',    true_pos,  false_neg, covid],
    ['No COVID',     false_pos, true_neg,  no_covid],
    ['Total',        true_pos+false_pos, false_neg+true_neg, N]
]
table = ax.table(cellText=table_data, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(13)
table.scale(1.5, 2)
ax.set_title('Bayes Theorem: Frequency Grid (N=1000 people)', fontweight='bold', pad=20)
plt.tight_layout()
fig.savefig('/tmp/day09_bayes.png', dpi=120)
print('Saved /tmp/day09_bayes.png')
plt.close()
print(f'True Positive Rate (PPV): {true_pos/(true_pos+false_pos):.3f}')
```

## M-Pesa Fraud Detection Example

Bayes' theorem is used in M-Pesa fraud detection at Safaricom:

```python
# P(fraud) = 0.001 (0.1% of transactions are fraudulent)
# P(flagged | fraud) = 0.90   (90% of fraud is caught)
# P(flagged | legit) = 0.001  (0.1% false alarm rate)

P_fraud  = 0.001
P_legit  = 1 - P_fraud
P_flag_fraud = 0.90
P_flag_legit = 0.001

P_flag = P_flag_fraud*P_fraud + P_flag_legit*P_legit
P_fraud_given_flag = P_flag_fraud*P_fraud / P_flag

print(f'P(fraud | flagged transaction): {P_fraud_given_flag:.4f}')
print(f'That is {P_fraud_given_flag*100:.1f}% — most flagged transactions are NOT fraud (false alarms)!')
print(f'This is why fraud systems need human review.')
```
