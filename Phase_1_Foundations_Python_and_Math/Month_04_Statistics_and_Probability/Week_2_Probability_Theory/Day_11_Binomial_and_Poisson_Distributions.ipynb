# Day 11: Binomial and Poisson Distributions

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# BINOMIAL: n independent trials, each with P(success)=p
# Example: Out of 50 loan applications, each has P(approved)=0.70
n, p = 50, 0.70
x = np.arange(0, n+1)
bin_pmf = stats.binom.pmf(x, n, p)

print('=== Binomial(n=50, p=0.70) — Loan Approvals ===')
print(f'E[X] = np = {n*p:.1f}')
print(f'Std  = sqrt(np(1-p)) = {np.sqrt(n*p*(1-p)):.2f}')
print(f'P(exactly 35 approved) = {stats.binom.pmf(35, n, p):.4f}')
print(f'P(at least 40 approved) = {1-stats.binom.cdf(39, n, p):.4f}')
```

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# POISSON: events in fixed time/space interval
# Example: M-Pesa system errors per hour (λ=2)
lam = 2
x_pois = np.arange(0, 12)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Binomial
n, p = 50, 0.70
x_bin = np.arange(0, 51)
axes[0].bar(x_bin, stats.binom.pmf(x_bin, n, p), color='steelblue', alpha=0.8, edgecolor='white')
axes[0].axvline(n*p, color='red', lw=2, ls='--', label=f'μ={n*p}')
axes[0].set_title('Binomial(n=50, p=0.7)\nKCB Loan Approvals', fontweight='bold')
axes[0].set_xlabel('Number of Approvals')
axes[0].legend()

# Poisson
axes[1].bar(x_pois, stats.poisson.pmf(x_pois, lam), color='tomato', alpha=0.8, edgecolor='white')
axes[1].axvline(lam, color='navy', lw=2, ls='--', label=f'λ={lam}')
axes[1].set_title('Poisson(λ=2)\nM-Pesa System Errors/Hour', fontweight='bold')
axes[1].set_xlabel('Number of Errors')
axes[1].legend()

plt.tight_layout()
fig.savefig('/tmp/day11_discrete.png', dpi=120)
print('Saved /tmp/day11_discrete.png')
plt.close()
```

## Poisson Approximation to Binomial

When n is large and p is small, Binomial(n,p) ≈ Poisson(λ=np).

```python
import numpy as np
from scipy import stats

# SIM card defect example: n=10000, p=0.0003 → λ=3
n, p = 10_000, 0.0003
lam = n * p

print('Binomial vs Poisson approximation:')
for k in range(8):
    binom_p = stats.binom.pmf(k, n, p)
    pois_p  = stats.poisson.pmf(k, lam)
    print(f'  P(X={k}): Binom={binom_p:.6f}  Poisson={pois_p:.6f}  diff={abs(binom_p-pois_p):.2e}')
```
