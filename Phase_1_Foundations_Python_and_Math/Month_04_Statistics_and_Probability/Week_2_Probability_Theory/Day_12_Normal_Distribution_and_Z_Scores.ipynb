# Day 12: Normal Distribution and Z-Scores

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Normal distribution properties
mu, sigma = 52, 12   # KCSE mean and std dev
dist = stats.norm(mu, sigma)
x = np.linspace(mu-4*sigma, mu+4*sigma, 400)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, dist.pdf(x), 'navy', lw=2.5, label='N(52, 12)')

# Shade regions
for lo, hi, color, label in [
        (mu-sigma, mu+sigma, 'steelblue', '68.3% within ±1σ'),
        (mu-2*sigma, mu+2*sigma, 'lightblue', '95.4% within ±2σ'),
]:
    ax.fill_between(x, dist.pdf(x), where=((x>=lo)&(x<=hi)), alpha=0.3, color=color, label=label)

ax.axvline(mu, color='red', lw=2, ls='--', label=f'μ={mu}')
ax.set_title('KCSE Score Distribution — Normal Curve', fontweight='bold')
ax.set_xlabel('Score')
ax.set_ylabel('Probability Density')
ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day12_normal.png', dpi=120)
print('Saved /tmp/day12_normal.png')
plt.close()
print(f'P(40 < X < 64): {dist.cdf(64)-dist.cdf(40):.4f}')
```

```python
import numpy as np
from scipy import stats

# Z-scores: standardise to N(0,1)
mu, sigma = 52, 12
scores = np.array([30, 45, 52, 65, 78])
z_scores = (scores - mu) / sigma

print('Score → Z-score → Percentile')
for s, z in zip(scores, z_scores):
    pct = stats.norm.cdf(z) * 100
    print(f'  {s:3d}  → z={z:+.2f}  → {pct:.1f}th percentile')
```

## 68-95-99.7 Rule (Empirical Rule)

| Range | Probability |
|-------|------------|
| μ ± 1σ | 68.3% |
| μ ± 2σ | 95.4% |
| μ ± 3σ | 99.7% |

This is used in statistical process control (SPC) at Kenyan manufacturing plants.

```python
import numpy as np
from scipy import stats

# Real example: M-Pesa response time (ms) N(μ=120, σ=20)
mu, sigma = 120, 20
print('M-Pesa API Response Time SLA Analysis:')
print(f'P(response < 150ms)  : {stats.norm.cdf(150, mu, sigma):.4f}')
print(f'P(response > 180ms)  : {1-stats.norm.cdf(180, mu, sigma):.6f}')
print(f'99th percentile      : {stats.norm.ppf(0.99, mu, sigma):.1f}ms')
print(f'SLA 99.9% at         : {stats.norm.ppf(0.999, mu, sigma):.1f}ms')
```
