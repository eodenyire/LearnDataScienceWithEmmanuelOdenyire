# Day 4: Data Distributions and Frequency Tables

## Phase 1 | Month 4 | Week 1 | Day 4

**🎯 Goal:** Build frequency tables, histograms, and understand the shapes of distributions.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Simulated KCSE scores for 200 students in Nairobi
np.random.seed(0)
scores = np.random.normal(loc=52, scale=15, size=200).clip(0, 100).astype(int)

# Frequency table
bins   = [0, 20, 40, 60, 80, 100]
labels = ['E (0-20)', 'D (21-40)', 'C (41-60)', 'B (61-80)', 'A (81-100)']
df = pd.DataFrame({'Score': scores})
df['Grade'] = pd.cut(df['Score'], bins=bins, labels=labels, include_lowest=True)

freq = df['Grade'].value_counts().sort_index()
freq_df = pd.DataFrame({
    'Frequency': freq,
    'Relative Freq': (freq / freq.sum()).round(3),
    'Cumulative Freq': (freq / freq.sum()).cumsum().round(3)
})
print(freq_df)
```

```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(0)
scores = np.random.normal(loc=52, scale=15, size=200).clip(0, 100).astype(int)
bins   = [0, 20, 40, 60, 80, 100]
labels = ['E', 'D', 'C', 'B', 'A']
df = pd.DataFrame({'Score': scores})
df['Grade'] = pd.cut(df['Score'], bins=bins, labels=labels, include_lowest=True)
freq = df['Grade'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(freq.index, freq.values, color='steelblue', edgecolor='white')
axes[0].set_title('KCSE Grade Frequency', fontweight='bold')
axes[0].set_xlabel('Grade')
axes[0].set_ylabel('Number of Students')

axes[1].hist(scores, bins=20, cumulative=True, color='tomato', alpha=0.7, edgecolor='white')
axes[1].set_title('Cumulative Frequency (Ogive)', fontweight='bold')
axes[1].set_xlabel('Score')
axes[1].set_ylabel('Cumulative Count')

plt.tight_layout()
fig.savefig('/tmp/day04_freq.png', dpi=120)
print('Saved /tmp/day04_freq.png')
plt.close()
```

## Distribution Shapes

| Shape | Mean vs Median | Example |
|-------|---------------|--------|
| Symmetric (Normal) | Mean ≈ Median | Test scores |
| Right-skewed (Positive) | Mean > Median | Income, house prices |
| Left-skewed (Negative) | Mean < Median | Age at retirement |
| Bimodal | Two peaks | Commute times (AM/PM rush) |
