# Day 3: Measures of Variability

## Phase 1 | Month 4 | Week 1 | Day 3

**🎯 Goal:** Quantify how spread out data is using range, variance, standard deviation, and IQR.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
# Daily returns (%) of Safaricom stock vs a stable bond
safaricom = np.random.normal(0.05, 1.8, 252)   # volatile
bond      = np.random.normal(0.02, 0.3, 252)   # stable

for name, data in [('Safaricom', safaricom), ('Bond', bond)]:
    print(f'\n=== {name} ===')
    print(f'  Range  : {data.max()-data.min():.4f}%')
    print(f'  Variance (s²): {np.var(data, ddof=1):.4f}')
    print(f'  Std Dev (s)  : {np.std(data, ddof=1):.4f}%')
    q1, q3 = np.percentile(data, [25, 75])
    print(f'  IQR          : {q3-q1:.4f}%')
    print(f'  CV           : {np.std(data,ddof=1)/abs(np.mean(data))*100:.1f}%')
```

## Why Variability Matters

- **Standard Deviation** is the most common risk measure in finance
- **IQR** is robust to outliers — used in box plots
- **Coefficient of Variation (CV)** compares variability across different scales

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
safaricom = np.random.normal(0.05, 1.8, 252)
bond = np.random.normal(0.02, 0.3, 252)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, data, name, color in zip(axes, [safaricom, bond], ['Safaricom Returns', 'Bond Returns'], ['tomato', 'steelblue']):
    ax.hist(data, bins=40, color=color, alpha=0.7, edgecolor='white')
    ax.axvline(data.mean(), color='black', lw=2, label=f'Mean: {data.mean():.3f}%')
    ax.axvline(data.mean()+data.std(ddof=1), color='navy', lw=1.5, ls='--', label=f'±1σ: {data.std(ddof=1):.3f}%')
    ax.axvline(data.mean()-data.std(ddof=1), color='navy', lw=1.5, ls='--')
    ax.set_title(name, fontweight='bold')
    ax.set_xlabel('Daily Return (%)')
    ax.legend(fontsize=9)
plt.tight_layout()
fig.savefig('/tmp/day03_variability.png', dpi=120)
print('Saved /tmp/day03_variability.png')
plt.close()
```
