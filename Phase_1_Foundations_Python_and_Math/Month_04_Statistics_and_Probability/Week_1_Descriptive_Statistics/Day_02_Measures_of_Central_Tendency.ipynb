3. Mode for categorical and multimodal data

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Nairobi employee salaries (KES thousands/month) — with CEO outlier
salaries = np.array([25, 28, 32, 35, 40, 42, 45, 50, 55, 60, 65, 500])

print('=== Measures of Central Tendency ===')
print(f'Mean   : KES {salaries.mean():.1f}k  ← pulled up by 500k CEO')
print(f'Median : KES {np.median(salaries):.1f}k  ← not affected by outlier')
print(f'Mode   : KES {stats.mode(salaries, keepdims=True).mode[0]:.1f}k')
print(f'\nMean without outlier: KES {salaries[:-1].mean():.1f}k')
```

## When to Use Each Measure

| Measure | Best When | Avoid When |
|---------|-----------|------------|
| **Mean** | Symmetric data, no outliers | Skewed data, outliers present |
| **Median** | Skewed data (income, house prices) | Need algebraic properties |
| **Mode** | Categorical data, most-common value | Continuous data (rarely useful) |

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)
# Simulate Nairobi house prices (KES millions)
prices = np.concatenate([np.random.normal(8, 2, 500), np.random.normal(50, 5, 20)])

fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(prices, bins=60, color='steelblue', edgecolor='white', alpha=0.7)
ax.axvline(prices.mean(),   color='red',    lw=2, label=f'Mean:   KES {prices.mean():.1f}M')
ax.axvline(np.median(prices), color='green', lw=2, label=f'Median: KES {np.median(prices):.1f}M')
ax.set_title('Nairobi House Prices — Mean vs Median', fontweight='bold')
ax.set_xlabel('Price (KES Millions)')
ax.set_ylabel('Frequency')
ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day02_central.png', dpi=120)
print('Saved /tmp/day02_central.png')
plt.close()
```

## Weighted Mean

Used when observations have different importance (e.g., student grades with different credit hours).

```python
import numpy as np

# University of Nairobi student grades
courses      = ['Math', 'Stats', 'Python', 'Databases', 'Ethics']
grades       = np.array([78, 85, 92, 70, 88])   # marks out of 100
credit_hours = np.array([4,  3,   3,  2,  1])    # credit weighting

simple_mean   = grades.mean()
weighted_mean = np.average(grades, weights=credit_hours)

print(f'Simple mean  : {simple_mean:.2f}')
print(f'Weighted mean: {weighted_mean:.2f}')
print(f'The weighted mean rewards performance in heavier courses (Math, Stats).')
```
