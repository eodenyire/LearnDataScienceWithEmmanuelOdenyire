# Day 5: Percentiles, Quartiles, and Box Plots

## Phase 1 | Month 4 | Week 1 | Day 5

**🎯 Goal:** Understand quartiles, percentiles, the five-number summary, and build box plots.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(1)
# Monthly salaries of software engineers in Nairobi (KES thousands)
salaries = np.concatenate([np.random.normal(120, 30, 200), [500, 600, 750]])

q1, q2, q3 = np.percentile(salaries, [25, 50, 75])
iqr = q3 - q1

print('Five-Number Summary:')
print(f'  Min    : KES {salaries.min():.1f}k')
print(f'  Q1     : KES {q1:.1f}k')
print(f'  Median : KES {q2:.1f}k')
print(f'  Q3     : KES {q3:.1f}k')
print(f'  Max    : KES {salaries.max():.1f}k')
print(f'  IQR    : KES {iqr:.1f}k')
print(f'  Lower fence: KES {q1-1.5*iqr:.1f}k')
print(f'  Upper fence: KES {q3+1.5*iqr:.1f}k')
print(f'  Outliers > upper fence: {(salaries > q3+1.5*iqr).sum()} values')
```

```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(1)
nai = np.concatenate([np.random.normal(120, 30, 200), [500, 600, 750]])
msa = np.concatenate([np.random.normal(90, 25, 200), [400]])
kis = np.random.normal(75, 20, 200)

fig, ax = plt.subplots(figsize=(9, 5))
bp = ax.boxplot([nai, msa, kis], labels=['Nairobi', 'Mombasa', 'Kisumu'],
                patch_artist=True, notch=True)
colors = ['steelblue', 'tomato', 'green']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
ax.set_title('Software Engineer Salaries by City — Kenya', fontweight='bold')
ax.set_ylabel('Monthly Salary (KES thousands)')
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
fig.savefig('/tmp/day05_boxplot.png', dpi=120)
print('Saved /tmp/day05_boxplot.png')
plt.close()
```

## Percentiles in Practice

- **P90** (90th percentile) — 90% of values fall below this
- KNBS uses percentiles to report income inequality
- Exam performance: 'Scored in the 85th percentile' means outperformed 85% of students

```python
import numpy as np

np.random.seed(0)
# KCPE scores across Kenya
scores = np.random.normal(250, 60, 10000).clip(0, 500).astype(int)

for p in [10, 25, 50, 75, 90, 95, 99]:
    print(f'  P{p:2d}: {np.percentile(scores, p):.0f} marks')
```
