```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(2024)
n = 2000
df = pd.DataFrame({
    'transaction_id': range(1, n+1),
    'amount_kes': np.random.lognormal(mean=7.5, sigma=1.2, size=n).round(2),
    'county': np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n, p=[0.35,0.20,0.15,0.15,0.15]),
    'channel': np.random.choice(['Send Money','Buy Goods','Pay Bill','Withdraw','Deposit'], n, p=[0.30,0.25,0.20,0.15,0.10]),
    'hour': np.random.randint(0, 24, n),
})

# Complete descriptive summary
summary = df['amount_kes'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9, 0.99])
summary['skewness']      = stats.skew(df['amount_kes'])
summary['kurtosis']      = stats.kurtosis(df['amount_kes'])
summary['IQR']           = df['amount_kes'].quantile(0.75) - df['amount_kes'].quantile(0.25)
summary['CV (%)']        = df['amount_kes'].std(ddof=1) / df['amount_kes'].mean() * 100
print(summary.round(2))
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(2024)
n = 2000
df = pd.DataFrame({
    'amount_kes': np.random.lognormal(mean=7.5, sigma=1.2, size=n).round(2),
    'county': np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n, p=[0.35,0.20,0.15,0.15,0.15]),
    'channel': np.random.choice(['Send Money','Buy Goods','Pay Bill','Withdraw','Deposit'], n, p=[0.30,0.25,0.20,0.15,0.10]),
    'hour': np.random.randint(0, 24, n),
})

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1 – Amount distribution
axes[0,0].hist(df['amount_kes'], bins=50, color='steelblue', edgecolor='white', alpha=0.8)
axes[0,0].axvline(df['amount_kes'].median(), color='red', lw=2, label=f'Median: KES {df.amount_kes.median():,.0f}')
axes[0,0].set_title('Transaction Amount Distribution', fontweight='bold')
axes[0,0].set_xlabel('Amount (KES)')
axes[0,0].legend()

# 2 – Box plots by county
counties = ['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret']
data_by_county = [df[df['county']==c]['amount_kes'].values for c in counties]
bp = axes[0,1].boxplot(data_by_county, labels=counties, patch_artist=True)
for p in bp['boxes']: p.set_facecolor('lightblue')
axes[0,1].set_title('Transaction Amount by County', fontweight='bold')
axes[0,1].set_ylabel('Amount (KES)')

# 3 – Channel frequency
channel_counts = df['channel'].value_counts()
axes[1,0].bar(channel_counts.index, channel_counts.values, color='tomato', edgecolor='white')
axes[1,0].set_title('Transactions by Channel', fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=30)

# 4 – Hourly pattern
hourly = df.groupby('hour')['amount_kes'].count()
axes[1,1].bar(hourly.index, hourly.values, color='green', alpha=0.7)
axes[1,1].set_title('Transaction Volume by Hour', fontweight='bold')
axes[1,1].set_xlabel('Hour of Day')

plt.tight_layout()
fig.savefig('/tmp/day07_project.png', dpi=120)
print('Dashboard saved to /tmp/day07_project.png')
plt.close()
```
