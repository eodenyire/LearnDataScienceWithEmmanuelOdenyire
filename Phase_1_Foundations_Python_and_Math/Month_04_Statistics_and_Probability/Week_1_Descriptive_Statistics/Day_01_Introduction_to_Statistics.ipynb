## Population vs Sample

| Concept | Symbol | Description |
|---------|--------|-------------|
| Population mean | μ | Mean of ALL observations |
| Sample mean | x̄ | Mean of a SUBSET |
| Population std dev | σ | Std of ALL observations |
| Sample std dev | s | Std of a SUBSET (ddof=1) |

**Rule of thumb:** Use `ddof=1` (Bessel's correction) whenever working with a sample.

```python
import numpy as np

np.random.seed(42)
# Simulate M-Pesa daily transactions (KES billions) for a year
population = np.random.normal(loc=12.5, scale=2.3, size=365)

# Draw a sample of 30 days
sample = np.random.choice(population, size=30, replace=False)

print('=== Population ===')
print(f'  Mean (μ): {population.mean():.4f}')
print(f'  Std  (σ): {population.std(ddof=0):.4f}')

print('\n=== Sample (n=30) ===')
print(f'  Mean (x̄): {sample.mean():.4f}')
print(f'  Std  (s): {sample.std(ddof=1):.4f}')
```

## 🌍 Real-World Example: KNBS Household Survey

KNBS cannot survey every Kenyan household (47 million people). Instead they sample ~60,000 households and use those statistics to estimate population parameters.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(0)
# Simulated monthly household income (KES) for Nairobi
population_income = np.random.lognormal(mean=10.5, sigma=0.6, size=100_000)

# KNBS-style sample
sample_income = np.random.choice(population_income, size=600, replace=False)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, data, title in zip(axes, [population_income, sample_income],
                            ['Population (100,000)', 'KNBS Sample (600)']):
    ax.hist(data, bins=50, color='steelblue', edgecolor='white', alpha=0.8)
    ax.axvline(data.mean(), color='red', linestyle='--', label=f'Mean: KES {data.mean():,.0f}')
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Monthly Income (KES)')
    ax.set_ylabel('Frequency')
    ax.legend()
plt.tight_layout()
fig.savefig('/tmp/day01_income.png', dpi=120)
print('Saved /tmp/day01_income.png')
print(f'Population mean: KES {population_income.mean():,.0f}')
print(f'Sample mean:     KES {sample_income.mean():,.0f}')
plt.close()
```
