# Day 6: Skewness, Kurtosis, and Shape of Distributions

## Phase 1 | Month 4 | Week 1 | Day 6

**🎯 Goal:** Quantify the shape of a distribution using skewness and kurtosis.
**⏱️ Study Session:** ~2 hours

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
symmetric    = np.random.normal(50, 10, 1000)
right_skewed = np.random.exponential(scale=10, size=1000) + 20
left_skewed  = 100 - np.random.exponential(scale=10, size=1000)

for name, data in [('Symmetric', symmetric), ('Right-skewed', right_skewed), ('Left-skewed', left_skewed)]:
    sk = stats.skew(data)
    ku = stats.kurtosis(data)   # excess kurtosis (normal=0)
    print(f'{name:15s}  skew={sk:+.3f}  kurt={ku:+.3f}  mean={data.mean():.1f}  median={np.median(data):.1f}')
```

```python
import numpy as np
from scipy import stats
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
distributions = {
    'Symmetric\n(Normal)':     np.random.normal(50, 10, 1000),
    'Right-Skewed\n(Exponential)': np.random.exponential(10, 1000)+20,
    'Left-Skewed':             100 - np.random.exponential(10, 1000),
    'Heavy-Tailed\n(Student-t df=3)': stats.t.rvs(df=3, size=1000)*10+50,
}

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, (name, data) in zip(axes, distributions.items()):
    ax.hist(data, bins=40, color='steelblue', edgecolor='white', alpha=0.7)
    ax.axvline(data.mean(),      color='red',   lw=2, label='Mean')
    ax.axvline(np.median(data),  color='green', lw=2, ls='--', label='Median')
    ax.set_title(f'{name}\nskew={stats.skew(data):+.2f}', fontsize=9, fontweight='bold')
    ax.legend(fontsize=7)
plt.tight_layout()
fig.savefig('/tmp/day06_shape.png', dpi=120)
print('Saved /tmp/day06_shape.png')
plt.close()
```

## Interpreting Skewness and Kurtosis

| Value | Skewness | Distribution |
|-------|----------|-------------|
| 0 | Symmetric | Normal |
| > 0 (positive) | Right-skewed | Income, rainfall |
| < 0 (negative) | Left-skewed | Exam scores (when easy) |

| Excess Kurtosis | Type | Tails |
|----------------|------|-------|
| 0 | Mesokurtic | Normal tails |
| > 0 | Leptokurtic | Heavy/fat tails (stock returns) |
| < 0 | Platykurtic | Light/thin tails |
