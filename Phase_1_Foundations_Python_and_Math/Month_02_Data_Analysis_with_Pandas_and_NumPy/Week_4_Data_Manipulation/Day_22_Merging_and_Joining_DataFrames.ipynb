```python
import pandas as pd
import numpy as np
np.random.seed(0)

customers = pd.DataFrame({'customer_id':[101,102,103,104,105,106],'name':['Alice','Bob','Carol','David','Eve','Frank'],'county':['Nairobi','Kisumu','Nakuru','Nairobi','Mombasa','Eldoret'],'tier':['Gold','Silver','Bronze','Gold','Silver','Bronze']})
transactions = pd.DataFrame({'txn_id':[1001,1002,1003,1004,1005,1006,1007,1008],'customer_id':[101,102,103,101,104,102,999,105],'amount_kes':[5000,2000,8000,3000,12000,1500,700,4000],'type':['Deposit','Withdraw','Transfer','Deposit','Withdraw','Deposit','Transfer','Deposit']})

# INNER JOIN — only matching rows
inner = pd.merge(customers, transactions, on='customer_id', how='inner')
print(f'INNER JOIN: {len(inner)} rows (txn 1007 excluded, cust 106 excluded)')
print(inner[['name','amount_kes','type']])

# LEFT JOIN — all customers, matched transactions
left = pd.merge(customers, transactions, on='customer_id', how='left')
print(f'\nLEFT JOIN: {len(left)} rows (Frank has NaN transactions)')
print(left[left['txn_id'].isna()][['name','tier']])

# RIGHT JOIN — all transactions, matched customers
right = pd.merge(customers, transactions, on='customer_id', how='right')
print(f'\nRIGHT JOIN: {len(right)} rows (txn 1007 has NaN customer info)')
print(right[right['name'].isna()][['txn_id','customer_id','amount_kes']])

# OUTER JOIN — everything
outer = pd.merge(customers, transactions, on='customer_id', how='outer')
print(f'\nOUTER JOIN: {len(outer)} rows')
```

```python
import pandas as pd
import numpy as np

# pd.concat — stacking DataFrames
np.random.seed(1)
jan = pd.DataFrame({'month':['Jan']*5,'sales':np.random.randint(100,1000,5),'agent':[f'A{i}' for i in range(1,6)]})
feb = pd.DataFrame({'month':['Feb']*5,'sales':np.random.randint(100,1000,5),'agent':[f'A{i}' for i in range(1,6)]})
mar = pd.DataFrame({'month':['Mar']*5,'sales':np.random.randint(100,1000,5),'agent':[f'A{i}' for i in range(1,6)]})

# Vertical stack
all_months = pd.concat([jan, feb, mar], ignore_index=True)
print('Concatenated (15 rows):')
print(all_months)
print('\nTotal sales by month:')
print(all_months.groupby('month')['sales'].sum())
```
