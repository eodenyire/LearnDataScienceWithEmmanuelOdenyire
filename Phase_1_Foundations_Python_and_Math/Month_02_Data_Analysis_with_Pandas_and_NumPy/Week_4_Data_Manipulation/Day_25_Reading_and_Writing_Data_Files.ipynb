```python
import pandas as pd
import numpy as np
import io, os
np.random.seed(42)

# Create sample dataset
df = pd.DataFrame({
    'date'    : pd.date_range('2024-01-01', periods=100, freq='D'),
    'county'  : np.random.choice(['Nairobi','Mombasa','Kisumu'],100),
    'product' : np.random.choice(['Tea','Coffee','Maize','Beans'],100),
    'quantity': np.random.randint(10,500,100),
    'price'   : np.random.uniform(50,500,100).round(2),
})
df['revenue'] = (df['quantity'] * df['price']).round(2)

# Write CSV
df.to_csv('/tmp/sales.csv', index=False)
df.to_csv('/tmp/sales_indexed.csv', index=True)

# Read back with options
df_loaded = pd.read_csv('/tmp/sales.csv', parse_dates=['date'])
print('Loaded CSV shape:', df_loaded.shape)
print('dtypes:', df_loaded.dtypes.to_dict())

# Read subset of columns and rows
df_subset = pd.read_csv('/tmp/sales.csv', usecols=['county','product','revenue'], nrows=10)
print('\nSubset read:')
print(df_subset)
```

```python
import pandas as pd
import numpy as np
import json
np.random.seed(42)

df = pd.DataFrame({'id':range(1,6),'name':['Alice','Bob','Carol','David','Eve'],'score':[85,73,90,55,88],'county':['Nairobi','Kisumu','Nakuru','Mombasa','Eldoret']})

# JSON formats
json_records = df.to_json(orient='records', indent=2)
print('JSON records format (first 200 chars):')
print(json_records[:200])

df_from_json = pd.read_json(io.StringIO(json_records), orient='records')
print('\nLoaded back from JSON:', df_from_json.shape)

# Parquet (columnar, compressed, fast)
try:
    df.to_parquet('/tmp/data.parquet', index=False)
    df_parquet = pd.read_parquet('/tmp/data.parquet')
    csv_size = os.path.getsize('/tmp/data.parquet')
    print(f'\nParquet file size: {csv_size} bytes (very compact!)')
    print('Parquet loaded:', df_parquet.shape)
except Exception as e:
    print(f'Parquet requires pyarrow: pip install pyarrow ({e})')

# SQLite
try:
    import sqlite3
    conn = sqlite3.connect('/tmp/data.db')
    df.to_sql('employees', conn, if_exists='replace', index=False)
    df_sql = pd.read_sql('SELECT * FROM employees WHERE score > 80', conn)
    print('\nSQL query result:', df_sql.shape)
    print(df_sql)
    conn.close()
except Exception as e:
    print(f'SQLite error: {e}')
```
