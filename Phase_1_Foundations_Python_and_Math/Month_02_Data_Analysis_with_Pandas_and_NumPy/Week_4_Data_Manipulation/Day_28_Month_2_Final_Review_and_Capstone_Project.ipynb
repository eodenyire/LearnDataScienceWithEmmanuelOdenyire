## Capstone Project: Comprehensive Kenya Economic Data Analysis

**Background:** You have been hired by the Kenya National Bureau of Statistics (KNBS) to analyse and report on economic indicators across Kenya's 47 counties.

**Deliverables:**
1. Load and clean the multi-source dataset
2. Exploratory Data Analysis (EDA)
3. Statistical analysis with NumPy
4. Visualisation-ready summary tables
5. Save complete report to CSV and JSON

```python
import pandas as pd
import numpy as np
np.random.seed(2024)

# DATASET 1: County Demographics
counties = ['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret','Kisii','Thika','Meru',
            'Kakamega','Machakos','Nyeri','Garissa','Isiolo','Turkana','Wajir']

demographics = pd.DataFrame({
    'county'        : counties,
    'population'    : [4_700_000,1_400_000,1_100_000,2_200_000,900_000,1_300_000,
                       1_000_000,850_000,1_800_000,1_200_000,750_000,850_000,
                       350_000,1_100_000,900_000],
    'area_km2'      : [695,212,2085,7496,3015,1317,1490,6936,3325,
                       6208,3337,45720,25336,77000,56685],
    'urban_pct'     : [98,80,60,55,65,35,70,42,38,55,58,42,30,25,20],
    'literacy_pct'  : [92,82,78,80,75,77,83,74,72,79,85,45,60,42,38],
    'region'        : ['Nairobi','Coast','Nyanza','Rift Valley','Rift Valley',
                       'Nyanza','Central','Eastern','Western','Eastern',
                       'Central','NEP','Eastern','Rift Valley','NEP'],
})

# DATASET 2: Economic Indicators
np.random.seed(1)
economics = pd.DataFrame({
    'county'        : counties,
    'gdp_billion_kes': np.random.uniform(10, 500, len(counties)).round(1),
    'unemployment_pct': np.random.uniform(8, 35, len(counties)).round(1),
    'poverty_pct'   : np.random.uniform(10, 60, len(counties)).round(1),
    'investment_2023': np.random.uniform(1, 100, len(counties)).round(1),
})

# DATASET 3: Infrastructure
np.random.seed(2)
infrastructure = pd.DataFrame({
    'county'        : counties,
    'hospitals'     : np.random.randint(5, 80, len(counties)),
    'schools'       : np.random.randint(50, 1500, len(counties)),
    'roads_km'      : np.random.randint(200, 5000, len(counties)),
    'internet_pct'  : np.random.uniform(5, 85, len(counties)).round(1),
})

print('Dataset 1 — Demographics:')
print(demographics.head(3))
print('\nDataset 2 — Economics:')
print(economics.head(3))
print('\nDataset 3 — Infrastructure:')
print(infrastructure.head(3))
```

```python
import pandas as pd
import numpy as np
np.random.seed(2024)
counties = ['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret','Kisii','Thika','Meru','Kakamega','Machakos','Nyeri','Garissa','Isiolo','Turkana','Wajir']
demographics = pd.DataFrame({'county':counties,'population':[4700000,1400000,1100000,2200000,900000,1300000,1000000,850000,1800000,1200000,750000,850000,350000,1100000,900000],'area_km2':[695,212,2085,7496,3015,1317,1490,6936,3325,6208,3337,45720,25336,77000,56685],'urban_pct':[98,80,60,55,65,35,70,42,38,55,58,42,30,25,20],'literacy_pct':[92,82,78,80,75,77,83,74,72,79,85,45,60,42,38],'region':['Nairobi','Coast','Nyanza','Rift Valley','Rift Valley','Nyanza','Central','Eastern','Western','Eastern','Central','NEP','Eastern','Rift Valley','NEP']})
np.random.seed(1)
economics = pd.DataFrame({'county':counties,'gdp_billion_kes':np.random.uniform(10,500,15).round(1),'unemployment_pct':np.random.uniform(8,35,15).round(1),'poverty_pct':np.random.uniform(10,60,15).round(1),'investment_2023':np.random.uniform(1,100,15).round(1)})
np.random.seed(2)
infrastructure = pd.DataFrame({'county':counties,'hospitals':np.random.randint(5,80,15),'schools':np.random.randint(50,1500,15),'roads_km':np.random.randint(200,5000,15),'internet_pct':np.random.uniform(5,85,15).round(1)})

# STEP 1: Merge all three datasets
df = demographics.merge(economics, on='county').merge(infrastructure, on='county')
print(f'Merged dataset: {df.shape}')

# STEP 2: Feature engineering
df['gdp_per_capita']   = (df['gdp_billion_kes'] * 1e9 / df['population']).round(0)
df['pop_density']      = (df['population'] / df['area_km2']).round(1)
df['hospitals_per_100k'] = (df['hospitals'] / df['population'] * 100_000).round(2)
df['schools_per_100k']   = (df['schools']   / df['population'] * 100_000).round(2)
df['development_index']  = (
    0.3 * (df['literacy_pct'] / 100)
    + 0.3 * (1 - df['poverty_pct'] / 100)
    + 0.2 * (df['gdp_per_capita'] / df['gdp_per_capita'].max())
    + 0.2 * (df['internet_pct'] / 100)
).round(4)

# STEP 3: Rankings
df['gdp_rank']  = df['gdp_per_capita'].rank(ascending=False).astype(int)
df['dev_rank']  = df['development_index'].rank(ascending=False).astype(int)

print('\nTop 5 counties by Development Index:')
cols = ['county','region','development_index','gdp_per_capita','literacy_pct','internet_pct']
print(df.nlargest(5, 'development_index')[cols].to_string())

print('\nBottom 5 counties:')
print(df.nsmallest(5, 'development_index')[cols].to_string())
```

```python
import pandas as pd
import numpy as np
np.random.seed(2024)
counties = ['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret','Kisii','Thika','Meru','Kakamega','Machakos','Nyeri','Garissa','Isiolo','Turkana','Wajir']
demographics = pd.DataFrame({'county':counties,'population':[4700000,1400000,1100000,2200000,900000,1300000,1000000,850000,1800000,1200000,750000,850000,350000,1100000,900000],'area_km2':[695,212,2085,7496,3015,1317,1490,6936,3325,6208,3337,45720,25336,77000,56685],'urban_pct':[98,80,60,55,65,35,70,42,38,55,58,42,30,25,20],'literacy_pct':[92,82,78,80,75,77,83,74,72,79,85,45,60,42,38],'region':['Nairobi','Coast','Nyanza','Rift Valley','Rift Valley','Nyanza','Central','Eastern','Western','Eastern','Central','NEP','Eastern','Rift Valley','NEP']})
np.random.seed(1)
economics = pd.DataFrame({'county':counties,'gdp_billion_kes':np.random.uniform(10,500,15).round(1),'unemployment_pct':np.random.uniform(8,35,15).round(1),'poverty_pct':np.random.uniform(10,60,15).round(1),'investment_2023':np.random.uniform(1,100,15).round(1)})
np.random.seed(2)
infrastructure = pd.DataFrame({'county':counties,'hospitals':np.random.randint(5,80,15),'schools':np.random.randint(50,1500,15),'roads_km':np.random.randint(200,5000,15),'internet_pct':np.random.uniform(5,85,15).round(1)})
df = demographics.merge(economics, on='county').merge(infrastructure, on='county')
df['gdp_per_capita'] = (df['gdp_billion_kes']*1e9/df['population']).round(0)
df['development_index'] = (0.3*(df['literacy_pct']/100)+0.3*(1-df['poverty_pct']/100)+0.2*(df['gdp_per_capita']/df['gdp_per_capita'].max())+0.2*(df['internet_pct']/100)).round(4)

# STEP 4: Regional analysis
print('=== Regional Summary ===')
regional = df.groupby('region').agg(
    counties        =('county', 'count'),
    total_pop       =('population', 'sum'),
    mean_gdp_pc     =('gdp_per_capita', 'mean'),
    mean_literacy   =('literacy_pct', 'mean'),
    mean_dev_index  =('development_index', 'mean'),
    mean_poverty    =('poverty_pct', 'mean'),
).round(2)
print(regional.sort_values('mean_dev_index', ascending=False).to_string())

# STEP 5: Correlation analysis with NumPy
numeric_cols = ['gdp_per_capita','literacy_pct','urban_pct','internet_pct',
                'poverty_pct','unemployment_pct','hospitals_per_100k']
df['hospitals_per_100k'] = df['hospitals']/df['population']*100_000
X = df[numeric_cols].values
corr = np.corrcoef(X.T)
pd_corr = pd.DataFrame(corr, index=numeric_cols, columns=numeric_cols)
print('\n=== Correlation Matrix ===')
print(pd_corr.round(2).to_string())

# Save report
df.to_csv('/tmp/kenya_county_report.csv', index=False)
print('\nReport saved to /tmp/kenya_county_report.csv')
```
