# Day 23: Reshaping and Pivoting Data

## Phase 1 | Month 2 | Week 4 | Day 23

**Goal:** Transform wide data to long and back — essential for visualisation, modelling, and reporting.

### What You Will Learn
1. `pivot()` — long to wide
2. `melt()` — wide to long
3. `pivot_table()` — with aggregation
4. `stack()` and `unstack()`
5. When to use each

### Resources

```python
import pandas as pd
import numpy as np
np.random.seed(42)

# Wide format — each row is a student, columns are subjects
wide = pd.DataFrame({
    'student': ['Alice','Bob','Carol','David'],
    'maths'  : [85, 73, 90, 55],
    'english': [78, 91, 84, 65],
    'science': [92, 88, 67, 78],
})
print('Wide format:')
print(wide)

# MELT — wide to long (tidy format)
long = pd.melt(
    wide,
    id_vars   =['student'],
    value_vars=['maths','english','science'],
    var_name  ='subject',
    value_name='score'
)
print('\nLong (tidy) format:')
print(long)
print('\nMean score by subject:')
print(long.groupby('subject')['score'].mean().round(1))
```

```python
import pandas as pd
import numpy as np
np.random.seed(42)

# PIVOT — long back to wide
long = pd.DataFrame({
    'student': ['Alice','Alice','Alice','Bob','Bob','Bob'],
    'subject': ['maths','english','science']*2,
    'score'  : [85,78,92,73,91,88],
})

pivot = long.pivot(index='student', columns='subject', values='score')
print('Pivot table:')
print(pivot)
print('Column names:', pivot.columns.tolist())
pivot.columns.name = None  # remove 'subject' label
pivot = pivot.reset_index()
print('\nCleaned pivot:')
print(pivot)

# PIVOT TABLE with aggregation
np.random.seed(1)
df = pd.DataFrame({
    'county' : np.random.choice(['Nairobi','Kisumu','Nakuru'],30),
    'product': np.random.choice(['Maize','Beans','Rice'],30),
    'sales'  : np.random.randint(100,1000,30),
})
pvt = df.pivot_table(values='sales',index='county',columns='product',aggfunc='sum',fill_value=0)
print('\nSales pivot table:')
print(pvt)
```
