## 🏗️ Mini-Project: Kenya County Development Index Analysis

**Scenario:** You are a data analyst at the Kenya National Bureau of Statistics (KNBS). Using NumPy, build a comprehensive analysis of Kenya's 47 counties across 5 development indicators.

**Indicators:**
1. Literacy rate (%)
2. Access to clean water (%)
3. Healthcare coverage (%)
4. Road density (km/km²)
5. Mobile connectivity (%)

**Tasks:**
1. Generate/load the data
2. Compute a composite Development Index
3. Rank counties
4. Identify outliers
5. Run a bootstrap confidence interval
6. Generate a statistical report

```python
import numpy as np
np.random.seed(2024)

# ── Step 1: Simulate county data ──
n_counties = 47
county_names = [f"County_{i:02d}" for i in range(1, n_counties + 1)]
# (In practice you would load real KNBS data with np.loadtxt or pd.read_csv)

# Simulate 5 development indicators (values 0–100)
np.random.seed(2024)
literacy     = np.random.beta(7, 2, n_counties) * 100   # skewed high
water        = np.random.beta(5, 3, n_counties) * 100
healthcare   = np.random.beta(4, 4, n_counties) * 100
road_density = np.random.exponential(30, n_counties).clip(0, 100)
connectivity = np.random.beta(6, 2, n_counties) * 100

# Stack into matrix: 47 rows × 5 features
X = np.column_stack([literacy, water, healthcare, road_density, connectivity])
indicators = ["Literacy", "Water", "Healthcare", "Roads", "Connectivity"]

print("Data shape:", X.shape)
print("\nIndicator summaries:")
for name, col in zip(indicators, X.T):
    print(f"  {name:15}: mean={col.mean():.1f}  std={col.std():.1f}  min={col.min():.1f}  max={col.max():.1f}")
```

```python
import numpy as np
np.random.seed(2024)

# Recreate data
n_counties = 47
literacy     = np.random.beta(7, 2, n_counties) * 100
water        = np.random.beta(5, 3, n_counties) * 100
healthcare   = np.random.beta(4, 4, n_counties) * 100
road_density = np.random.exponential(30, n_counties).clip(0, 100)
connectivity = np.random.beta(6, 2, n_counties) * 100
X = np.column_stack([literacy, water, healthcare, road_density, connectivity])

# ── Step 2: Normalise to 0–1 ──
X_min  = X.min(axis=0)
X_max  = X.max(axis=0)
X_norm = (X - X_min) / (X_max - X_min)

# ── Step 3: Composite Development Index (equal weights) ──
weights = np.array([0.25, 0.20, 0.20, 0.15, 0.20])   # must sum to 1
assert abs(weights.sum() - 1.0) < 1e-9, "Weights must sum to 1"
CDI = X_norm @ weights   # matrix multiply: (47,5) @ (5,) = (47,)

# ── Step 4: Rank counties ──
ranks = np.argsort(CDI)[::-1]   # descending

print("=== Kenya County Development Index ===")
print("\nTop 5 Counties:")
for i, idx in enumerate(ranks[:5], 1):
    print(f"  {i}. County_{idx+1:02d}: CDI = {CDI[idx]:.3f}")

print("\nBottom 5 Counties:")
for i, idx in enumerate(ranks[-5:][::-1], 1):
    print(f"  {i}. County_{idx+1:02d}: CDI = {CDI[idx]:.3f}")

# ── Step 5: Outlier detection (IQR method) ──
q1, q3 = np.percentile(CDI, 25), np.percentile(CDI, 75)
iqr = q3 - q1
outliers = np.where((CDI < q1 - 1.5*iqr) | (CDI > q3 + 1.5*iqr))[0]
print(f"\nOutlier counties: {len(outliers)} detected")

# ── Step 6: Bootstrap 95% CI of mean CDI ──
n_boot = 10_000
boot_means = np.array([
    np.random.choice(CDI, len(CDI), replace=True).mean()
    for _ in range(n_boot)
])
ci_low, ci_high = np.percentile(boot_means, [2.5, 97.5])
print(f"\n95% CI for mean CDI: [{ci_low:.3f}, {ci_high:.3f}]")
print(f"Observed mean CDI  : {CDI.mean():.3f}")
```
