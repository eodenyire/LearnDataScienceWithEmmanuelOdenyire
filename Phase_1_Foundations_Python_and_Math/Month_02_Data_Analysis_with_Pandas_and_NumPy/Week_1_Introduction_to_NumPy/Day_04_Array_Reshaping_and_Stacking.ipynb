## 1. Reshaping Arrays

`arr.reshape(new_shape)` changes the layout of elements **without copying data**.

**Key rule:** The total number of elements must stay the same.
- `arr.reshape(3, 4)` works only if `arr.size == 12`
- Use `-1` for NumPy to auto-calculate one dimension

```python
import numpy as np

arr = np.arange(1, 25)   # 24 elements
print("Original:", arr)

# Different reshape options
print("\n4×6:\n", arr.reshape(4, 6))
print("\n6×4:\n", arr.reshape(6, 4))
print("\n2×3×4:\n", arr.reshape(2, 3, 4))
print("\nAuto-rows (-1, 4):\n", arr.reshape(-1, 4))   # NumPy computes rows=6

# ── MNIST-like image flattening ──
image = np.random.randint(0, 256, (28, 28))   # 28×28 grayscale image
flat = image.reshape(-1)                       # flatten to 784
print(f"\nImage shape : {image.shape}")
print(f"Flattened   : {flat.shape}")
print(f"Back to 2D  : {flat.reshape(28, 28).shape}")
```

## 2. Flatten, Ravel, Transpose

| Method | Returns | Notes |
|--------|---------|-------|
| `.flatten()` | Always a copy | Safe to modify |
| `.ravel()` | Usually a view | Faster, modifies original |
| `.T` | View of transposed | Swaps all axes |
| `np.transpose(arr)` | Configurable | Specify axis order |

```python
import numpy as np

m = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Original (2×3):\n", m)
print("Transposed (3×2):\n", m.T)
print("Flatten:", m.flatten())
print("Ravel :", m.ravel())

# ── Transposing for matrix multiplication ──
A = np.array([[1, 2], [3, 4], [5, 6]])   # 3×2
B = np.array([[7, 8, 9], [10, 11, 12]])  # 2×3
C = A @ B                                # 3×2 @ 2×3 = 3×3
print(f"\nA ({A.shape}) @ B ({B.shape}) = C ({C.shape}):")
print(C)
```

## 3. Stacking and Splitting

Build combined datasets from separate arrays — exactly what you do when merging features in a data pipeline.

```python
import numpy as np
np.random.seed(1)

# ── Vertical stacking — add rows ──
batch1 = np.array([[1, 2, 3], [4, 5, 6]])
batch2 = np.array([[7, 8, 9], [10, 11, 12]])
combined = np.vstack([batch1, batch2])
print("vstack:\n", combined)

# ── Horizontal stacking — add columns ──
features = np.array([[85, 92], [78, 88], [90, 75]])     # 3 students, 2 features
extra    = np.array([[1], [0], [1]])                     # 1 more feature
full     = np.hstack([features, extra])
print("\nhstack (add feature column):\n", full)

# ── Concatenate (general) ──
a = np.array([[1, 2, 3]])
b = np.array([[4, 5, 6]])
print("\nconcat axis=0 (rows):", np.concatenate([a, b], axis=0))
print("concat axis=1 (cols):", np.concatenate([a, b], axis=1))

# ── Splitting ──
big = np.arange(1, 25).reshape(4, 6)
parts = np.hsplit(big, 3)   # split into 3 equal column blocks
print("\nSplit into 3 horizontal parts:")
for i, p in enumerate(parts):
    print(f"  Part {i}: {p}")

# ── Train / test split (80/20) ──
data = np.arange(100)
split = int(0.8 * len(data))
train, test = data[:split], data[split:]
print(f"\nTrain: {len(train)} samples, Test: {len(test)} samples")
```
