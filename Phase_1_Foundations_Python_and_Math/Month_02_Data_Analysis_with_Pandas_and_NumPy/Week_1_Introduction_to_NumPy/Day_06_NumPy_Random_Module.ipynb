````python
import numpy as np
np.random.seed(42)   # Always set seed for reproducibility!

# ── Key distributions ──
print("=== Random Number Generation ===")
print("Uniform [0,1)  :", np.random.rand(5).round(3))
print("Standard normal:", np.random.randn(5).round(3))
print("Integers [1,7) :", np.random.randint(1, 7, 10))  # dice rolls
print("Normal(μ=170,σ=10):", np.random.normal(170, 10, 5).round(1))
print("Exponential(λ=2)  :", np.random.exponential(2, 5).round(3))
print("Binomial(n=10,p=0.3):", np.random.binomial(10, 0.3, 5))

# ── New-style API (preferred in NumPy ≥ 1.17) ──
rng = np.random.default_rng(seed=2024)
samples = rng.normal(loc=0, scale=1, size=(3, 4))
print("\nNew-style RNG (3×4 standard normal):\n", samples.round(3))
````

## Reproducibility — Critical in Data Science

**Always set a random seed** before any stochastic operation. This ensures:
- Your results can be reproduced by others
- Your experiments are comparable
- Your code reviews are meaningful

```python
np.random.seed(42)          # Legacy API
rng = np.random.default_rng(42)  # Modern API (preferred)
```

Every major ML paper sets a random seed. Get in the habit now.

````python
import numpy as np

# ── Random sampling ──
np.random.seed(42)

population = np.arange(1, 1001)   # 1000 Kenyan households

# Simple random sample (without replacement)
sample = np.random.choice(population, size=100, replace=False)
print(f"Population mean : {population.mean():.1f}")
print(f"Sample mean     : {sample.mean():.1f}")   # should be close

# Weighted sampling (e.g., richer counties sampled more)
weights = np.random.dirichlet(np.ones(1000))  # random non-uniform weights
weighted_sample = np.random.choice(population, size=100, p=weights)
print(f"Weighted sample mean: {weighted_sample.mean():.1f}")

# ── Shuffling ──
deck = np.arange(1, 53)          # 52 cards
np.random.shuffle(deck)
print("\nShuffled deck (first 10):", deck[:10])

# ── permutation (doesn't modify original) ──
arr = np.array([1, 2, 3, 4, 5])
perm = np.random.permutation(arr)
print("\nOriginal  :", arr)
print("Permuted  :", perm)
````

## Monte Carlo Simulation

Monte Carlo methods use repeated random sampling to estimate numerical results. Named after the famous casino — using randomness to solve deterministic problems.

**Used at:** Google (A/B testing), Netflix (recommendation exploration), insurance companies (risk modelling), central banks (economic forecasting).

````python
import numpy as np
np.random.seed(42)

# ── Classic: Estimate π ──
n = 1_000_000
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)
inside = (x**2 + y**2) <= 1
pi_est = 4 * inside.mean()
print(f"Monte Carlo estimate of π: {pi_est:.6f}")
print(f"True value             : {np.pi:.6f}")
print(f"Error                  : {abs(pi_est - np.pi):.6f}")

# ── Business simulation: Safaricom daily revenue ──
n_days = 365
# Revenue from mobile data (billions KES)
data_revenue = np.random.normal(loc=1.2, scale=0.15, size=n_days)
# Revenue from M-Pesa
mpesa_revenue = np.random.normal(loc=2.1, scale=0.25, size=n_days)
# Revenue from voice calls
voice_revenue = np.random.normal(loc=0.9, scale=0.10, size=n_days)

total = data_revenue + mpesa_revenue + voice_revenue
print(f"\n=== Safaricom Revenue Simulation (365 days) ===")
print(f"Daily mean    : KES {total.mean():.3f} billion")
print(f"Annual total  : KES {total.sum():.1f} billion")
print(f"Best day      : KES {total.max():.3f} billion")
print(f"Worst day     : KES {total.min():.3f} billion")
print(f"Days > 5B KES : {(total > 5).sum()}")
````
