```python
import numpy as np

# ── Trigonometric ──
angles = np.array([0, 30, 45, 60, 90])
rad    = np.deg2rad(angles)
print("Degrees:", angles)
print("sin    :", np.sin(rad).round(3))
print("cos    :", np.cos(rad).round(3))
print("tan    :", np.tan(rad).round(3))

# ── Exponential and logarithm ──
x = np.array([1, 2, 4, 8, 16, 32])
print("\nx      :", x)
print("log2   :", np.log2(x))
print("log10  :", np.log10(x).round(3))
print("exp    :", np.exp([0, 1, 2]).round(3))

# ── Rounding ──
vals = np.array([1.234, 2.678, -3.456, 4.500])
print("\nfloor  :", np.floor(vals))
print("ceil   :", np.ceil(vals))
print("round2 :", np.round(vals, 2))
print("abs    :", np.abs(vals))
```

## Statistical Functions

NumPy provides fast statistical summaries across full arrays or along specific axes.

| Function | Description |
|----------|-------------|
| `np.mean(arr)` | Arithmetic mean |
| `np.median(arr)` | Middle value |
| `np.std(arr)` | Standard deviation |
| `np.var(arr)` | Variance |
| `np.percentile(arr, q)` | qth percentile |
| `np.quantile(arr, q)` | Same, 0–1 scale |
| `np.corrcoef(a, b)` | Correlation matrix |
| `np.histogram(arr)` | Frequency counts |

```python
import numpy as np
np.random.seed(42)

# Simulate KCSE exam scores for 1000 Form 4 students
scores = np.random.normal(loc=58, scale=14, size=1000).clip(0, 84)

print("=== KCSE Score Statistics (n=1000) ===")
print(f"Mean          : {np.mean(scores):.2f}")
print(f"Median        : {np.median(scores):.2f}")
print(f"Std deviation : {np.std(scores):.2f}")
print(f"Min / Max     : {scores.min():.1f} / {scores.max():.1f}")
print(f"25th pctile   : {np.percentile(scores, 25):.1f}")
print(f"75th pctile   : {np.percentile(scores, 75):.1f}")
print(f"IQR           : {np.percentile(scores, 75) - np.percentile(scores, 25):.1f}")

# Grade distribution
grades = ["A (75-84)", "B (60-74)", "C (45-59)", "D (30-44)", "E (<30)"]
thresholds = [(75, 84), (60, 74), (45, 59), (30, 44), (0, 29)]
print("\nGrade Distribution:")
for g, (lo, hi) in zip(grades, thresholds):
    n = ((scores >= lo) & (scores <= hi)).sum()
    pct = n / len(scores) * 100
    bar = "█" * int(pct // 2)
    print(f"  {g}: {n:4d} ({pct:.1f}%) {bar}")
```

```python
import numpy as np
np.random.seed(0)

# ── Sorting ──
data = np.array([64, 32, 99, 15, 47, 83, 21, 56])
print("Original  :", data)
print("Sorted    :", np.sort(data))
print("Argsort   :", np.argsort(data))   # indices that sort array
print("Top-3 idx :", np.argsort(data)[-3:])

# ── Searching ──
print("\nWhere > 50:", np.where(data > 50))
print("Values > 50:", data[np.where(data > 50)])

# ── Correlation — key for feature selection ──
np.random.seed(10)
study_hours  = np.array([2, 4, 6, 8, 10, 12, 5, 7, 9, 3])
exam_scores  = study_hours * 5 + np.random.normal(0, 5, 10)
sleepHrs     = np.random.uniform(5, 9, 10)

corr_study   = np.corrcoef(study_hours, exam_scores)[0, 1]
corr_sleep   = np.corrcoef(sleepHrs, exam_scores)[0, 1]
print(f"\nCorrelation(study hours, exam score) = {corr_study:.3f}")
print(f"Correlation(sleep hours, exam score) = {corr_sleep:.3f}")
print("Strong positive correlation with study, weak with sleep 📚")
```
