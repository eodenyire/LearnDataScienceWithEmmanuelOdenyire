```python
import numpy as np
np.random.seed(42)

# Simulate exam scores: 4 subjects, 100 students
n = 100
maths   = np.random.normal(70, 12, n)
science = 0.7 * maths + np.random.normal(0, 8, n)  # correlated
english = np.random.normal(65, 15, n)              # independent
coding  = 0.5*maths + 0.3*english + np.random.normal(0, 10, n)

X = np.column_stack([maths, science, english, coding])
labels = ['Maths', 'Science', 'English', 'Coding']

corr = np.corrcoef(X.T)
print('Correlation Matrix:')
print(f'{'':10}', '  '.join(f'{l:8}' for l in labels))
for i, name in enumerate(labels):
    row = '  '.join(f'{corr[i,j]:8.3f}' for j in range(4))
    print(f'{name:10} {row}')
print(f'\nMaths-Science correlation: {corr[0,1]:.3f} (expected ~0.7)')
```

```python
import numpy as np
np.random.seed(0)

# PCA FROM SCRATCH
n, p = 200, 5
X = np.random.randn(n, p)
X[:, 1] = 0.8 * X[:, 0] + np.random.randn(n) * 0.4  # corr
X[:, 2] = 0.6 * X[:, 0] + 0.3 * X[:, 1] + np.random.randn(n) * 0.3

# Step 1: Centre
X_c = X - X.mean(axis=0)

# Step 2: Covariance matrix
cov = np.cov(X_c.T)

# Step 3: Eigendecomposition
vals, vecs = np.linalg.eigh(cov)

# Step 4: Sort descending
idx  = np.argsort(vals)[::-1]
vals = vals[idx]
vecs = vecs[:, idx]

# Step 5: Explained variance
exp_var = vals / vals.sum()
cum_var = np.cumsum(exp_var)

print('PCA Explained Variance:')
for i, (ev, cv) in enumerate(zip(exp_var, cum_var)):
    bar = chr(9608) * int(ev * 40)
    print(f'  PC{i+1}: {ev*100:5.1f}%  (cum {cv*100:5.1f}%)  {bar}')

# Project onto first 2 PCs
X_pca = X_c @ vecs[:, :2]
print(f'\nReduced from {X.shape[1]}D to {X_pca.shape[1]}D')
print(f'Variance retained: {cum_var[1]*100:.1f}%')
print(f'Components for 95% variance: {np.searchsorted(cum_var, 0.95)+1}')
```

```python
import numpy as np
np.random.seed(5)

# PERMUTATION TEST — non-parametric hypothesis testing
control   = np.random.normal(65, 12, 50)
treatment = np.random.normal(70, 12, 50)

observed_diff = treatment.mean() - control.mean()
print(f'Observed mean difference: {observed_diff:.3f}')

all_scores = np.concatenate([control, treatment])
n_perm = 10_000
perm_diffs = np.zeros(n_perm)
for i in range(n_perm):
    perm = np.random.permutation(all_scores)
    perm_diffs[i] = perm[50:].mean() - perm[:50].mean()

p_value = (np.abs(perm_diffs) >= np.abs(observed_diff)).mean()
print(f'p-value: {p_value:.4f}')
print(f'Result : {"Significant (reject H0)" if p_value < 0.05 else "Not significant"}')
print(f'\n95th percentile of permutation distribution: {np.percentile(perm_diffs, 95):.3f}')
```
