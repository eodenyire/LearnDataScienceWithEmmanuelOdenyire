```python
import numpy as np
np.random.seed(1)

# Matrix multiplication — the core of neural networks
A = np.array([[1, 2], [3, 4], [5, 6]])   # 3x2
B = np.array([[7, 8, 9], [10, 11, 12]])  # 2x3
C = A @ B
print('A @ B =\n', C)
print('Shape:', C.shape)

# Neural net forward pass (simplified)
X  = np.random.randn(5, 3)
W1 = np.random.randn(3, 4)
b1 = np.zeros(4)
W2 = np.random.randn(4, 1)
b2 = np.zeros(1)
H  = np.maximum(0, X @ W1 + b1)  # ReLU
out = H @ W2 + b2
print('\n2-layer NN output:', out.flatten().round(3))
```

## Matrix Inverse and Solving Linear Systems

Solving `Ax = b` is at the core of:
- Linear regression (normal equations)
- Portfolio optimisation
- Physics simulations
- Computer graphics

```python
import numpy as np

# Solve: 3x + y = 9,  2x + 4y = 20
A = np.array([[3, 1], [2, 4]], dtype=float)
b = np.array([9.0, 20.0])

# Method 1: Direct solve (numerically stable, preferred)
x = np.linalg.solve(A, b)
print('Solution:', x)
print('Verify Ax:', A @ x)

# Method 2: Inverse (less stable, avoid for large systems)
A_inv = np.linalg.inv(A)
print('x via inverse:', A_inv @ b)

# Eigenvalues — foundation of PCA, PageRank, Markov chains
M = np.array([[4, 1], [2, 3]])
vals, vecs = np.linalg.eig(M)
print('\nEigenvalues:', vals)
print('Eigenvectors:\n', vecs.round(4))

# Least squares regression
x_data = np.arange(1, 11, dtype=float)
y_data = 2.5 * x_data + 1.0 + np.random.randn(10) * 0.5
X_des  = np.column_stack([x_data, np.ones(10)])
theta, _, _, _ = np.linalg.lstsq(X_des, y_data, rcond=None)
print(f'\nFitted: y = {theta[0]:.3f}x + {theta[1]:.3f}  (true: 2.5, 1.0)')
```

## Norms and Distances
