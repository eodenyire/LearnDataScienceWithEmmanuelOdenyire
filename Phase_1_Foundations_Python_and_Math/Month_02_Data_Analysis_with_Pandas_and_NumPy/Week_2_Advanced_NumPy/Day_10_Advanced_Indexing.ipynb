```python
import numpy as np
np.random.seed(42)

matrix = np.arange(1, 26).reshape(5, 5)
print('5x5 Matrix:\n', matrix)

# Fancy indexing — select arbitrary rows
print('\nRows [0,2,4]:\n', matrix[[0, 2, 4]])

# Select specific (row, col) pairs
rows = [0, 1, 2, 3, 4]
cols = [4, 3, 2, 1, 0]
print('Anti-diagonal:', matrix[rows, cols])

# np.ix_ — cross product of rows and cols
sel_rows = np.array([0, 2, 4])
sel_cols = np.array([1, 3])
ix = np.ix_(sel_rows, sel_cols)
print('\nnp.ix_ submatrix:\n', matrix[ix])
```

```python
import numpy as np
np.random.seed(7)

# EMBEDDING LOOKUP — used in Word2Vec, BERT, recommendation systems!
vocab_size = 10
embed_dim  = 4
embeddings = np.random.randn(vocab_size, embed_dim)

# Sentence as word IDs: 'the cat sat on the mat'
sentence = np.array([3, 1, 7, 3, 9, 5])

# Lookup: one line replaces a for-loop
sentence_vecs = embeddings[sentence]   # (6, 4)
print('Embedding table shape:', embeddings.shape)
print('Sentence IDs:', sentence)
print('Sentence embeddings (6x4):\n', sentence_vecs.round(3))

# Mean pooling — simple sentence representation
sentence_rep = sentence_vecs.mean(axis=0)
print('Mean-pooled sentence vector:', sentence_rep.round(3))
```

```python
import numpy as np

# STRUCTURED ARRAYS — typed tabular data without Pandas
dtype = np.dtype([
    ('name',   'U20'),
    ('age',    np.int32),
    ('salary', np.float64),
    ('county', 'U20'),
])

employees = np.array([
    ('Alice Njeri',   28, 95000.0,  'Nairobi'),
    ('Bob Otieno',    35, 120000.0, 'Kisumu'),
    ('Carol Wanjiku', 31, 85000.0,  'Nakuru'),
    ('David Kamau',   42, 200000.0, 'Nairobi'),
    ('Eve Achieng',   26, 78000.0,  'Mombasa'),
], dtype=dtype)

print('Names:', employees['name'])
print('Nairobi:', employees[employees['county'] == 'Nairobi']['name'])
print('High earners:', employees[employees['salary'] > 90000]['name'])
print(f'Mean salary: KES {employees["salary"].mean():,.0f}')
```
