## 1. Boolean Indexing

```python
import numpy as np

scores = np.array([78, 92, 55, 67, 88, 45, 95, 73])

# Boolean mask
passing = scores >= 70
print('Mask:', passing)
print('Passing scores:', scores[passing])
print('Failing scores:', scores[~passing])

# Conditional update
adjusted = scores.copy()
adjusted[adjusted < 50] = 50  # Floor at 50
print('Adjusted:', adjusted)
```

## 2. Broadcasting

Broadcasting lets NumPy operate on arrays of different shapes:

```python
# Adding a scalar to an array
arr = np.array([[1, 2, 3], [4, 5, 6]])
print('Original:\n', arr)
print('arr + 10:\n', arr + 10)

# Adding a 1D array to each row
row = np.array([100, 200, 300])
print('\narr + row:\n', arr + row)

# Column broadcasting
col = np.array([[10], [20]])
print('\narr + col:\n', arr + col)
```

## 3. Reshaping and Stacking

```python
a = np.arange(12)
print('Original:', a)
print('\nReshaped to (3,4):\n', a.reshape(3, 4))
print('\nReshaped to (2,2,3):\n', a.reshape(2, 2, 3))

# Stacking
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print('\nvstack:\n', np.vstack([x, y]))
print('hstack:', np.hstack([x, y]))
```

## 4. Random Number Generation

```python
rng = np.random.default_rng(seed=42)

# Uniform distribution
uniform = rng.uniform(0, 1, size=(2, 3))
print('Uniform [0,1):\n', uniform)

# Normal distribution
normal = rng.normal(loc=0, scale=1, size=8)
print('\nNormal (mean=0, std=1):', np.round(normal, 2))

# Random integers
ints = rng.integers(1, 7, size=10)  # Dice rolls
print('\nDice rolls:', ints)
```

## 5. Linear Algebra

```python
A = np.array([[2, 1], [5, 3]])
b = np.array([4, 7])

# Matrix multiplication
print('A @ A:\n', A @ A)

# Solve Ax = b
x = np.linalg.solve(A, b)
print('\nSolution to Ax=b:', x)

# Determinant and inverse
print('det(A):', np.linalg.det(A))
print('inv(A):\n', np.linalg.inv(A))
```
