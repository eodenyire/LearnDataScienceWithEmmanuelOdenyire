```python
import numpy as np
import time

N = 1_000_000
a = np.random.rand(N)
b = np.random.rand(N)

# Python loop
start = time.perf_counter()
result_loop = [a[i] + b[i] for i in range(N)]
loop_time = time.perf_counter() - start

# NumPy vectorised
start = time.perf_counter()
result_np = a + b
np_time = time.perf_counter() - start

print(f'Loop time : {loop_time*1000:.1f} ms')
print(f'NumPy time: {np_time*1000:.2f} ms')
print(f'Speedup   : {loop_time/np_time:.0f}x faster with NumPy!')
print(f'Results match: {np.allclose(result_loop, result_np)}')
```

## np.where — Vectorised If/Else

This is the NumPy equivalent of a list comprehension with a condition.
Used constantly in data cleaning, feature engineering, and model scoring.

```python
import numpy as np
np.random.seed(42)

scores = np.random.randint(40, 100, 20)
print('Scores:', scores)

# Simple conditional
passed = np.where(scores >= 60, 'Pass', 'Fail')
print('Pass/Fail:', passed)

# Multi-condition with np.select
conditions = [scores >= 80, scores >= 60, scores >= 50]
choices    = ['A', 'B', 'C']
grades     = np.select(conditions, choices, default='D')
print('Grades:', grades)

# np.where with arrays (ternary: if condition use a else b)
a_arr = np.array([1, -2, 3, -4, 5])
relu  = np.where(a_arr > 0, a_arr, 0)  # This is the ReLU activation!
print('\nReLU:', relu)
```

```python
import numpy as np
import time

# Demonstrate: NEVER use Python loops for element-wise ops
np.random.seed(0)
X = np.random.randn(10_000, 50)

# BAD: Python loop normalisation
start = time.perf_counter()
result_slow = np.zeros_like(X)
for i in range(X.shape[0]):
    row_min = X[i].min()
    row_max = X[i].max()
    result_slow[i] = (X[i] - row_min) / (row_max - row_min)
slow_t = time.perf_counter() - start

# GOOD: Vectorised normalisation
start = time.perf_counter()
row_min = X.min(axis=1, keepdims=True)
row_max = X.max(axis=1, keepdims=True)
result_fast = (X - row_min) / (row_max - row_min)
fast_t = time.perf_counter() - start

print(f'Slow (loop)  : {slow_t*1000:.1f} ms')
print(f'Fast (vector): {fast_t*1000:.2f} ms')
print(f'Speedup      : {slow_t/fast_t:.0f}x')
print(f'Results match: {np.allclose(result_slow, result_fast)}')
```
