# Day 12: NumPy File I/O

## Phase 1 | Month 2 | Week 2 | Day 12

**Goal:** Save and load NumPy arrays efficiently — essential for caching processed data and sharing work.

### What You Will Learn
1. `.npy` format — fastest binary storage
2. `.npz` — save multiple named arrays
3. `np.genfromtxt` for reading text/CSV
4. Compressed storage

### Resources
- 📖 [NumPy I/O Reference](https://numpy.org/doc/stable/reference/routines.io.html)
- 🎥 [Loading Data with NumPy](https://www.youtube.com/watch?v=WG-cHlZS_TY)

```python
import numpy as np
import os

np.random.seed(42)
X = np.random.randn(1000, 50)
y = np.random.randint(0, 3, 1000)

# Save and load single arrays
np.save('/tmp/features.npy', X)
np.save('/tmp/labels.npy', y)
X_loaded = np.load('/tmp/features.npy')
y_loaded = np.load('/tmp/labels.npy')
print('Single array save/load:')
print(f'  Shapes match: {X.shape == X_loaded.shape}')
print(f'  Data matches: {np.allclose(X, X_loaded)}')
print(f'  File size   : {os.path.getsize("/tmp/features.npy") / 1024:.1f} KB')

# Save multiple arrays in one file
np.savez('/tmp/dataset.npz',
         X_train=X[:800], y_train=y[:800],
         X_test=X[800:],  y_test=y[800:])

data = np.load('/tmp/dataset.npz')
print('\nMulti-array .npz file:')
print('  Keys:', list(data.keys()))
print(f'  X_train: {data["X_train"].shape},  X_test: {data["X_test"].shape}')

# Compressed
np.savez_compressed('/tmp/dataset_c.npz',
                    X_train=X[:800], y_train=y[:800],
                    X_test=X[800:],  y_test=y[800:])
orig = os.path.getsize('/tmp/dataset.npz') / 1024
comp = os.path.getsize('/tmp/dataset_c.npz') / 1024
print(f'\nUncompressed: {orig:.1f} KB,  Compressed: {comp:.1f} KB  ({comp/orig*100:.0f}%)')
```

```python
import numpy as np
import io

# Reading CSV-like text files with genfromtxt
# We build the CSV content as a regular Python string
csv_content = (
    'county,gdp_billion,population_million,literacy_pct\n'
    'Nairobi,3.5,4.7,92.1\n'
    'Mombasa,1.2,1.4,82.5\n'
    'Kisumu,0.6,1.1,78.3\n'
    'Nakuru,0.8,2.2,80.1\n'
    'Eldoret,0.5,0.9,75.4\n'
)

data = np.genfromtxt(
    io.StringIO(csv_content),
    delimiter=',',
    names=True,
    dtype=None,
    encoding='utf8'
)

print('Loaded county data:')
print('Columns:', data.dtype.names)
print('\nCounties:', data['county'])
print('GDP     :', data['gdp_billion'])
print('Literacy:', data['literacy_pct'])
highest = data['county'][data['gdp_billion'].argmax()].decode()
print(f'\nHighest GDP county: {highest}')
```
