# Week 4: Reading and Writing Data with Pandas

## Learning Objectives
- Read CSV, Excel, and JSON files
- Write data to various formats
- Handle common file reading issues

```python
import pandas as pd
import numpy as np
import io
```

## 1. Reading CSV Files

```python
# Simulated CSV data
csv_data = """date,product,quantity,price,region
2024-01-01,Laptop,2,1200,North
2024-01-02,Mouse,5,25,South
2024-01-02,Keyboard,3,75,East
2024-01-03,Monitor,1,300,West
2024-01-03,Laptop,1,1200,South
2024-01-04,Mouse,8,25,North
2024-01-04,Keyboard,2,75,West
2024-01-05,Monitor,3,300,East
2024-01-05,Laptop,2,1200,North"""

df = pd.read_csv(io.StringIO(csv_data), parse_dates=['date'])
print(df.head())
print('\nDtypes:\n', df.dtypes)
```

```python
# read_csv parameters
# pd.read_csv('data.csv',
#     sep=',',           # delimiter
#     header=0,          # row for column names
#     index_col=None,    # column to use as index
#     usecols=['col1'],  # subset of columns
#     nrows=100,         # read only first 100 rows
#     skiprows=[1,2],    # skip rows
#     na_values=['NA'],  # additional NA strings
#     dtype={'col': str}# specify dtypes
# )
print('Common read_csv parameters demonstrated above (commented).')
```

## 2. Writing Data

```python
# Write to CSV (to string for demo)
csv_out = df.to_csv(index=False)
print('CSV output (first 200 chars):')
print(csv_out[:200])

# Write to JSON
json_out = df.head(3).to_json(orient='records', indent=2)
print('\nJSON output:\n', json_out)
```

## 3. Data Exploration After Loading

```python
print('Shape:', df.shape)
print('\nFirst 3 rows:\n', df.head(3))
print('\nLast 2 rows:\n', df.tail(2))
print('\nColumn info:')
print(df.info())
print('\nStatistics:\n', df.describe())
```

## 4. Grouping and Aggregating

```python
# Revenue per region
df['revenue'] = df['quantity'] * df['price']
regional = df.groupby('region')['revenue'].sum().sort_values(ascending=False)
print('Revenue by region:\n', regional)

# Multiple aggregations
summary = df.groupby('product').agg(
    total_qty=('quantity', 'sum'),
    total_rev=('revenue', 'sum'),
    num_orders=('date', 'count')
).reset_index()
print('\nProduct summary:\n', summary)
```

## 5. Pivot Tables

```python
pivot = df.pivot_table(
    values='revenue',
    index='product',
    columns='region',
    aggfunc='sum',
    fill_value=0
)
print('Revenue pivot table:\n', pivot)
```
