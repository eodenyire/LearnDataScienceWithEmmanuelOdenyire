```python
import pandas as pd
import numpy as np
np.random.seed(42)

# Create a messy real-world dataset
n = 200
df = pd.DataFrame({
    'age'        : np.where(np.random.rand(n) < 0.05, np.nan, np.random.randint(18, 65, n)),
    'income'     : np.where(np.random.rand(n) < 0.15, np.nan, np.random.exponential(80_000, n)),
    'education'  : np.where(np.random.rand(n) < 0.08, np.nan, np.random.choice(['Primary','Secondary','University','Masters'], n)),
    'county'     : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru'], n),
    'credit_score': np.where(np.random.rand(n) < 0.20, np.nan, np.random.randint(300, 850, n)),
    'loan_amount': np.where(np.random.rand(n) < 0.30, np.nan, np.random.exponential(200_000, n)),
})

# Missing data audit
print('=== Missing Data Audit ===')
missing = df.isnull().sum()
missing_pct = (df.isnull().mean() * 100).round(1)
for col in df.columns:
    print(f'  {col:15}: {missing[col]:4d} missing ({missing_pct[col]:5.1f}%)')

print(f'\nTotal missing cells: {df.isnull().sum().sum()}')
print(f'Rows with ANY missing: {df.isnull().any(axis=1).sum()}')
print(f'Rows with ALL present: {df.notnull().all(axis=1).sum()}')
```

```python
import pandas as pd
import numpy as np
np.random.seed(42)

n = 200
df = pd.DataFrame({
    'age'   : np.where(np.random.rand(n) < 0.05, np.nan, np.random.randint(18,65,n).astype(float)),
    'income': np.where(np.random.rand(n) < 0.15, np.nan, np.random.exponential(80_000,n)),
    'education': np.where(np.random.rand(n) < 0.08, np.nan, np.random.choice(['Primary','Secondary','University','Masters'],n)),
    'credit_score': np.where(np.random.rand(n) < 0.20, np.nan, np.random.randint(300,850,n).astype(float)),
    'loan_amount': np.where(np.random.rand(n) < 0.30, np.nan, np.random.exponential(200_000,n)),
})

# Strategy 1: Drop
df_dropna = df.dropna()  # all rows with any NaN
df_dropna_thresh = df.dropna(thresh=4)  # keep rows with at least 4 non-NaN
print(f'Original : {len(df)} rows')
print(f'dropna() : {len(df_dropna)} rows (dropped {len(df)-len(df_dropna)})')
print(f'thresh=4 : {len(df_dropna_thresh)} rows (dropped {len(df)-len(df_dropna_thresh)})')

# Strategy 2: Imputation
df_imp = df.copy()

# Numerical: mean/median
df_imp['age'].fillna(df_imp['age'].median(), inplace=True)
df_imp['income'].fillna(df_imp['income'].median(), inplace=True)
df_imp['credit_score'].fillna(df_imp['credit_score'].mean(), inplace=True)
df_imp['loan_amount'].fillna(df_imp['loan_amount'].median(), inplace=True)

# Categorical: mode
mode_edu = df_imp['education'].mode()[0]
df_imp['education'].fillna(mode_edu, inplace=True)

print(f'\nAfter imputation — missing values: {df_imp.isnull().sum().sum()}')
print('Imputed age median:', df_imp['age'].median().round(1))
print('Imputed edu mode  :', mode_edu)
```

```python
import pandas as pd
import numpy as np

# Forward fill and backward fill — ideal for time series
dates = pd.date_range('2024-01-01', periods=10, freq='D')
prices = pd.Series([100.0, np.nan, np.nan, 105.0, np.nan, 108.0, np.nan, 110.0, np.nan, 112.0], index=dates)
print('Original prices (with NaN):')
print(prices)
print('\nForward fill (use last known price):')
print(prices.ffill())
print('\nBackward fill:')
print(prices.bfill())
print('\nLinear interpolation:')
print(prices.interpolate())
```

## Summary

- Always audit missing data before doing anything else
- Choose imputation strategy based on data type and % missing
- Numerical: mean (normal distribution), median (skewed/outliers)
- Categorical: mode
- Time series: forward/backward fill or interpolation
- NEVER blindly drop rows — you may lose patterns!
