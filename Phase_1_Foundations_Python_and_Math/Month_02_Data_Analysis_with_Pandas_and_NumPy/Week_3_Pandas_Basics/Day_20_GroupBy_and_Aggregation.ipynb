```python
import pandas as pd
import numpy as np
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'market': np.random.choice(['Gikomba','Ngara','City Market','Toi','Marikiti'],n),
    'product': np.random.choice(['Tomatoes','Onions','Potatoes','Maize','Beans'],n),
    'quantity_kg': np.random.randint(10,500,n),
    'price_kes': np.random.randint(30,200,n),
    'seller_gender': np.random.choice(['M','F'],n),
})
df['revenue'] = df['quantity_kg'] * df['price_kes']

# Basic groupby
print('Total revenue by market:')
print(df.groupby('market')['revenue'].sum().sort_values(ascending=False))

# Multiple aggregations
print('\nRevenue stats by market:')
aggs = df.groupby('market')['revenue'].agg(['sum','mean','count','std'])
print(aggs.round(0).astype(int))

# Group by multiple columns
print('\nMean price by market and product:')
print(df.groupby(['market','product'])['price_kes'].mean().round(1).head(10))

# Named aggregations
print('\nCustom aggregations:')
result = df.groupby('market').agg(
    total_revenue=('revenue', 'sum'),
    avg_price    =('price_kes', 'mean'),
    transactions =('revenue', 'count'),
    top_qty      =('quantity_kg', 'max')
).round(1)
print(result)
```

```python
import pandas as pd
import numpy as np
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'market': np.random.choice(['Gikomba','Ngara','City Market','Toi','Marikiti'],n),
    'product': np.random.choice(['Tomatoes','Onions','Potatoes','Maize','Beans'],n),
    'quantity_kg': np.random.randint(10,500,n),
    'price_kes': np.random.randint(30,200,n),
    'seller_gender': np.random.choice(['M','F'],n),
})
df['revenue'] = df['quantity_kg'] * df['price_kes']

# transform() — adds group stats back to original DataFrame
df['market_avg_revenue'] = df.groupby('market')['revenue'].transform('mean')
df['above_market_avg']   = df['revenue'] > df['market_avg_revenue']
print('Above-market-average transactions per market:')
print(df.groupby('market')['above_market_avg'].sum())

# Pivot table
print('\nPivot table: mean price by market (rows) and product (cols)')
pivot = df.pivot_table(
    values='price_kes',
    index='market',
    columns='product',
    aggfunc='mean',
    fill_value=0
)
print(pivot.round(1))
```

## Summary
