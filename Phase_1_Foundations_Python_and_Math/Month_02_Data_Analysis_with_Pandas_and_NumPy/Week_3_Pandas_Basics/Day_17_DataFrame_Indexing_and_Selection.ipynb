```python
import pandas as pd
import numpy as np
np.random.seed(42)

# Create a rich employee dataset
np.random.seed(0)
n = 20
df = pd.DataFrame({
    'name'      : [f'Employee_{i:02d}' for i in range(1, n+1)],
    'dept'      : np.random.choice(['Engineering','Finance','Marketing','Sales'], n),
    'county'    : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n),
    'salary_kes': np.random.randint(60_000, 350_000, n),
    'age'       : np.random.randint(22, 58, n),
    'yrs_exp'   : np.random.randint(0, 20, n),
    'rating'    : np.random.choice([1,2,3,4,5], n),
})
df = df.set_index('name')
print('Employee DataFrame:')
print(df.head(8))
print(f'\nShape: {df.shape}')
```

```python
import pandas as pd
import numpy as np

# Setup (same as above)
np.random.seed(0)
n = 20
df = pd.DataFrame({
    'name'      : [f'Employee_{i:02d}' for i in range(1, n+1)],
    'dept'      : np.random.choice(['Engineering','Finance','Marketing','Sales'], n),
    'county'    : np.random.choice(['Nairobi','Mombasa','Kisumu','Nakuru','Eldoret'], n),
    'salary_kes': np.random.randint(60_000, 350_000, n),
    'age'       : np.random.randint(22, 58, n),
    'yrs_exp'   : np.random.randint(0, 20, n),
    'rating'    : np.random.choice([1,2,3,4,5], n),
}).set_index('name')

# .loc — label based
print('.loc[row_label, col]:')
print(df.loc['Employee_01', 'salary_kes'])
print('\n.loc slicing:')
print(df.loc['Employee_01':'Employee_03', ['dept', 'salary_kes']])

# .iloc — position based
print('\n.iloc[0:3, 0:2]:')
print(df.iloc[0:3, 0:2])

# Boolean mask
print('\nHigh earners in Engineering:')
mask = (df['salary_kes'] > 200_000) & (df['dept'] == 'Engineering')
print(df[mask][['dept','salary_kes','county']])

# .query() — readable, SQL-like
print('\n.query() — Nairobi employees under 35:')
result = df.query('county == "Nairobi" and age < 35')
print(result[['dept','age','salary_kes']])
```

```python
import pandas as pd
import numpy as np

# MultiIndex — hierarchical indexing
np.random.seed(1)

# Sales data: county × year × quarter
arrays = [
    ['Nairobi']*4 + ['Mombasa']*4 + ['Kisumu']*4,
    ['Q1','Q2','Q3','Q4'] * 3
]
index = pd.MultiIndex.from_arrays(arrays, names=['County', 'Quarter'])
sales = pd.Series(
    np.random.randint(500_000, 5_000_000, 12),
    index=index,
    name='Sales_KES'
)
print('Multi-Index Series:')
print(sales)
print('\nNairobi only:')
print(sales.loc['Nairobi'])
print('\nAll Q1 sales:')
print(sales.loc[:, 'Q1'])
```
