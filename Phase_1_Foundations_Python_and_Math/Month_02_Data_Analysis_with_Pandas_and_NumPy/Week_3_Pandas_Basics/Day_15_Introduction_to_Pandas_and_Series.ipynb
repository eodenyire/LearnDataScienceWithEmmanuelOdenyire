## Pandas Series

A Series is a **labelled one-dimensional array**. Think of it as a column in a spreadsheet.

| Feature | NumPy array | Pandas Series |
|---------|------------|----------------|
| Index | 0, 1, 2 ... | Any hashable label |
| Data types | Homogeneous | Homogeneous per series |
| Missing values | NaN (float only) | pd.NA / NaN (all types) |
| Operations | Positional | Position OR label |

```python
import pandas as pd
import numpy as np

# Create a Series from a list
temps = pd.Series([29.5, 27.3, 31.0, 26.8, 30.1],
                  index=['Mon','Tue','Wed','Thu','Fri'],
                  name='Nairobi_Temp_C')

print('Nairobi temperatures:')
print(temps)
print(f'\ndtype: {temps.dtype}')
print(f'index: {temps.index.tolist()}')
print(f'values: {temps.values}')
print(f'name: {temps.name}')

# Indexing
print(f'\nWednesday: {temps["Wed"]}')
print(f'Mon-Wed   : {temps["Mon":"Wed"].values}')
print(f'Positional [0]: {temps.iloc[0]}')

# From a dict
pop = pd.Series({'Kenya': 54.0, 'Nigeria': 223.8, 'Ghana': 33.5, 'Ethiopia': 126.5})
print('\nAfrican populations (millions):')
print(pop.sort_values(ascending=False))
```

```python
import pandas as pd
import numpy as np

# Series arithmetic — automatic index alignment!
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([100, 200, 300, 400], index=['a', 'b', 'c', 'd'])

print('s1:', s1.to_dict())
print('s2:', s2.to_dict())
print('s1 + s2:', (s1 + s2).to_dict())  # NaN for unaligned

# Fill NaN during arithmetic
print('fill NaN:', s1.add(s2, fill_value=0).to_dict())

# Boolean masking
scores = pd.Series([45, 78, 92, 55, 88, 71, 60, 95],
                   index=['Alice','Bob','Charlie','Diana','Eve','Frank','Grace','Henry'])
print('\nTop scorers (>80):')
print(scores[scores > 80])
print('\nBottom 3:')
print(scores.nsmallest(3))
```

```python
import pandas as pd

# String Series operations with .str accessor
cities = pd.Series(['Nairobi', 'mombasa', 'KISUMU', 'nakuru', 'Eldoret'])

print('Original :', cities.tolist())
print('Upper    :', cities.str.upper().tolist())
print('Lower    :', cities.str.lower().tolist())
print('Title    :', cities.str.title().tolist())
print('Length   :', cities.str.len().tolist())
print('Contains N:', cities.str.contains('N', case=False).tolist())
print('Replace  :', cities.str.replace('a', '@').tolist())

# Splitting
full_names = pd.Series(['Alice Njeri', 'Bob Otieno', 'Carol Wanjiku', 'David Kamau'])
print('\nFirst names:', full_names.str.split().str[0].tolist())
print('Last names :', full_names.str.split().str[1].tolist())
```
