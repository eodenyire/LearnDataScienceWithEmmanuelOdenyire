```python
import pandas as pd
import numpy as np
np.random.seed(42)

# Creating DataFrames

# From a dictionary
df_dict = pd.DataFrame({
    'name'  : ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'age'   : [28, 35, 31, 42, 26],
    'salary': [95000, 120000, 85000, 200000, 78000],
    'county': ['Nairobi', 'Kisumu', 'Nakuru', 'Nairobi', 'Mombasa'],
})
print('From dict:')
print(df_dict)
print(f'\nShape: {df_dict.shape}  ({df_dict.shape[0]} rows, {df_dict.shape[1]} cols)')

# From NumPy array
X = np.random.randn(4, 3)
df_np = pd.DataFrame(X, columns=['Feature_A', 'Feature_B', 'Feature_C'])
print('\nFrom NumPy array:')
print(df_np.round(3))
```

```python
import pandas as pd
import numpy as np
import io

# Reading CSV (we simulate a CSV in memory)
csv_data = (
    'student_id,name,maths,science,english,county\n'
    '1,Alice,85,92,78,Nairobi\n'
    '2,Bob,73,88,91,Kisumu\n'
    '3,Carol,90,67,84,Nakuru\n'
    '4,David,55,78,65,Mombasa\n'
    '5,Eve,88,95,72,Nairobi\n'
    '6,Frank,65,71,88,Eldoret\n'
    '7,Grace,92,83,79,Nairobi\n'
)

df = pd.read_csv(io.StringIO(csv_data))

print('=== Student Data ===')
print(df)
print('\n--- .info() ---')
print(df.info())
print('\n--- .describe() ---')
print(df.describe().round(1))
```

```python
import pandas as pd
import numpy as np
import io

csv = (
    'student_id,name,maths,science,english,county\n'
    '1,Alice,85,92,78,Nairobi\n2,Bob,73,88,91,Kisumu\n'
    '3,Carol,90,67,84,Nakuru\n4,David,55,78,65,Mombasa\n'
    '5,Eve,88,95,72,Nairobi\n6,Frank,65,71,88,Eldoret\n'
    '7,Grace,92,83,79,Nairobi\n'
)
df = pd.read_csv(io.StringIO(csv))

# Column selection
print('Maths column:\n', df['maths'].tolist())
print('Multiple columns:\n', df[['name', 'maths', 'science']].head(3))

# Row selection
print('\nRow 0 (label):', df.loc[0, ['name', 'maths']].to_dict())
print('Row 1 (position):', df.iloc[1][['name', 'maths']].to_dict())
print('\nRows 2-4:\n', df.iloc[2:5][['name', 'county']])

# Add new column
df['average'] = df[['maths', 'science', 'english']].mean(axis=1).round(1)
df['grade']   = pd.cut(df['average'], bins=[0,60,70,80,101],
                       labels=['C','B','A-','A'])
print('\nWith average and grade:')
print(df[['name', 'average', 'grade']])

# Drop column
df_no_id = df.drop('student_id', axis=1)
print('\nColumns after drop:', df_no_id.columns.tolist())
```
