# Week 4: Interactive Visualizations with Plotly

## Learning Objectives
- Create interactive charts with Plotly Express
- Build charts using the Plotly Graph Objects API
- Add hover information and sliders
- Create animated visualizations

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

print('Plotly imported successfully')
```

## 1. Plotly Express — Quick Charts

```python
# Scatter plot
iris = px.data.iris()
fig = px.scatter(iris, x='sepal_width', y='sepal_length',
                 color='species', size='petal_length',
                 hover_data=['petal_width'],
                 title='Iris Dataset — Interactive Scatter')
fig.show()
```

```python
# Bar chart
gapminder = px.data.gapminder().query('year == 2007').nlargest(10, 'gdpPercap')
fig = px.bar(gapminder, x='country', y='gdpPercap', color='continent',
             title='Top 10 Countries by GDP per Capita (2007)',
             text_auto='.0f')
fig.show()
```

## 2. Line and Area Charts

```python
gdp_data = px.data.gapminder().query('continent == "Americas"')
fig = px.line(gdp_data, x='year', y='lifeExp', color='country',
              title='Life Expectancy in the Americas',
              labels={'lifeExp': 'Life Expectancy', 'year': 'Year'})
fig.show()
```

## 3. Graph Objects — Fine Control

```python
rng = np.random.default_rng(42)
x = np.arange(1, 13)
month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
product_a = rng.integers(80, 200, 12)
product_b = rng.integers(60, 180, 12)

fig = go.Figure()
fig.add_trace(go.Bar(x=month_names, y=product_a, name='Product A', marker_color='steelblue'))
fig.add_trace(go.Bar(x=month_names, y=product_b, name='Product B', marker_color='coral'))
fig.update_layout(barmode='group', title='Monthly Sales by Product',
                  xaxis_title='Month', yaxis_title='Units Sold',
                  template='plotly_white')
fig.show()
```

## 4. Animated Charts

```python
gapminder = px.data.gapminder()
fig = px.scatter(gapminder,
                 x='gdpPercap', y='lifeExp',
                 size='pop', color='continent',
                 hover_name='country',
                 animation_frame='year',
                 animation_group='country',
                 log_x=True, size_max=60,
                 range_x=[100, 100000], range_y=[25, 90],
                 title='Gapminder Animated: GDP vs Life Expectancy')
fig.show()
```
