# Regression From Scratch with NumPy
## Learning Objectives
- Implement linear regression using gradient descent
- Extend to polynomial regression
- Evaluate with MSE and R² metrics


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = 2.5 * X + 1.0 + np.random.normal(0, 1.5, 100)
print(f'X shape: {X.shape}, y shape: {y.shape}')
```

```python
# Linear Regression from scratch
class LinearRegressionScratch:
    def __init__(self, lr=0.01, epochs=1000):
        self.lr, self.epochs = lr, epochs
        self.w, self.b = 0.0, 0.0

    def fit(self, X, y):
        n = len(X)
        for _ in range(self.epochs):
            y_pred = self.w * X + self.b
            self.w -= self.lr * (-2/n) * np.dot(X, (y - y_pred))
            self.b -= self.lr * (-2/n) * np.sum(y - y_pred)

    def predict(self, X):
        return self.w * X + self.b

model = LinearRegressionScratch(lr=0.01, epochs=2000)
model.fit(X, y)
print(f'Learned w={model.w:.3f}, b={model.b:.3f}')
print(f'True    w=2.500,    b=1.000')
```

```python
# Evaluation metrics
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def r2(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - y_true.mean())**2)
    return 1 - ss_res/ss_tot

y_pred = model.predict(X)
print(f'MSE: {mse(y, y_pred):.4f}')
print(f'R²:  {r2(y, y_pred):.4f}')
```

```python
# Polynomial regression using numpy
from numpy.polynomial import polynomial as P

degrees = [1, 2, 3, 5]
for deg in degrees:
    coeffs = np.polyfit(X, y, deg)
    y_fit = np.polyval(coeffs, X)
    print(f'Degree {deg}: R² = {r2(y, y_fit):.4f}')
```
