# Week 3: ARIMA and SARIMA Models

## Learning Objectives
- Understand AR, MA, and ARIMA components
- Use ACF and PACF to determine order parameters
- Fit ARIMA and SARIMA models
- Evaluate and forecast with ARIMA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. ACF and PACF for Order Selection

```python
months = pd.date_range('1949-01', periods=144, freq='ME')
passengers = np.array([
    112,118,132,129,121,135,148,148,136,119,104,118,
    115,126,141,135,125,149,170,170,158,133,114,140,
    145,150,178,163,172,178,199,199,184,162,146,166,
    171,180,193,181,183,218,230,242,209,191,172,194,
    196,196,236,235,229,243,264,272,237,211,180,201,
    204,188,235,227,234,264,302,293,259,229,203,229,
    242,233,267,269,270,315,364,347,312,274,237,278,
    284,277,317,313,318,374,413,405,355,306,271,306,
    315,301,356,348,355,422,465,467,404,347,305,336,
    340,318,362,348,363,435,491,505,404,359,310,337,
    360,342,406,396,420,472,548,559,463,407,362,405,
    417,391,419,461,472,535,622,606,508,461,390,432
])
ts = pd.Series(passengers, index=months, name='passengers')

log_diff = np.log(ts).diff(1).diff(12).dropna()

fig, axes = plt.subplots(2, 1, figsize=(12, 7))
plot_acf(log_diff, lags=40, ax=axes[0])
plot_pacf(log_diff, lags=40, ax=axes[1])
plt.tight_layout(); plt.show()
```

## 2. Fitting SARIMA

```python
train = ts[:'1959']
test  = ts['1960':]

# SARIMA(p=1, d=1, q=1)(P=1, D=1, Q=1, S=12)
model = SARIMAX(np.log(train), order=(1,1,1), seasonal_order=(1,1,1,12),
                enforce_stationarity=False, enforce_invertibility=False)
result = model.fit(disp=False)
print(result.summary().tables[1])
```

```python
forecast_log = result.forecast(steps=len(test))
forecast = np.exp(forecast_log)

rmse = np.sqrt(mean_squared_error(test, forecast))
print(f'RMSE: {rmse:.2f}')

plt.figure(figsize=(12, 5))
train.plot(label='Train')
test.plot(label='Actual')
forecast.plot(label='Forecast', linestyle='--')
plt.title('SARIMA Forecast — Airline Passengers')
plt.legend(); plt.tight_layout(); plt.show()
```

## 3. Residual Diagnostics

```python
result.plot_diagnostics(figsize=(12, 8))
plt.tight_layout(); plt.show()
```
