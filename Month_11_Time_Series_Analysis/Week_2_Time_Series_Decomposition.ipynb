# Week 2: Time Series Decomposition

## Learning Objectives
- Decompose time series into trend, seasonal, and residual components
- Compare additive and multiplicative decomposition
- Apply STL decomposition
- Use differencing to achieve stationarity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.stattools import adfuller

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Classical Decomposition

```python
# Airline passenger dataset (classic time series)
months = pd.date_range('1949-01', periods=144, freq='ME')
passengers = np.array([
    112,118,132,129,121,135,148,148,136,119,104,118,
    115,126,141,135,125,149,170,170,158,133,114,140,
    145,150,178,163,172,178,199,199,184,162,146,166,
    171,180,193,181,183,218,230,242,209,191,172,194,
    196,196,236,235,229,243,264,272,237,211,180,201,
    204,188,235,227,234,264,302,293,259,229,203,229,
    242,233,267,269,270,315,364,347,312,274,237,278,
    284,277,317,313,318,374,413,405,355,306,271,306,
    315,301,356,348,355,422,465,467,404,347,305,336,
    340,318,362,348,363,435,491,505,404,359,310,337,
    360,342,406,396,420,472,548,559,463,407,362,405,
    417,391,419,461,472,535,622,606,508,461,390,432
])
ts = pd.Series(passengers, index=months, name='passengers')

result = seasonal_decompose(ts, model='multiplicative', period=12)

fig = result.plot()
fig.set_size_inches(12, 9)
plt.suptitle('Multiplicative Decomposition: Airline Passengers', y=1.01)
plt.tight_layout(); plt.show()
```

## 2. STL Decomposition

```python
stl = STL(ts, period=12, robust=True)
stl_result = stl.fit()

fig, axes = plt.subplots(4, 1, figsize=(12, 10))
stl_result.observed.plot(ax=axes[0]);  axes[0].set_ylabel('Observed')
stl_result.trend.plot(ax=axes[1]);     axes[1].set_ylabel('Trend')
stl_result.seasonal.plot(ax=axes[2]);  axes[2].set_ylabel('Seasonal')
stl_result.resid.plot(ax=axes[3]);     axes[3].set_ylabel('Residual')
plt.suptitle('STL Decomposition', y=1.01)
plt.tight_layout(); plt.show()
```

## 3. Differencing for Stationarity

```python
def adf_test(series, name):
    r = adfuller(series.dropna())
    print(f'{name}: ADF={r[0]:.4f}, p={r[1]:.4f}, stationary={r[1]<0.05}')

log_ts = np.log(ts)
adf_test(log_ts, 'Log passengers')
adf_test(log_ts.diff().dropna(), 'First diff of log')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
log_ts.plot(ax=axes[0], title='Log Passengers')
log_ts.diff().dropna().plot(ax=axes[1], title='First Difference of Log')
plt.tight_layout(); plt.show()
```
