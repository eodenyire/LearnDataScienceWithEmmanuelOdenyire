# Week 1: Understanding Time Series Data

## Learning Objectives
- Understand components of time series (trend, seasonality, noise)
- Resample and manipulate time series with Pandas
- Compute rolling statistics
- Test for stationarity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Time Series Components

```python
# Synthetic time series with trend + seasonality + noise
dates = pd.date_range(start='2018-01-01', periods=365*4, freq='D')
t = np.arange(len(dates))
trend = 0.05 * t
seasonality = 10 * np.sin(2 * np.pi * t / 365)
noise = rng.normal(0, 2, len(t))
ts = pd.Series(trend + seasonality + noise, index=dates, name='value')

plt.figure(figsize=(12, 4))
ts.plot()
plt.title('Synthetic Time Series: Trend + Seasonality + Noise')
plt.ylabel('Value'); plt.tight_layout(); plt.show()
```

## 2. Rolling Statistics

```python
window = 30
rolling_mean = ts.rolling(window=window).mean()
rolling_std  = ts.rolling(window=window).std()

plt.figure(figsize=(12, 5))
plt.plot(ts, alpha=0.5, label='Original')
plt.plot(rolling_mean, label=f'Rolling Mean ({window}d)', linewidth=2)
plt.fill_between(ts.index,
                  rolling_mean - rolling_std,
                  rolling_mean + rolling_std, alpha=0.2, label='±1 Std')
plt.title('Time Series with Rolling Statistics')
plt.legend(); plt.tight_layout(); plt.show()
```

## 3. Resampling

```python
monthly_mean = ts.resample('ME').mean()
monthly_std  = ts.resample('ME').std()

fig, axes = plt.subplots(2, 1, figsize=(12, 7))
ts.plot(ax=axes[0], alpha=0.6, title='Daily Data')
monthly_mean.plot(ax=axes[1], marker='o', title='Monthly Mean')
plt.tight_layout(); plt.show()
```

## 4. Stationarity Test (ADF)

```python
def adf_test(series, name=''):
    result = adfuller(series.dropna())
    print(f'{name}: ADF={result[0]:.4f}, p-value={result[1]:.4f}, Stationary={result[1]<0.05}')

adf_test(ts, 'Original TS')
adf_test(ts.diff().dropna(), 'First Difference')
```
