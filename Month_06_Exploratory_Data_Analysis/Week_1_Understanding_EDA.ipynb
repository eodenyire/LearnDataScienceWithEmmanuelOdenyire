```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
```

## 1. What is EDA?

Exploratory Data Analysis (EDA) is the process of:
1. **Understanding** the data structure, types, and scale
2. **Cleaning** obvious issues (missing values, duplicates, types)
3. **Summarising** distributions, central tendency, and spread
4. **Visualising** relationships and patterns
5. **Forming hypotheses** to test with statistical models

## 2. A Systematic EDA Framework

```python
# Step 1: Load and inspect
rng = np.random.default_rng(42)
n = 300
df = pd.DataFrame({
    'age':          rng.integers(18, 65, n),
    'education':    rng.choice(['HS', 'Bachelor', 'Master', 'PhD'], n,
                               p=[0.3, 0.4, 0.2, 0.1]),
    'hours_work':   rng.normal(40, 8, n).clip(20, 80),
    'income':       rng.lognormal(10.5, 0.5, n),
    'satisfaction': rng.integers(1, 11, n)
})
df['high_earner'] = (df['income'] > df['income'].quantile(0.75)).astype(int)

print('Shape:', df.shape)
print('\nDtypes:\n', df.dtypes)
print('\nHead:\n', df.head(3))
```

```python
# Step 2: Summary statistics
print(df.describe().round(2))
print('\nMissing values:\n', df.isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
```

```python
# Step 3: Distribution overview
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()
num_cols = ['age', 'hours_work', 'income', 'satisfaction']
for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
for j, col in enumerate(['education', 'high_earner']):
    df[col].value_counts().plot(kind='bar', ax=axes[4+j], rot=0)
    axes[4+j].set_title(f'Count of {col}')
plt.suptitle('Variable Distributions Overview', fontsize=14)
plt.tight_layout(); plt.show()
```

## 3. Key EDA Questions

1. What is the data shape and types?
2. Are there missing values or duplicates?
3. What are the distributions (normal, skewed, bimodal)?
4. Are there outliers?
5. How are variables related to each other?
6. Are there any data quality issues?
7. What patterns or clusters exist?
