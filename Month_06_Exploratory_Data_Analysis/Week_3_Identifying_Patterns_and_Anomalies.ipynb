# Week 3: Identifying Patterns and Anomalies

## Learning Objectives
- Detect outliers using statistical methods
- Visualise clusters and subgroups
- Find temporal and spatial patterns
- Use isolation forest for anomaly detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Statistical Outlier Detection

```python
n = 200
normal = rng.normal(50, 10, n)
outliers = rng.choice([-30, 120, 150, -50, 200], 10)
data = np.append(normal, outliers)

z_scores = np.abs((data - data.mean()) / data.std())
iqr = np.percentile(data, 75) - np.percentile(data, 25)
iqr_mask = (data < np.percentile(data, 25) - 1.5*iqr) | (data > np.percentile(data, 75) + 1.5*iqr)

print(f'Z-score outliers (|z|>3): {(z_scores>3).sum()}')
print(f'IQR outliers: {iqr_mask.sum()}')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(range(len(data)), data, c=['red' if z>3 else 'steelblue' for z in z_scores], alpha=0.7)
axes[0].set_title('Z-Score Outliers (red)')
axes[1].boxplot(data)
axes[1].set_title('Boxplot')
plt.tight_layout(); plt.show()
```

## 2. Isolation Forest

```python
X = rng.normal(0, 1, (300, 2))
X_outliers = rng.uniform(-6, 6, (20, 2))
X_all = np.vstack([X, X_outliers])

iso = IsolationForest(contamination=0.05, random_state=42)
labels = iso.fit_predict(X_all)  # -1 = outlier, 1 = normal

colors = ['red' if l == -1 else 'steelblue' for l in labels]
plt.figure(figsize=(7, 5))
plt.scatter(X_all[:, 0], X_all[:, 1], c=colors, alpha=0.7)
plt.title(f'Isolation Forest: {(labels==-1).sum()} outliers detected')
plt.tight_layout(); plt.show()
```

## 3. Temporal Patterns

```python
dates = pd.date_range('2024-01-01', periods=180, freq='D')
values = 100 + 20 * np.sin(2*np.pi * np.arange(180) / 30) + rng.normal(0, 5, 180)
values[60:65] += 80  # Inject anomaly

df_ts = pd.Series(values, index=dates)
rolling_mean = df_ts.rolling(7).mean()
rolling_std = df_ts.rolling(7).std()

plt.figure(figsize=(12, 5))
plt.plot(df_ts, label='Observed', alpha=0.6)
plt.plot(rolling_mean, label='7-day MA', linewidth=2)
plt.fill_between(df_ts.index, rolling_mean - 2*rolling_std, rolling_mean + 2*rolling_std,
                  alpha=0.2, label='±2σ band')
plt.title('Time Series with Anomaly Band')
plt.legend(); plt.tight_layout(); plt.show()
```
