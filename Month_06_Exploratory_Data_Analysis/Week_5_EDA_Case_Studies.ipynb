```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_theme(style='whitegrid')
pd.set_option('display.max_columns', 20)
```

## Case Study 1: Titanic Survival Analysis

```python
titanic = sns.load_dataset('titanic')
print('Shape:', titanic.shape)
print('\nInfo:')
print(titanic.dtypes)
print('\nMissing values:\n', titanic.isnull().sum()[titanic.isnull().sum() > 0])
```

```python
print('Survival rate:', titanic['survived'].mean().round(3))
print('\nSurvival by class:')
print(titanic.groupby('pclass')['survived'].mean().round(3))
print('\nSurvival by sex:')
print(titanic.groupby('sex')['survived'].mean().round(3))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))

titanic['survived'].value_counts().plot(kind='bar', ax=axes[0,0], rot=0, color=['coral','steelblue'])
axes[0,0].set_title('Survival Count'); axes[0,0].set_xticklabels(['Died','Survived'])

sns.barplot(data=titanic, x='pclass', y='survived', ax=axes[0,1], errorbar=None)
axes[0,1].set_title('Survival Rate by Class'); axes[0,1].set_ylabel('Rate')

sns.barplot(data=titanic, x='sex', y='survived', ax=axes[0,2], errorbar=None)
axes[0,2].set_title('Survival Rate by Sex')

sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30, ax=axes[1,0])
axes[1,0].set_title('Age Distribution by Survival')

sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1,1], palette='Set2')
axes[1,1].set_title('Fare Distribution by Class')

corr_num = titanic[['survived','age','fare','pclass','sibsp','parch']].corr()
sns.heatmap(corr_num, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1,2], linewidths=0.5)
axes[1,2].set_title('Numeric Correlations')

plt.suptitle('Titanic EDA Dashboard', fontsize=16, fontweight='bold')
plt.tight_layout(); plt.show()
```

## Case Study 2: Tips Dataset

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill'] * 100

print('Summary statistics:')
print(tips[['total_bill','tip','tip_pct']].describe().round(2))

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
sns.histplot(tips['tip_pct'], kde=True, ax=axes[0])
axes[0].set_title('Tip % Distribution')
axes[0].axvline(tips['tip_pct'].mean(), color='red', linestyle='--', label=f"Mean: {tips['tip_pct'].mean():.1f}%")
axes[0].legend()

sns.boxplot(data=tips, x='day', y='tip_pct', ax=axes[1], order=['Thur','Fri','Sat','Sun'])
axes[1].set_title('Tip % by Day')

sns.scatterplot(data=tips, x='total_bill', y='tip_pct', hue='time', alpha=0.7, ax=axes[2])
axes[2].set_title('Bill vs Tip %')
plt.tight_layout(); plt.show()
```
