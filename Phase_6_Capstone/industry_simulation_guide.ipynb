# Industry Simulation Guide
## Learning Objectives
- Experience a realistic data science project workflow
- Practice Agile/Scrum for data science teams
- Prepare for technical interviews and whiteboard challenges


```python
# Industry roles and responsibilities
ds_roles = {
    'Data Analyst': ['SQL', 'Excel', 'Tableau', 'Basic stats', 'Reporting'],
    'Data Scientist': ['ML', 'Python', 'Statistics', 'A/B testing', 'Modeling'],
    'ML Engineer': ['MLOps', 'Docker', 'APIs', 'Model deployment', 'Monitoring'],
    'Data Engineer': ['Pipelines', 'Spark', 'Kafka', 'Airflow', 'Data warehouse'],
    'AI Researcher': ['Deep learning', 'Papers', 'Novel architectures', 'PyTorch']
}
for role, skills in ds_roles.items():
    print(f'{role:20s}: {', '.join(skills)}')
```

```python
# Sprint planning simulation
project_sprints = [
    {'sprint': 1, 'week': '1-2', 'goal': 'Data collection, exploration, and problem scoping'},
    {'sprint': 2, 'week': '3-4', 'goal': 'Feature engineering and baseline model'},
    {'sprint': 3, 'week': '5-6', 'goal': 'Model optimization and validation'},
    {'sprint': 4, 'week': '7',   'goal': 'Deployment and monitoring setup'},
    {'sprint': 5, 'week': '8',   'goal': 'Presentation and retrospective'}
]
print('Project Sprint Plan:')
print(f'{"Sprint":8} {"Weeks":8} {"Goal"}')
print('-' * 60)
for s in project_sprints:
    print(f'Sprint {s["sprint"]:2d}  Week {s["week"]:5s}  {s["goal"]}')
```

```python
# Technical interview prep: common DS questions
interview_topics = {
    'Statistics': ['P-value', 'CLT', 'Hypothesis testing', 'Confidence intervals', 'Bayesian vs Frequentist'],
    'ML Concepts': ['Bias-variance tradeoff', 'Regularization', 'Cross-validation', 'Overfitting'],
    'Algorithms': ['Decision tree splitting', 'Gradient descent', 'Backpropagation', 'K-means++'],
    'SQL': ['JOINs', 'Window functions', 'GROUP BY', 'CTEs', 'Query optimization'],
    'Case Studies': ['A/B testing design', 'Recommendation system', 'Fraud detection']
}
print('Interview Preparation Topics:')
for category, topics in interview_topics.items():
    print(f'  {category:15s}: {', '.join(topics)}')
```

```python
# Portfolio project checklist
checklist = [
    ('GitHub Repo',       'Clean code, README, requirements, reproducible'),
    ('EDA Notebook',      'Visualizations, insights, business context'),
    ('Model Pipeline',    'Preprocessing + model + evaluation in sklearn pipeline'),
    ('Results Summary',   'Clear metrics, comparison, limitations'),
    ('Presentation',      'Non-technical 5-10 slide deck'),
    ('Live Demo',         'Streamlit or FastAPI app (bonus)'),
    ('Blog Post',         'Medium/dev.to article explaining your work (bonus)')
]
print('Portfolio Checklist:')
for item, desc in checklist:
    print(f'  □ {item:20s}: {desc}')
```
