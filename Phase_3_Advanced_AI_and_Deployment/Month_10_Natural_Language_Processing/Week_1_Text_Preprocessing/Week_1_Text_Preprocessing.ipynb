# Week 1: Text Preprocessing

## Learning Objectives
- Clean raw text (lowercasing, punctuation removal, stopwords)
- Apply stemming and lemmatization
- Understand TF-IDF and Bag-of-Words
- Build a text preprocessing pipeline

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
from collections import Counter

try:
    import nltk
    nltk.download('stopwords', quiet=True)
    nltk.download('punkt', quiet=True)
    nltk.download('wordnet', quiet=True)
    from nltk.corpus import stopwords
    from nltk.stem import PorterStemmer, WordNetLemmatizer
    from nltk.tokenize import word_tokenize
    NLTK = True
except ImportError:
    NLTK = False
    print('NLTK not installed: pip install nltk')

sns.set_theme(style='whitegrid')
```

## 1. Text Cleaning

```python
sample_texts = [
    'Natural Language Processing (NLP) is AMAZING! It powers chatbots, search, & translation.',
    'Machine learning models can analyse text data very efficiently!!!',
    'Text cleaning removes noise: HTML tags, punctuation, numbers123, and extra spaces.'
]

def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)           # remove HTML
    text = re.sub(r'http\S+', '', text)         # remove URLs
    text = re.sub(r'[^a-z\s]', '', text)        # keep letters only
    text = re.sub(r'\s+', ' ', text).strip()
    return text

for t in sample_texts:
    print('Before:', t[:60])
    print('After: ', clean_text(t)[:60])
    print()
```

## 2. Tokenization, Stemming, and Lemmatization

```python
if NLTK:
    stop_words = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    text = 'The quick brown foxes are jumping over the lazy dogs running'
    tokens = word_tokenize(text.lower())
    tokens_ns = [t for t in tokens if t not in stop_words]
    stems = [stemmer.stem(t) for t in tokens_ns]
    lemmas = [lemmatizer.lemmatize(t) for t in tokens_ns]

    print('Tokens:  ', tokens_ns)
    print('Stems:   ', stems)
    print('Lemmas:  ', lemmas)
else:
    print('NLTK not available.')
```

## 3. Bag-of-Words and TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

corpus = [
    'Data science is the study of data',
    'Machine learning is a subset of data science',
    'Deep learning is a subset of machine learning',
    'Natural language processing is part of data science'
]

# BoW
bow = CountVectorizer()
X_bow = bow.fit_transform(corpus)
print('BoW vocab size:', len(bow.vocabulary_))

# TF-IDF
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(corpus)

# Compare
df_bow   = pd.DataFrame(X_bow.toarray(),   columns=bow.vocabulary_)
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.vocabulary_)
print('\nTF-IDF (first row):')
print(df_tfidf.iloc[0].sort_values(ascending=False).head(5).round(4))
```

## 4. Word Frequency Visualisation

```python
all_words = ' '.join(corpus).split()
freq = Counter(all_words).most_common(10)
words, counts = zip(*freq)

plt.figure(figsize=(8, 4))
plt.bar(words, counts, color='steelblue')
plt.title('Top 10 Word Frequencies')
plt.xticks(rotation=45)
plt.tight_layout(); plt.show()
```
