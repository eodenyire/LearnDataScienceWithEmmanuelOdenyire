# Week 5: Recurrent Neural Networks (RNNs)

## Learning Objectives
- Understand RNN architecture and vanishing gradients
- Build LSTM and GRU models
- Apply sequence models to time series prediction
- Compare LSTM and GRU performance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    TF = True
    print(f'TensorFlow {tf.__version__}')
except ImportError:
    TF = False
    print('TensorFlow not installed.')

rng = np.random.default_rng(42)
```

## 1. RNN Concepts

- RNNs maintain a **hidden state** updated at each time step
- Problem: **vanishing gradients** over long sequences
- **LSTM** adds cell state with gating (forget, input, output gates)
- **GRU** is a simplified version with fewer parameters

## 2. Sequence Data Preparation

```python
t = np.linspace(0, 100, 1000)
signal = np.sin(t) + 0.1 * rng.normal(size=1000)

window = 30
X_seq, y_seq = [], []
for i in range(len(signal) - window):
    X_seq.append(signal[i:i+window])
    y_seq.append(signal[i+window])
X_seq = np.array(X_seq)[..., np.newaxis]
y_seq = np.array(y_seq)

split = int(0.8 * len(X_seq))
X_tr, X_te = X_seq[:split], X_seq[split:]
y_tr, y_te = y_seq[:split], y_seq[split:]
print(f'Train: {X_tr.shape}, Test: {X_te.shape}')
```

## 3. LSTM and GRU

```python
if TF:
    from sklearn.metrics import mean_squared_error

    for name, rnn_layer in [('LSTM', layers.LSTM), ('GRU', layers.GRU)]:
        model = keras.Sequential([
            rnn_layer(64, return_sequences=True, input_shape=(window, 1)),
            rnn_layer(32),
            layers.Dense(1)
        ])
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_tr, y_tr, epochs=20, batch_size=32,
                   validation_split=0.1, verbose=0)
        rmse = np.sqrt(mean_squared_error(y_te, model.predict(X_te, verbose=0)))
        print(f'{name} RMSE: {rmse:.4f}')
else:
    print('TensorFlow not available.')
```
