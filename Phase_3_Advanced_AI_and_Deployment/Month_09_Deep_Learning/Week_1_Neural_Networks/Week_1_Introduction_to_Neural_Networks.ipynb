```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rng = np.random.default_rng(42)
sns.set_theme(style='whitegrid')
```

## 1. Activation Functions

```python
x = np.linspace(-5, 5, 200)
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

fig, axes = plt.subplots(1, 3, figsize=(13, 3))
for ax, y, name in zip(axes, [sigmoid, relu, tanh], ['Sigmoid', 'ReLU', 'Tanh']):
    ax.plot(x, y, linewidth=2, color='steelblue')
    ax.set_title(name)
    ax.axhline(0, color='gray', lw=0.8)
    ax.axvline(0, color='gray', lw=0.8)
plt.tight_layout(); plt.show()
```

## 2. A Simple Neural Network from Scratch

```python
class SimpleNN:
    def __init__(self, n_input, n_hidden, n_output, lr=0.01):
        self.lr = lr
        self.W1 = rng.normal(0, 0.5, (n_input, n_hidden))
        self.b1 = np.zeros(n_hidden)
        self.W2 = rng.normal(0, 0.5, (n_hidden, n_output))
        self.b2 = np.zeros(n_output)

    def sigmoid(self, z):        return 1 / (1 + np.exp(-z))
    def sigmoid_d(self, a):      return a * (1 - a)

    def forward(self, X):
        self.z1 = X @ self.W1 + self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = self.a1 @ self.W2 + self.b2
        self.a2 = self.sigmoid(self.z2)
        return self.a2

    def backward(self, X, y):
        m = X.shape[0]
        dz2 = self.a2 - y.reshape(-1, 1)
        dW2 = self.a1.T @ dz2 / m
        db2 = dz2.mean(axis=0)
        dz1 = (dz2 @ self.W2.T) * self.sigmoid_d(self.a1)
        dW1 = X.T @ dz1 / m
        db1 = dz1.mean(axis=0)
        self.W2 -= self.lr * dW2; self.b2 -= self.lr * db2
        self.W1 -= self.lr * dW1; self.b1 -= self.lr * db1

    def loss(self, y_pred, y):
        eps = 1e-9
        y = y.reshape(-1, 1)
        return (-y * np.log(y_pred + eps) - (1-y) * np.log(1-y_pred + eps)).mean()

print('SimpleNN class defined')
```

```python
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_c, y_c = make_circles(n_samples=500, noise=0.2, random_state=42)
X_c = StandardScaler().fit_transform(X_c)
X_tr, X_te, y_tr, y_te = train_test_split(X_c, y_c, test_size=0.2, random_state=42)

nn = SimpleNN(n_input=2, n_hidden=8, n_output=1, lr=0.5)
losses = []
for epoch in range(2000):
    preds = nn.forward(X_tr)
    losses.append(nn.loss(preds, y_tr))
    nn.backward(X_tr, y_tr)

acc = ((nn.forward(X_te) > 0.5).flatten() == y_te).mean()
print(f'Test Accuracy: {acc:.4f}')

plt.figure(figsize=(7, 3))
plt.plot(losses)
plt.title('Training Loss')
plt.xlabel('Epoch'); plt.ylabel('Binary Cross-Entropy')
plt.tight_layout(); plt.show()
```
