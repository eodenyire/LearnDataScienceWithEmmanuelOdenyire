# Week 4: Prophet Forecasting

## Learning Objectives
- Understand the Facebook Prophet model
- Fit Prophet to real-world time series
- Handle holidays and custom seasonalities
- Visualise uncertainty intervals

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    from prophet import Prophet
    PROPHET = True
except ImportError:
    try:
        from fbprophet import Prophet
        PROPHET = True
    except ImportError:
        PROPHET = False
        print('Prophet not installed: pip install prophet')

sns.set_theme(style='whitegrid')
```

## 1. Preparing Data for Prophet

Prophet requires a DataFrame with two columns:
- `ds` — datetime
- `y`  — the numeric value to forecast

```python
months = pd.date_range('1949-01-01', periods=144, freq='ME')
passengers = np.array([
    112,118,132,129,121,135,148,148,136,119,104,118,
    115,126,141,135,125,149,170,170,158,133,114,140,
    145,150,178,163,172,178,199,199,184,162,146,166,
    171,180,193,181,183,218,230,242,209,191,172,194,
    196,196,236,235,229,243,264,272,237,211,180,201,
    204,188,235,227,234,264,302,293,259,229,203,229,
    242,233,267,269,270,315,364,347,312,274,237,278,
    284,277,317,313,318,374,413,405,355,306,271,306,
    315,301,356,348,355,422,465,467,404,347,305,336,
    340,318,362,348,363,435,491,505,404,359,310,337,
    360,342,406,396,420,472,548,559,463,407,362,405,
    417,391,419,461,472,535,622,606,508,461,390,432
])
df = pd.DataFrame({'ds': months, 'y': np.log(passengers)})
print(df.head())
```

## 2. Fitting and Forecasting with Prophet

```python
if PROPHET:
    model = Prophet(yearly_seasonality=True, weekly_seasonality=False,
                    daily_seasonality=False, interval_width=0.95)
    model.fit(df)

    future = model.make_future_dataframe(periods=24, freq='ME')
    forecast = model.predict(future)

    fig = model.plot(forecast)
    plt.title('Prophet Forecast')
    plt.tight_layout(); plt.show()
else:
    print('Prophet not installed.')
```

```python
if PROPHET:
    fig2 = model.plot_components(forecast)
    plt.tight_layout(); plt.show()
else:
    print('Prophet not installed.')
```

## 3. Evaluating Prophet

```python
if PROPHET:
    from sklearn.metrics import mean_absolute_error, mean_squared_error
    actual = df['y'].values
    predicted = forecast.loc[:len(df)-1, 'yhat'].values
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae  = mean_absolute_error(actual, predicted)
    print(f'RMSE: {rmse:.4f}')
    print(f'MAE:  {mae:.4f}')
else:
    print('Prophet not installed.')
```
